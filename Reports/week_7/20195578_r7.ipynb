## Deflating the Peruvian economy

The academic work "The Monetary and Fiscal History of Peru, 1960-2017: Radical Experiments in Politics, Inflation and Stabilization" proposes the hyperinflation of Peru during the 1970s and 1980s as the result of the need for inflationary taxes in a regime of fiscal preponderance of monetary policy. Both the unavailability of the debt accumulation and the populist administration made the policy counterproductive in terms of price control and lax credit.

Let's know history to understand the present and thus make better decisions for the future: considering recent Peruvian history, the contribution of this article lies in the political and socioeconomic learning process regarding the association between the credibility of change of political regime in the 1990s and the change in public opinion that provided the right incentives to political actors. Both the respect for the independence of the central bank and the rejection of fiscal imprudence (characteristic features of the Peruvian economy and politics today) go back to some extent to the effect of the traumatic events of the 1980s. point in favor of this document is the confirmation and reaffirmation of the influence of institutional stability on macroeconomic stability, that is, if internal political struggles erode the institutional framework for the arbitration of disputes between the powers of the State, they can position in a critical situation for the macroeconomic stability of the country, which has cost so much to achieve.

Three strengths of this academic work stand out. The first corresponds to the figure of those who elaborated it: César Martinelli and Marco Vega. They have a Doctorate and are experts in the science of Economics, an aptitude that can allow them to prepare ideal academic papers in their field of study like this one, that is, that are distanced from biases, and provide true results. The second strength is that it is a work whose object of study spans decades, as the title reveals. This aspect means a better analysis in political and macroeconomic matters, which presents contrasts, correspondences, and particularities of the Peruvian economic reality before and after 1993, a year marked by the brake on the spending capacity of the legislative power and the delegation of monetary policy. to an autonomous central bank. The third and last strength consists in the validity of the period of independence of monetary policy and fiscal moderation, characteristics of the contemporary Peruvian economic model, to which reference is made.

The weakness of the work of both economists lies, on the one hand, in the fact that the publication of this document is relatively young, since it corresponds to the year 2018. This aspect implies that the figure of the opinions and/or constructive criticism expressed in academic works of other experts on the subject is extremely poor. On the other hand, this work, which collects data only up to 2018, distances itself to a certain extent from how the current economic reality is affected in the time after that year. At first glance, this aspect may seem like a limit, but if it is observed in depth, it can be considered as a valuable opportunity to encourage the development of future academic works that incorporate the analysis of how our economy is affected independently of monetary policy by factors, such as the covid 19 pandemic, natural phenomena, the political crisis manifested in the confrontation between the executive and the legislature, among others.

## Effects of the increase in Government Spending

```python
import numpy as np               
import matplotlib.pyplot as plt  
Y_size = 100
c = 20                 
b = 0.2               
m= 0.2            
T = 1                  
I_bar = 10             
G_bar = 20             
X_bar = 2              
h = 5                  
Y = np.arange(Y_size)  
def i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS
def Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS
i = i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y)
Y_size = 100
c = 20                 
b = 0.2                
m = 0.2            
T = 1           
I_bar = 10             
G_bar = 30              
X_bar = 2              
h = 5
def i_IS_1(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS
def Y_IS_1(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS
i_1 = i_IS_1(c, b, m, T, I_bar, G_bar, X_bar, h, Y)
y_max = np.max(i)
x_max = Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, 0)
v = [0, x_max, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Effects of the increase in Government Spending", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i, "k-",Label="IS_0")
ax.plot(Y, i_1, "b-",linestyle='dotted',Label="IS_1")
plt.axis(v)         
plt.legend()        
plt.show() 
```

In the proposed case, the curve expands, because public spending increases. Consequently, the aggregate demand becomes bigger than the equilibrium product. Therefore, it must increase to generate the perfect balance.

## The LM model

### The function of the real money supply

$M^{s}$$P^{-1}$ = $M_{0}^{s}$$P_{0}^{-1}$

### The real money demand function

$M^{d}$ = $kY$-$ji$

### The equilibrium equation

$M^{s}$$P^{-1}$ = $kY$-$ji$

### The equilibrium in the Monetary Market

```python
import numpy as np               
import matplotlib.pyplot as plt  
Y_size = 200
a = 40                 
b = 0.4                
alpha = 10              
beta  = 0.4           
T = 2                  
I_bar = 20             
G_bar = 16              
X_bar = 4            
d = 10                  
Y = np.arange(Y_size)  
def i_IS(a, alpha, b, beta, T, I_bar, G_bar, X_bar, d, Y):
    i_IS = ((a-alpha)-(b-beta)*T + I_bar + G_bar + X_bar - (1-b+beta)*Y)/d
    return i_IS
def Y_IS(a, alpha, b, beta, T, I_bar, G_bar, X_bar, d, i):
    Y_IS = ((a-alpha)-(b-beta)*T + I_bar + G_bar + X_bar - d*i)/(1-b+beta)
    return Y_IS
i = i_IS(a, alpha, b, beta, T, I_bar, G_bar, X_bar, d, Y)
y_max = np.max(i)
x_max = Y_IS(a, alpha, b, beta, T, I_bar, G_bar, X_bar, d, 0)
v = [0, x_max, 0, y_max]                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="The equilibrium in the Monetary Market", xlabel=r'real money supply', ylabel=r'interest rate')
ax.plot(Y, i, "k-")
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())   
plt.axis(v)  
plt.vlines(x=35, ymin=0, ymax=90, colors='green', ls=':', lw=2, label='vline_single - partial height')
plt.show() 
```

```python
"1|IMPORT PACKAGES"
import numpy as np               
import matplotlib.pyplot as plt  
Y_size = 100
c1 = 40              
c2 = 9                
c3 = 9                
Ms = 350             
P  = 20                
Y = np.arange(Y_size)  
def i_LM(c1, c2, c3, Ms, P, Y):
    i_LM = (c1 - Ms/P)/c3 + c2/c3*Y
    return i_LM
i = i_LM(c1, c2, c3, Ms, P, Y)
y_max = np.max(i)
v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="LM SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i, "k-")
plt.axis(v)                                     
plt.show()
```
