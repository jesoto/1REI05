# 2.1. Effects of the increase in Government Spending

```python
import numpy as np               
import matplotlib.pyplot as plt  

"2|DEFINE PARAMETERS AND ARRAYS of IS_0"
# Parameters
Y_size = 100
C_0 = 20                 # Autonomous consumption
b = 0.2                  # Marginal propensity to consume
alpha = 5                # Autonomous imports
m  = 0.2                 # Marginal propensity to import
T = 1                    # Taxes
I_0 = 10                 # Investment intercept (when i = 0)
G_0 = 8                  # Government spending
X_0 = 2                  # Exports (given)
h = 5                    # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE IS_0-SCHEDULE"
def r_IS_0(C_0, alpha, b, m, T, I_0, G_0, X_0, h, Y):
    r_IS_0 = ((C_0-alpha)-(b-m)*T + I_0 + G_0 + X_0 - (1-b+m)*Y)/h
    return r_IS_0

def Y_IS_0(a, alpha, b, m, T, I_0, G_0, X_0, d, r_0):
    Y_IS_0 = ((C_0-alpha)-(b-m)*T + I_0 + G_0 + X_0 - h*r_0)/(1-b+m)
    return Y_IS_0

r_0 = r_IS_0(C_0, alpha, b, m, T, I_0, G_0, X_0, h, Y)

"2|DEFINE PARAMETERS AND ARRAYS of IS_1"
# Parameters
Y_size = 100
C_1 = 20                   # Autonomous consumption
b_1 = 0.15                  # Marginal propensity to consume
alpha_1 = 5                # Autonomous imports
m_1  = 0.2                 # Marginal propensity to import
T_1 = 1                    # Taxes
I_1 = 10                   # Investment intercept (when i = 0)
G_1 = 16                    # Government spending
X_1 = 2                    # Exports (given)
h_1 = 5                    # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE IS_1-SCHEDULE"
def r_IS_1(C_1, alpha_1, b_1, m_1, T_1, I_1, G_1, X_1, h_1, Y):
    r_IS_1 = ((C_1-alpha_1)-(b_1-m_1)*T_1 + I_1 + G_1 + X_1 - (1-b_1+m_1)*Y)/h_1
    return r_IS_1

def Y_IS_1(C_1, alpha_1, b_1, m_1, T_1, I_1, G_1, X_1, d_1, r_1):
    Y_IS_1 = ((C_1-alpha_1)-(b_1-m_1)*T_1 + I_1 + G_1 + X_1 - h_1*r_1)/(1-b_1+m_1)
    return Y_IS_1

r_1 = r_IS_1(C_1, alpha_1, b_1, m_1, T_1, I_1, G_1, X_1, h_1, Y)

"4|PLOT THE IS-SCHEDULE"
y_max = np.max(r_0)
x_max = Y_IS_0(C_0, alpha, b, m, T, I_0, G_0, X_0, h, 0)
                        
v = [0, x_max, 0, y_max]
fig, ax = plt.subplots(figsize=(14, 12))
ax.set(title="Effects of the increase in Government Spending", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r_0,"k-", scalex = T, scaley = T, label = "IS_0")
ax.plot(Y, r_1, "k-",scalex= T, scaley = T, color = 'green',linestyle="--", label = "IS_1")

ax.legend()
plt.text(18, 4, '∆Go', fontsize=20, color='black')
plt.text(18, 3.7, '→', fontsize=25, color='black')
plt.axis(v)                                     
plt.show()
```

In the graph of effects of the increase in Government Spending, there is a parallel shift to the right. Which allows us to understand that there has been some change in the intercept, due to the increase in the autonomous component of Government Expenditure of aggregate demand. It is assumed to be an increase by the positive shift to the right.

When there is an increase in Government Spending -> Aggregate demand increases -> This generates excess demand -> Giving the parallel shift of the curve to the left. This displacement is represented by the line IS_1

# 2.2. Effects of an increase in the tax rate

```python
"2|DEFINE PARAMETERS AND ARRAYS"
# Parameters
Y_size = 100
c = 20                 # Autonomous consumption
b = 0.20                # Marginal propensity to consume
             # Autonomous imports
m = 0.2            # Marginal propensity to import
T = 1                  # Taxes
I_bar = 10             # Investment intercept (when i = 0)
G_bar = 8              # Government spending
X_bar = 2              # Exports (given)
h = 5                  # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE IS-SCHEDULE"
def i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS

def Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS


i = i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y)
#--------------------------------------------------------------
Y_size = 100
c = 20                 # Autonomous consumption
b = 0.02                # Marginal propensity to consume
             # Autonomous imports
m = 0.2            # Marginal propensity to import
T = 1                  # Taxes
I_bar = 10             # Investment intercept (when i = 0)
G_bar = 8              # Government spending
X_bar = 2              # Exports (given)
h = 5  



def i_IS_1(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS

def Y_IS_1(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS

i_1 = i_IS_1(c, b, m, T, I_bar, G_bar, X_bar, h, Y)

"4|PLOT THE IS-SCHEDULE"
y_max = np.max(i)
x_max = Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, 0)

v = [0, x_max, 0, y_max]                        # Set the axes range
fig, ax = plt.subplots(figsize=(14, 12))

ax.set(title="Effects of an increase in the tax rate", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i, "k-",label="IS_0")
ax.plot(Y, i_1, "r-",color = 'blue',linestyle='dashed', label="IS_1")
ax.yaxis.set_major_locator(plt.NullLocator())   # Hide ticks
ax.xaxis.set_major_locator(plt.NullLocator())   # Hide ticks

ax.legend ()
plt.text(28, 2, '∆t', fontsize=20, color='black')
plt.text(28, 1.7, '←', fontsize=25, color='black')
plt.axis(v)                 
plt.show()
```

In this second case, the change produced is affecting the slope of the curve. Then, the autonomous variables that must have undergone some changes are:
$$ \ (b-m)(1-t)Y \ $$

In the case of this graph, there is an increase in the tax rate. When the government increases the tax rate ($$ \ t \ $$), the slope of the IS curve decreases, since there is a non-parallel rotation of the curve to the left. That is, the rotation occurs in a clockwise direction.

This slope reduction occurs because when there is an increase in the tax rate, the size of the multiplier is reduced. As such, the slope of aggregate demand. Although the interest rate is maintained due to the reduction in the multiplier, the equilibrium income will be lower. The change produced in the IS curve is represented by the green line.

However, it should be considered that this increase in the tax rate actually ends up producing two changes in the IS curve. This is because the increase in the tax rate affects disposable income.
So, it has repercussions on two elements of the IS equation, aggregate demand, and foreign savings. In the first case, of aggregate demand, the increase in the tax rate will reduce consumption. Which consequently reduces aggregate demand and shifts to the left.

On the other hand, when it affects foreign savings, it generates a positive effect. Since this will allow the reduction of imports, as such, there would be an increase in the curve, and a shift to the right. Of these two effects, the one that will predominate as evidenced in the graph with the blue line is the first effect that is addressed. Since the number of people who import is less than the entire population that you consume.

# 3. THE LM SCHEDULE (LIQUIDITY PREFERENCE-MONEY SUPPLY)

Real Money Offer Feature

\begin{equation}
M^s P^-1 = M^s_0 P^-1_0
\end{equation}

The real money demand function

\begin{equation}
 M^d = k Y - ji
\end{equation}

The equilibrium equation

\begin{equation}
M^s P^-1  = k Y - ji
\end{equation}


The Equilibrium LM Curve in the Money Market

\begin{align}
    \ r &= \frac {-1}{j}\ \frac {M^S_0}{P_0}\ + \frac {k}{j}Y\
\end{align}

# 3.1. Equilibrium in the monetary market

```python
"1|DEFINE PARAMETERS AND ARRAYS of LM_0"
# Parameters
Md_size = 100
k = 0.8              # Marginal propensity to demand money
j = 0.8                # Elasticity demand for money
Y = 100                # Nominal interest rate = real interest rate
# Arrays
Md = np.arange(Md_size) 

"2|DEFINE AND POPULATE THE LM_0-SCHEDULE"
def r(k, j, Y, Md):
    r = (k*Y) - j*(Md)
    return r

r = r(k, j, Y, Md)

y_max = np.max(r)

v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(14, 12))
ax.set(title="Monetary Market Equlibrium", xlabel=r'$M^s$/P', ylabel=r'r')
ax.plot(Md, r, "k-", scalex = T, scaley = T, label = "Md")
ax.vlines(x = 40, ymin = 0, ymax = 100, label = "$M^s$/$P_0$")
plt.plot(40, 48, marker="o", color="black")
ax.legend()
plt.axis(v)                                    
plt.show()
```
