The authors wonder about the reasons that caused the chronic inflation in Peru between 1970 and 1980. In their study, they find that these would be the result of the need for inflationary taxation in a regime of fiscal dominance of monetary policy.The first strength of the text is found in the method that they used to demonstrate this conclusion. Throughout his proposal, a historical analysis of the policies that were carried out in different periods such as the governments of Belaúnde, García and Fujimori is presented; while, a quantitative analysis of the economic data of each period was presented to explain the results of such policies. Thus, the authors also show how the called “fujishock” ended up regularizing the economy in the medium term.  
Furthermore, in the analysis they propose a "historical learning" by society that led to popular rejection of state intervention in economics. This is one of the advances made by the authors on the subject, since they add a social variable related to perceptions to the study of the subject. This is important since it is often forgotten that macroeconomic decisions are the result of political processes that are at the same time a process of social conflict.  
Moreover, this "historical learning" can serve both to criticize the text itself and to propose future research on the subject. The results of a kind of "hyperinflation trauma" based on this learning, would also serve to check whether in the long term the population has learned and requested other principles of the state-society-economy relationship. From the text we know that the population rejected state interventionism, but in recent years, multiple social conflicts that stem from the current economic model and the country's political class are known. Based on this, a change could be seen in this "historical learning".

**Investment and Save - IS Curve**

```python
import numpy as np               
import matplotlib.pyplot as plt  

# Parameters
Y_size = 100
Co = 20             # Autonomous consumption
b = 0.2             # Marginal propensity to consume
m  = 0.2            # Marginal propensity to import
t = 1               # Taxes
Io = 10             # Investment intercept (when i = 0)
Go = 8              # Government spending
Xo = 2              # Exports (given)
h = 5               # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

# DEFINE AND POPULATE THE IS-SCHEDULE
def i_IS(Co, b, m, t, Io, Go, Xo, h, Y):
    i_IS = ((Co)-(b-m)*t + Io + Go + Xo - (1-b+m)*Y)/h
    return i_IS

def Y_IS(Co, b, m, t, Io, Go, Xo, h, r):
    Y_IS =  ((Co)-(b-m)*t + Io + Go + Xo - h*r)/(1-b+m)
    return Y_IS

r = i_IS(Co, b, m, t, Io, Go, Xo, h, Y)

# PLOT THE IS-SCHEDULE
y_max = np.max(r)
x_max = Y_IS(Co, b, m, t, Io, Go, Xo, h, 0)

v = [0, x_max, 0, y_max]                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE - IS_0", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r, "k-")
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())   
plt.axis(v)                                     
plt.show()
```
