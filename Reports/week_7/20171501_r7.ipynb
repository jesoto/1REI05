# 2. IS Curve

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
Y_size = 100
C = 10                 # Autonomous consumption
M = 0.2                # Marginal propensity to consume
a = 5              # Autonomous imports
b  = 0.2            # Marginal propensity to import
T = 0.1                  # Taxes
Io = 10             # Investment intercept (when i = 0)
Go = 10              # Government spending
Xo = 2              # Exports (given)
d = 5                  # Investment slope wrt to i

Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE IS-SCHEDULE"
def i_IS(C, a, M, b, T, Io, Go, Xo, d, Y):
    i_IS = ((C-a)-(M-b)*T + Io + Go + Xo - (1-M+b)*Y)/d
    return i_IS

def Y_IS(C, a, M, b, T, Io, Go, Xo, d, i):
    Y_IS = ((C-a)-(M-b)*T + Io + Go + Xo - d*i)/(1-M+b)
    return Y_IS

i = i_IS(C, a, M, b, T, Io, Go, Xo, d, Y)

"4|PLOT THE IS-SCHEDULE"
y_max = np.max(i)
x_max = Y_IS(C, a, M, b, T, Io, Go, Xo, d, 0)

v = [0, x_max, 0, y_max]                        # Set the axes range
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i, "k-")
ax.yaxis.set_major_locator(plt.NullLocator())   # Hide ticks
ax.xaxis.set_major_locator(plt.NullLocator())   # Hide ticks
plt.axis(v)                                     # Use 'v' as the axes range
plt.show()
```

# 2.1 Increase Goverment Spending (Go)

```python
Y_size = 100
C = 10                 # Autonomous consumption
M = 0.2                # Marginal propensity to consume
a = 5              # Autonomous imports
b  = 0.2            # Marginal propensity to import
T = 0.1                  # Taxes
Io = 10             # Investment intercept (when i = 0)
Go = 20              # Government spending
Xo = 2              # Exports (given)
d = 5                  # Investment slope wrt to i

Y = np.arange(Y_size)  # Array values of Y

def i_IS_G(C, a, M, b, T, Io, Go, Xo, d, i):
    i_IS=((C-a)-(M-b)*T + Io + Go + Xo - (1-M+b)*Y)/d
    return i_IS

def Y_IS_G(C, a, M, b, T, Io, Go, Xo, d, i):
    Y_IS=((C-a)-(M-b)*T + Io + Go + Xo - d*i)/(1-M+b)
    return Y_IS

i_G = i_IS_G(C, a, M, b, T, Io, Go, Xo, d, Y)

y_max = np.max(i)
x_max = Y_IS(C, a, M, b, T, Io, Go, Xo, d, 0)

v = [0, x_max, 0, y_max]                        # Set the axes range
fig, ax = plt.subplots(figsize=(16, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i, "k-", label='IS')
ax.plot(Y, i_G, "k-",label='IS_G',linestyle='--')

ax.legend()
plt.text(47, 0.-3, 'ΔGo', fontsize=13, color='black')
plt.text(48, 0.-4, '→', fontsize=27, color='gray')
plt.show()
```

In this case the cruce moves to the right, it's because in the case of Goverment spending, the increase makes the whole higher.

# 3.2 LM curve

```python
"1|IMPORT PACKAGES"
import numpy as np               # Package for scientific computing with Python
import matplotlib.pyplot as plt  # Matplotlib is a 2D plotting library

"2|DEFINE PARAMETERS AND ARRAYS"
# Parameters
Y_size = 100
c1 = 1000              # Precautionary money demand
c2 = 10                # Transaction money demand
c3 = 10                # Speculation money demand
Ms = 20000             # Nominal money supply
P  = 20                # Price level
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE LM-SCHEDULE"
def i_LM(c1, c2, c3, Ms, P, Y):
    i_LM = (c1 - Ms/P)/c3 + c2/c3*Y
    return i_LM

i = i_LM(c1, c2, c3, Ms, P, Y)
    
"4|PLOT THE IS-SCHEDULE"
y_max = np.max(i)

v = [0, Y_size, 0, y_max]                       # Set the axes range
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="LM SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i, "k-")
plt.axis(v)                                     # Use 'v' as the axes range
plt.show()
```
