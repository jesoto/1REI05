## I.  REPORT

The objective of this article is to answer the question of why there was chronic inflation in Peru between the 1970s and 1980s. To answer this question, the authors make a historical interpretation of the events of those years through a monetarist approach. Likewise, they carry out a quantitative analysis through three measurements: growth accounting, fiscal accounting and analysis of the policies adopted (institutional perspective). First, in the growth accounting part, what the authors did was disaggregate the changes in GDP per worker. It was observed that the period of stagflation coincided with a considerable slowdown in productivity, which shows a poor distribution of resources as a result of the radical reforms of the 1970s. Second, the exercise of fiscal accounting made it possible to disaggregate State financing. The conclusion reached was that fiscal deficits were financed through inflationary taxes and the accumulation of external public debt. Likewise, it is shown that the BCRP was the main source of internal financing and that its operations were marked by a complete lack of transparency. Finally, the authors conclude that the policies adopted by the government at the time were counterproductive and did not help or even aggravated the situation. 

As for strengths, several can be found in the article. On the one hand, it is remarkable that the research has combined both historical and quantitative analysis. Also, recognizing the particular characteristics of our country, the authors emphasize the institutional and political weakness of Peru and its role in the period of hyperinflation. Finally, it is noteworthy that the authors have mentioned the demand for redistribution, public services and public investment as the origins of the fiscal difficulties, since, although the monetary and fiscal policies are questionable, it is recognized that there was a need behind that had to solve (shows the social part of the economy). 

As for the weaknesses, in the part of the quantitative analysis two can be mentioned. First, as the authors themselves acknowledge, some data are not the best. This was the case of the fiscal accounting analysis, where the data available for public debt at a constant real interest rate are not reliable. Second, the difficulty in estimating data due to the high volatility of the period. For example, the authors mention that, due to high inflation, the measurement of relative prices and real exchange rates became "a daunting task." Although the researchers decided to either ignore certain variables or set them equal to zero to proceed with the analysis, the two points mentioned above can be considered a weakness. On the analysis side, it can be said that a weakness of the research is the little importance given to internal shocks. Although the research determines that fiscal policy, monetary policy, institutional weakness and the inefficient distribution of resources had a direct impact on hyperinflation, variables such as the internal armed conflict, natural disasters, massive migration and the centralization of the time could have influenced production, but were not evaluated in depth. 

In recent years, there has been a tendency in Peru to question the economic model that has been implemented since the 1990s. In that sense, one of the reasons why it may have been decided to publish this article could be to justify the continuity of the country's monetary and fiscal policy. The macroeconomic strengths that characterize Peru today (fiscal solvency, large reserves and independent BCRP) were built after the period of hyperinflation that it experienced. Therefore, what this paper does is make clear the benefits that an "adequate" fiscal and monetary policy brought to the country. Likewise, another contribution of this work was to include the role of expectations and that of public opinion/attitude in the adoption of a different economic policy. The latter is noteworthy because economic research is normally carried out on the basis of material and quantifiable variables, leaving aside social variables such as the perception and discomfort of society, variables that, in the Peruvian case, determined what type of policies were applied.

The first next step should be to carry out a comparative investigation between the fiscal and monetary policies of that time and those implemented in the last 4 years. This is because the country has gone through several crises of various kinds and it is necessary to identify the changes, continuities, failures and strengths of the Peruvian economic policy. Another valuable next step would be to carry out research to delve into the role of citizen support in economic policy, specifically to what extent public attitude can be decisive for economic stability. Finally, it would be interesting to carry out an investigation as an experiment where Peru is placed in a situation of excessive fiscal deficit, negative external shocks and inflation to analyze what our capacity would be to get out of the crisis with our current characteristics.

### Effects of Increased Government Spending

To begin with, what happens in this case is a change in one of the autonomous components of aggregate demand (Goverment spending). Therefore, what you see on the graph is a parallel shift of the IS curve. This means that there has been an increase in goverment spending. --> The aggregate demand increase, which generates excess demand --> Giving the rise to a shift in the curve.

In the following two cases, what happens are changes in the slope of the IS curve. 

Where the slope of the IS curve $Î²_1$ = 1 - ($b$ - $m$) (1 - $t$)

### Effects of an increase in the Tax Rate

```python
"2|DEFINE PARAMETERS AND ARRAYS of IS_1"
# Parameters
Y_size = 100
C_1 = 20                   # Autonomous consumption
b_1 = 0.2                  # Marginal propensity to consume
alpha_1 = 5                # Autonomous imports
m_1  = 0.2                 # Marginal propensity to import
T_1 = 5                    # Taxes
I_1 = 10                   # Investment intercept (when i = 0)
G_1 = 8                    # Government spending
X_1 = 2                    # Exports (given)
h_1 = 5                    # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE IS_1-SCHEDULE"
def r_IS_1(C_1, alpha_1, b_1, m_1, T_1, I_1, G_1, X_1, h_1, Y):
    r_IS_1 = ((C_1-alpha_1)-(b_1-m_1)*T_1 + I_1 + G_1 + X_1 - (1-b_1+m_1)*Y)/h_1
    return r_IS_1

def Y_IS_1(C_1, alpha_1, b_1, m_1, T_1, I_1, G_1, X_1, d_1, r_1):
    Y_IS_1 = ((C_1-alpha_1)-(b_1-m_1)*T_1 + I_1 + G_1 + X_1 - h_1*r_1)/(1-b_1+m_1)
    return Y_IS_1

r_1 = r_IS_1(C_1, alpha_1, b_1, m_1, T_1, I_1, G_1, X_1, h_1, Y)
```

```python
"4|PLOT THE IS-SCHEDULE"
y_max = np.max(r_0)
x_max = Y_IS_0(C_0, alpha, b, m, T, I_0, G_0, X_0, h, 0)
                        
v = [0, x_max, 0, y_max]
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r_0,"k-", scalex = T, scaley = T, label = "IS_0")
ax.plot(Y, r_1, "k-",scalex= T, scaley = T, color = 'red',linestyle="--", label = "IS_1")
ax.legend()
plt.axis(v)                                     
plt.show()
```

In this second case, what happens is an increase in the Tax Rate:

- If the government increases the tax rate ($t$), the slope of the IS curve increases. --> Which causes a rotation of the curve to the left.
- This is because the increase in the tax rate reduces $Î²_1$.
- However, from a deductive approach (and disaggregating the components of the curve), it must be taken into account that an increase of $t$ affects disposable income ($y^d$), which affects two components: private saving and foreign savings. On the one hand, the increase in $t$ will reduce consumption, which reduces aggregate demand, generating a contraction of the curve. But, on the other hand, increasing $t$ generates a positive effect, since it reduces imports, which would generate an expansion of the curve. In the end, it will depend on the magnitudes which effect is going to be superimposed on the other.

### Equilibrium in the monetary market

```python
"1|DEFINE PARAMETERS AND ARRAYS of LM_0"
# Parameters
Md_size = 100
k = 0.3              # Marginal propensity to demand money
j = 0.5                # Elasticity demand for money
Y = 100                # Nominal interest rate = real interest rate
# Arrays
Md = np.arange(Md_size) 

"2|DEFINE AND POPULATE THE LM_0-SCHEDULE"
def r(k, j, Y, Md):
    r = (k*Y) - j*(Md)
    return r

r = r(k, j, Y, Md)
```

### LM Curve

```python
"1|DEFINE PARAMETERS AND ARRAYS"
# Parameters
Y_size = 100
c1 = 1000              # Precautionary money demand
k = 10                # Transaction money demand
j = 10                # Speculation money demand
Ms = 20000             # Nominal money supply
P  = 20                # Price level
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"2|DEFINE AND POPULATE THE LM-SCHEDULE"
def i_LM(c1, k, j, Ms, P, Y):
    i_LM = (c1 - Ms/P)/j + k/j*Y
    return i_LM

r_2 = i_LM(c1, k, j, Ms, P, Y)
    
"3|PLOT THE IS-SCHEDULE"
y_max = np.max(r_2)

v = [0, Y_size, 0, y_max]                       # Set the axes range
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="LM SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r_2, "k-", scalex = T, scaley = T, color = "green", label = "LM")
plt.axis(v)                                    
plt.show()
```
