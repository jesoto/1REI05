```python
import matplotlib.pyplot as plt
import numpy as np               
import matplotlib.pyplot as plt
```

```python
# Parameters
Y_size = 100
b = 0.4                 # Autonomous consumption
m = 0.2                  # Marginal propensity to consume
t = 0.3                # Autonomous imports
Co  = 20                 # Marginal propensity to import                  # Taxes
Io = 10                 # Investment intercept (when i = 0)
Go = 8                  # Government spending
Xo = 2                  # Exports (given)
h = 5                    # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y
```

```python
"3|DEFINE AND POPULATE THE IS_0-SCHEDULE"
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
r
```

```python
"4|PLOT THE IS-SCHEDULE"
y_max = np.max(r)
x_max = Y_IS(b, m, t, Co, Io, Go, Xo, h, 0)
                        
v = [0, x_max, 0, y_max]
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Effects of Increased Government Spending", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r,"k-", scalex = t, scaley = t, label = "IS0")
ax.plot(Y, r_G, "k-",scalex= t, scaley = t, color = 'green',linestyle="--", label = "IS1")
ax.legend()


plt.text(47, 0.07, '∆Go', fontsize=12, color='black')
plt.text(49, -0.45, '→', fontsize=15, color='grey')
plt.text(99, -8.7, 'IS0', fontsize=13)
plt.text(99, -7.3, 'IS1', fontsize=13)

plt.show()
```

To build the LM schedule we need money supply (MS) and money demand (MD). Money supply is an exogenous variable (e.g. a policy decision of the central bank). Following Keynes, the three sources of money demand are (1) precautionary reasons (e.g. save for a rainy-day), (2) to perform transaction and (3) for speculation reasons. Expressing MS and MD in real terms:

$$ M^o = M^d $$

In real terms is expressed like this:

$$ \frac{M^s}{P} = kY - ji $$

Where:

$$ \M^s $$  = Nominal money supply 

$$ P $$  = Price level 

$$ k $$ = Transaction money demand 

$$ j $$ = Speculation money demand 


$$ i $$ = Nominal interest rate 

So the equation is:

$$ \frac{M^s}{P} = kY - jr $$

#### Equilibrium LM Curve in the Money Market

The equation is:

$$ jr = \frac{M^s_o}{P_o} - kY $$

$$ r = -\frac{1}{j} \frac{M^s_o}{P_o} + \frac{k}{j}Y $$

```python
# Parameters
Y_size = 100

k = 10                # Transaction money demand
j = 10                # Speculation money demand
Ms = 20000             # Nominal money supply
P  = 20                # Price level

Y = np.arange(Y_size)
```

```python
def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
i
```

```python
y_max = np.max(i)

v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Equilibrium LM Curve in the Money Market", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i, "b-", label="LM", color = 'black')
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.text(99, -8.7, 'LM', fontsize=13)

ax.legend()
plt.show()
```
