3. THE LM SCHEDULE (LIQUIDITY PREFERENCE-MONEY SUPPLY)

*Real money supply:
$$\ M^s/P = M_0^s/P \$$

*Real money demand:
$$\ M^d = kY − ji \$$

*Equilibrium:
$$\ M^s = M^d \$$

$$\ M^s/P = kY − ji \$$

$$\ M^s =  P(kY − ji) \$$

*The LM schedule 

$$\ r = − (1/j)(M_0^𝑠/P_0) + (k/j) Y \$$

Where: −(1/j)(M_0^𝑠/P_0) is the intercept and +(k/j) is the slope
