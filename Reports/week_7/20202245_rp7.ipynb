The article tries to explain the monetary and fiscal history of Peru in the years 1960 to 2017. It exposes periods of increasing inflation until hyperinflation in the mid-1980s. These were accompanied by a fall in GDP per capita. Given this, the author provides a historical interpretation of these events from a monetarist approach. He shows that the inflation of years prior to economic stabilization was the result of the fiscal need to count on inflationary taxes in a regime of fiscal predominance in monetary policy. There are recurrent cyclical fiscal deficits until 1990.
The years of economic stabilization correspond to periods of monetary policy independence and fiscal moderation.
The author includes two measurement exercises. A growth accounting approach, disaggregating GDP per worker into various components. The results of this indicate that a considerable slowdown in productivity coincides with the period of stagflation. The slowdown in productivity provides evidence that there was a misallocation of resources as a result of the policies implemented. Including an extensive intervention of the State in the economy and the interruption of the fiscal policies applied since 1990.
Another fiscal accounting exercise made it possible to disaggregate government financing into its different sources.
He finally exposes his conclusions and suggested challenges according to the Peruvian experience.

Among the points in favor of the article is that it provides a mathematical and quantitative framework that provides a different point of view to what is already known historically.
The second positive feature is the historical framework that the article has. Try to answer the research question with historical tools that give the article more interest and capture curiosity. This brings me to the next positive feature, the years covered by the investigation. These chosen years are key to explain one of the biggest crises that the country has gone through. Peruvians know the precarious conditions experienced in those years but few know the methodological reasons why we came to it. In this sense, the years covered are also an advantage in the article since it attracts the public to read it.

Some of the disadvantages found lie in the exclusion of GDP per capita per worker in some of the most important variables. Although, it is part of the author's methodology, it can make many significant differences by involving and even changing the results of the investigation. And as a second negative characteristic is the lack of inclusion of the methodology of measurement exercises in the main points of the article, in many parts of the text it goes unnoticed and should have more importance.

The main contribution of the article to the research question is the quantitative methodological version of the history known by most Peruvians in the years of hyperinflation. With the help of this article, people will realize from another point of view what policies and activities led to the crisis in Peru in the years mentioned.

Some steps to advance in the investigation lie in the first place, updating the article for current years and with current crises that the country is going through. Add a comparative analysis of the policies implemented during the years of high inflation and current inflation years as well as the policies implemented.
```python
import numpy as np               
import matplotlib.pyplot as plt 

Y_size = 100
Y   = np.arange(Y_size)    
# IS Parameters
a = 100                    
b = 0.2                    
alpha = 5                  
beta  = 0.15               
T = 1                       
I = 10                     
G = 20                     
X = 5                     
d = 2                       


def i_IS(a, alpha, b, beta, T, I, G, X, d, Y):
    i_IS = ((a-alpha)-(b-beta)*T + I + G + X - (1-b+beta)*Y)/d
    return i_IS

def i_LM(c1, c2, c3, Ms, P, Y):
    i_LM = (c1 - Ms/P)/c3 + c2/c3*Y
    return i_LM

iIS = i_IS(a, alpha, b, beta, T, I, G, X, d, Y)


Y_star1 = ((a-alpha) - (b-beta)*T + I + G + X)/d
Y_star2 = (1/c3) * (c1 - Ms/P)
Y_star3 = (1 - b + beta)/d + (c2/c3)
Y_star  = (Y_star1 - Y_star2)/Y_star3

i_star1 = (1/c3)*(c1 - Ms/P)
i_star2 = (c2/c3)*Y_star
i_star  = i_star1 + i_star2


y_max = np.max(iIS)
v = [0, Y_size, 0, y_max]                     
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Equilibrio en el mercado de Dinero", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, iIS, "b-")
ax.yaxis.set_major_locator(plt.NullLocator())  
ax.xaxis.set_major_locator(plt.NullLocator())  
plt.plot(Y_star, i_star, 'ko')                
plt.axvline(x=Y_star, ls=':', color='k')
plt.axhline(y=i_star, ls=':', color='k')
plt.text(95, 22, "L(Yo)", color='blue')
plt.text(1, 36, "ro^e", color='purple')
plt.text(58, 1, "Mo^5\Po", color='green')
plt.axis(v)                                   
plt.show()
```
