Report of the article: “The colonial origins of comparative level”

This article seeks to explain the current differences in GDP per capita around the world by pointing out the different types of institutions inherited from the colonial era. The authors believe that the difference in mortality index had an impact on the settlement, which affected the type of early institutions formed in the different colonies and that these institutions are the basis of the current institutions. In this way, they establish that ex-colonies like the United States, New Zeeland, Canada and Australia, because of less mortality index and “Neo-Europe” early institutions, have nowadays more GDP per capita than others countries in Latin America, Africa and Asia that had more mortality index and early “extractive institutions”. 

One important strength of this article is the amount of data that the authors used to construct their investigation. They used a sample of 64 countries to analyze each aspect of the investigation: mortality and settlements, types of colonization and institutional persistence. In this way, they used mortality rates expected by first European in colonies, protection against “risk of expropriation” index from Political Risk Services as a proxy for institutions and a regression of current performance on current institutions. In addition, they used different statistics tables and graphs to demonstration the relation between the mortality index, types of institutions and even GDP per capita today. Furthermore, the authors used other previous similar investigations and historical data to analyze the information. 

Another important strength is the consideration, and even test, of other many possible omitted factor that could affect the results. The authors used previous similar investigations to test if other factors like latitude, climate, current disease environment, religion, natural resources, soil quality, ethnolinguistic fragmentation, current racial composition, and others could have a significant effect on the results the found; they found that neither of them were mainly significant. In this way, both, the use of different sources of information and consideration of many factor, give this investigation huge validation, reliability and trust in the results. 

However, this article also presents some weaknesses. Because of lack of data in the 64 countries of the sample, the authors had to recreate the information of some countries, which could have included some biased or mistaken data. In addition, I consider that the investigation, because of the use of the big data of 64 different countries, could fall into generalities. In other words, they grouped the 64 countries in two main groups to prove their point in general, but I believe there are many other factors that could determine the actual differences in GDP per capita if we follow the history of each particular country. 

Nevertheless, despite of the weaknesses of this article, I believe this investigation offers valuable information to understand why nowadays there are big differences in GDP per capita between countries even though they all share a colonial background. Furthermore, as the authors explained, currently there are no other investigations with the same approach, so this investigation is important because it puts new topics and theories to discuss and debate on the table. Furthermore, I believe that the economic models, statistics graphics, data recollected and analysis provided by the author would be used as a valuable base to new researches and investigations in this topic.

In this way, and considering the weaknesses mentioned before, I consider a valuable next step to continue with this investigation is to select different important examples of ex-colonies to examine deeper the history after the independence and the development of their institutions. This in order to determine which other factors, as corruption, natural disasters, economic crisis, and others, could have contributed to the current difference in GDP per capita between ex-colonies. For example, in 2006, Cavalcanti, Magalhães and Tavares, used this investigation as a base to analyze the impact of European mortality rate in the GDP per capita, ratio of private credit to GDP and ratio of investment to GDP in Brazil, Argentina and Chile specifically.  Also, as the authors mentioned, this investigation indicates that reducing expropriation risk would result in significant gains in income per capita, but do not point out what concrete steps would lead to an improvement. Therefore, in this way, another valuable step would be to examine which different aspects of institutions are important to prioritize and how can be improved in order to increase the income per capita in each country.

As a conclusion, this article shows a new, complete and innovative approach to the differences in GDP per capita between countries that has been an important question in politics. In my opinion, it gives us the opportunity to look into our history to evaluate the weakness in the colonial inherited institutions in order to be able to fix them for a better future and development of our country.

```python
plt.figure(figsize=(12,6))
plt.plot(data.año, data.pbipercap)
plt.xlabel("Año")
plt.ylabel("PBI per cápita")
plt.title("PBI per capita en Perú 1922-2021")
```

#Observaciones:  En general, se puede apreciar un crecimiento progresivo en el PBI per capita con el pasar de los años. Sin embargo, para el año 1990, aproximadamente, se puede ver una  caída del PBI per capita que se debería a la gran regresión económica que sufrió el Perú entre 1987 y 1990. Con un disminución del PBI per cápita de 3000 soles aproximadamente, el Perú 
regresaría a los niveles de la década de los 60's. A partir de la recuperación de esa caída en el año 2003 aproximadamente, Perú lograría mantener un crecimiento constante y más rápido entre 2003-2019 que en 80 años de su historia.Debido a la pandemia del 2020, se puede apreciar una nueva caída del PBI que estaría en proceso de recuperación para llegar al nivel alcanzado en el año 2019.

```python
 data['growth_rate'] = data["pbipercap"].pct_change()*100
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x = 'año', y= 'growth_rate', data=data)
plt.xlabel("Año")
plt.ylabel("Tasa de crecimiento anual del PBI real per cápita")
plt.title("Tasa de crecimiento anual del PBI real per cápita en Perú 1922-2021")
```
