Graph 1:
In this graph the X-axis is presented as the years and the Y-axis as the GDP per capita.
In the 1920s, there was a growth in GDP per capita due to increasing mining and agro-industrial exports. Starting in 1930 there is a drop, which could be explained as a consequence of the Great Depression and therefore uncertainty within the Peruvian state. Subsequently, the graph does not show falls of the same magnitude until around 1950, when there is instability in GDP per capita, ranging between 6,000 and 10,000 dollars. This instability lasts until around 2010, when it exceeds 10,000 dollars. The GDP per capita continues to grow until before 2020, when there is an abrupt fall due to the consequences of the pandemic and the measures taken by the president of the republic such as the bonuses, paralysis of the economy, etc. In general the GDP per capita grew from 1920 to 1970 from approximately 2000 dollars to almost 8000 dollars. Subsequently, due to ups and downs, from 1970 to 2000 the GDP per capita did not grow at all, but remained at around 8000 dollars. From the year 2000 to 2020 the GDP per capita rises from 8000 to more than 16000 dollars, i.e. it doubles its value. 


```python
plt.figure(figsize=(14,6))
plt.axhline(y=1.95, color = 'red',linestyle="-")
plt.axhline(y=0, color = 'black',linestyle="--")
sns.lineplot(x = 'Year', y= 'growth_rate', data=data, color = "blue")
plt.xlabel("Year")
plt.ylabel("anual growth rate of real GPD")
plt.title("The Annual Growth Rate of real GDP per Capita in PERU Between 1922 and 2021")
```
