What causes the large differences in GDP per capita between the countries of the world? For Acemoglu, Johnson, and Robinson, this can be answered through the analysis of the influence of the institutions inherited from the colonial era on GDP per capita. Opting for the use of the exogenous variable of settler mortality rates, it is distinguished from the 75 sample countries the existence of two groups, the "Neo-European" colonies, and the extractive state colonies. The authors assume that the mortality rate was decisive for the differentiation between both mentioned groups. It is in the Neo-European colonies where, having a low mortality rate, the creation of replica institutions of their centers was encouraged and they successively forged the basis for the institutions we know today that finally influenced their GDP per capita. Countries such as New Zealand, Canada, and Australia belong to this first group, which currently have a higher GDP per capita than countries that belonged to the group of extractive state colonies.

Despite the fact that the same authors recognize in their conclusions that the role of institutions in the growth or not of the GDP per capita of the countries, is not a single variable and there are other explanatory variables on the subject. This article demonstrates peculiar strengths, first of all, the use of linear regressions and the consideration of the variable "ei" as a representation of all the elements that can affect the study variables. Likewise, the use of overidentification test to validate his research, since it uses the exogenous variable from an exclusion restriction. Where it is shown that the mortality rate only affects the institutions and not the results of GDP per capita.

Secondly, the use of various datas, finding its main data coming from the work of Philip Curtin insufficient for them to deal with the colonies outside Europe such as those that were from Latin America, Asia, or Africa, introduces data such as those of Hall and Jones or Political Risk Services. Which served for the elaboration of the different tables and graphs, in which they simultaneously introduced potential variables that could affect their results and demonstrated little or no influence on their study.

However, the present investigation also has some weaknesses. Among the most fundamental is the vague idea of defining institutions. This is mainly due to the fact that up to now the institutions variable does not have an exact way of being measured, which is why exogenous variables are also used when addressing it. Simultaneously, by grouping the different colonies from different continents into a single study group, there is a risk of omitting more essential variables in explaining the role of institutions and the growth of GDP per capita.

The contributions of this research, in addition to being one of the first to analyze the growth of GDP per capita available to institutions, influenced by their colonial experience, open the doors to research on this topic. This is even reflected when they share information about research that initiated and contributed new knowledge, although some differ from the results of the analyzed study. Moreover, even the variables that the authors considered to be of little relevance or without influence in their study, can serve as bases to analyze similar topics.

Important steps to follow for future research, first, the delimitation of the type of institutions and to a greater degree, the importance of elites or power groups for the development of these institutions. At the same time, it is important to highlight the special peculiarities of the study units. They may share common features such as the experience of colonialism, but there are also particular elements such as the way of governing of the heads of these colonies among others.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('peru_gdp_1922_2021.xlsx')
```

```python
data.head()
```

```python
data.describe()
```

```python
data['growth_rate'] = data["GDP"].pct_change()*100
```

```python
data
```

```python
plt.figure(figsize=(14,10))
plt.plot(data.Years, data.GDP)
sns.lineplot(x = 'Years', y= 'GDP', data=data, color = "blue")
plt.xlabel("Year")
plt.ylabel("Real GDP per capita")
plt.title("PERU GDP percapita from 1922-2021")
```

From the present graph, it is observed that in general from 1920 to 2020, the GDP per capita of Peru has predominantly maintained a positive growth.
 
Between the mid-1940s and 1960s, one of the largest growths in GDP per capita in Peru was recorded, which ranged between 2,200 soles. Which is extremely impressive, since at the beginning of this time one of the strongest earthquakes occurred in the Peruvian capital, which caused too much damage. At the same time, between 2000 and 2010, another great growth took place, which borders on 3,800 soles. It should be noted that this growth occurred during the transition government to democracy, governed by Valent√≠n, who promoted integration and the need for active participation, not only of governments and businessmen, but also of society as a whole.

The fluctuation cycle between 2000 and 2020 is the only one with lower decreases in GDP per capita in Peru. Although this favorable economic growth was slowed down by covid 19 and the pandemic produced. However, it should be noted that the boom during this period was also due to the promotion of export policies.

```python
plt.figure(figsize=(14,10))
sns.lineplot(x = 'Years', y= 'growth_rate', data=data, color = "blue")
plt.xlabel("Year")
plt.ylabel("Annual growth rate of real GDP per capita")
plt.title("The Annual Growth Rate of real GDP per Capita in PERU Between 1922 and 2021")
plt.axhline(y=0, color = 'gray',linestyle="-")
plt.axhline(y=1.95, color = 'red',linestyle="-")
```

The graph shows that between the beginning of 1940 and the end of 1970, the annual growth rate of GDP per capita in Peru did not exceed a range between approximately 6% and almost -5% per year. This means that during those years, the annual growth rate of GDP per capita increased, increased slightly, or decreased year over year within the range mentioned above.
Between 1980 and the end of 1990, on the one hand, the highest annual growth rate of GDP per capita was recorded (almost 10%). On the other hand, within those years the lowest annual growth rate was also recorded with a figure close to -15%.
What is particular about the period 2000 to the end of 2019 is that only a positive growth rate was recorded.
In 2020, one of the most negative growth rates was recorded, probably due to the start of the COVID-19 pandemic.

This graph shows the annual growth rate of the Peruvian GDP per capita between 1920 and 2020. As has been observed, the growth rate has been negative and in an approximate percentage between -12 and -14% in four moments. The first produced by the Crisis of 29 and the Arequipa rebellion led by Commander Luis Miguel S√°nchez Cerro during 1930s. The next two negative growth points are periodically closer together, both occurring between the 1980-1990 period. And finally, the year 2020 with the theme of covid 19 and the pandemic.

However, it is necessary to mention that there were two exceptional moments where the speed of growth of GDP per capita reached up to 10%. This occurred around 1935 and early 1990, although it did not last very long, and these growth booms were followed by a slowdown in growth velocity.
