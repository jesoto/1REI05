```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Part 1: PERU GDP percapita from 1922-2021

```python
data = pd.read_excel('peru_gdp_1922_2021.xlsx')
```

```python
data.head()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x = 'Year', y = 'GDP per capita', data=data, color = 'green')
plt.title('Figure 1: Real GDP per Capita in Peru 1922-2021 (2007 Constant Dollars)')
```

- In general, the GDP per capita of Peru has had an upward trend during the evaluated period.
- Between the beginning of 1930 and the end of 1970, Peru's GDP per capita showed sustained growth, between 4,000 and 9,000 dollars, although not very pronounced and with slight downward variations.
- Between the 1980s and the end of the 1990s, Peru's GDP per capita went through a period marked by mainly downward variations. GDP per capita showed its steepest decline in that period, falling from approximately $8,500 to $6,500. Probably, this unstable period in GDP per capita is due to the Internal Armed Conflict that the country experienced during those years.
- Since the beginning of 2000, the country's GDP per capita has shown sustained and pronounced growth until the beginning of 2020. Between 9,000 and more than 16,500 dollars. This period coincides with the export boom that Peru experienced.
- Finally, the graph shows the drastic drop in GDP per capita that the country experienced due to the start of the COVID-19 pandemic. However, a recovery can also be observed in 2021 when it reached the figure of more than 16,500 dollars, a figure close to the year before the pandemic.

Part 2: The Annual Growth Rate of real GDP per Capita in PERU between 1922 and 2021

```python
data['growth_rate'] = data['GDP per capita'].pct_change()*100
```

```python
data.describe()
```

```python
plt.figure(figsize=(13,6))
sns.lineplot(x = 'Year', y = 'growth_rate', data=data)
plt.ylabel('Growth Rate (in percentage %)')
plt.axhline(y=1.95, color = 'red',linestyle="--")
plt.axhline(y=0, color = 'black',linestyle=":")
plt.title('Figure 2: The Annual Growth Rate of real GDP per Capita in PERU Between 1922 and 2021 (2007 Constant Dollars)')
```

- In the graph it can be seen that between the beginning of 1940 and the end of 1970 the annual growth rate of GDP per capita in Peru did not exceed a range between approximately a rate of 6% and almost -5% per year. This means that during those years, the annual growth rate of GDP per capita increased, increased slightly or decreased year after year within the interval mentioned above.
- Between 1980 and the end of 1990, on the one hand, the highest annual growth rate of GDP per capita (almost 10%) was recorded. On the other hand, the lowest annual growth rate was also recorded within those years with a figure of almost -15%. 
- What is particular about the period 2000 to the end of 2019 is that only a positive growth rate was recorded. 
- In 2020, one of the most negative growth rates was recorded, probably due to the start of the COVID-19 pandemic.

In the academy there are discussions about the possibility that state institutions and policies are the cause of the great differences in the per capita income per  capita of each country. However, there is not enough evidence to estimate the effects of institutions on economic performance. What Acemoglu, Johnson and Robinson propose is a theory of institutional differences between countries colonized by Europeans, and they analyze it to derive a possible source of exogenous variation for the economic performance of countries today. The theory of these authors can be summarized as follows: the mortality of the settlers affected the type of settlements that were established; the settlements affected the first institutions; and the early institutions persisted and formed the basis of today's institutions, which in turn impact economic performance. 

Some of the results were as follows. On the one hand, institutions alone explain 20% of the variation in the index of current institutions. Also, the fraction of Europeans in 1900 explains about 30% of the variation in our institutions today. This means that colonies where Europeans faced higher death rates are now substantially poorer than colonies that were healthy for Europeans. Finally, institutions have a significant impact on per capita income according to the d result of 0.94 in the 2SLS carried out by the researchers. It is important to emphasize that the authors emphasize that their results do not imply that the institutions we have today have been predetermined by colonial policies or that they cannot be changed. The colonial experience is one of the many factors that affect institutions. 

Three strengths can be found in the paper. First, the authors check the validity of their approach using overidentification tests. They do this because the validity of their approach, that is, the exclusion constraint, may be affected if other factors correlated with estimates of settler mortality affect per capita income. Second, it is important to note that the authors attempt to use data from all continents. For example, most of the data is obtained from the work of Philip Curtin; however, this does not include South America, so they use data from Hector Gutierrez. Lastly, his research also makes it possible to analyze hypotheses that had been raised by other researchers. For example, they found that distance from the equator does not have an independent effect on economic performance. 

On the other hand, in the investigation three weaknesses can be found in this investigation. First, the fact of using institutions. The authors recognize that many institutions are involved in economic performance, so any measure will be able to capture only part of the institutions involved; therefore, there is a measurement error in reducing the proxy of the institutions to the index of protection against the "risk of expropriation". Second, the paper excludes social parameters that may have influenced the formation of the first institutions in the colonies, such as culture, religion, the level of education of the authorities at the time, among others. Finally, it can be considered a weakness of the paper that the sources of information for the colonial time are not necessarily very reliable. 

The contribution of this document is that it was pioneering for its time. On the one hand, it was the first investigation - or one of the first - to carry out a comparative analysis of the economic performance of countries (on various continents), based on their institutions. On the other hand, the authors were the first to point out the link between the mortality of colonists and the institutions of today. This is particularly interesting because it proposed that the colonial experience did have an effect on the current economy of each country. 

Although the results of the paper indicate that reducing the risk of expropriation can lead to a higher per capita income, it does not mention what steps should be taken to improve the institutions that guarantee this protection. Therefore, an important next step would be to investigate how institutions can improve. In addition, it would be particularly interesting to analyze if there is any case that has been an exception to what the authors propose. On the other hand, another important step would be to carry out this same research, but on a regional or continental scale to see if there are differences or if the trend is the same. Finally, another valuable next step to move forward would be to analyze whether the type of regime could have influenced the degree of protection against the risk of expropriation (proxy of the authors for the institutions) of the colonies.
