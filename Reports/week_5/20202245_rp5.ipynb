As is known, the minimum wage policy emerged to try to help improve the situation of low-income workers by establishing lower limits for salaried workers. This with the aim of improving income distribution and inclusion. However, is it entirely true that a minimum wage policy helps improve the situation of low-income workers?
To answer this question, Jaramillo's article investigates and provides a concise answer to the problem posed.
The objective of this article is to evaluate the effects of the minimum wage on income and employment between the years 2000. To do this, a descriptive analysis of wage distributions and econometrics is used. To extend the investigation, data from a decade were used in which the increase in RMV was five times. For the evolution of the RMV, the remuneration of informal and unskilled workers is used as a reference point, since these are the most affected by the change in the RMV. Data from the permanent employment survey carried out in metropolitan Lima are used. Furthermore, the analysis focuses on specific target groups: wage earners, formal wage earners, informal wage earners, and independent workers.
The article is built on a base already explored but not fully analyzed, it is oriented in Peru, uses and considers variables ignored by some authors in the same field. It stands out that Neumark and Wascher, who are oriented to investigate the effect of the minimum wage on employment and find that the effects are diverse, both negative and positive.
On the income side, related research such as Kristenser and Cunningham also stands out, which analyzed the role of the minimum wage in wage distribution in the formal and informal sectors of Latin American and Caribbean countries. From this, the institutional heterogeneity of the minimum wage policy by region stands out, since informality and high non-compliance with the law is something concurrent in the object of study. Other investigations such as that of Saavedra in 2004 and Céspedes in the same year, are criticized for some shortcomings centered on the absence of a control group, so the article hopes not to make the same mistakes and enhance the investigation.
The results of the investigation show that, based on the descriptive analysis, there is an association between the RMV level and the average remuneration and the default rate. That is, the higher the relative level of RMV with respect to the average salary, the higher the rate of non-compliance and thus the lower the probability that this policy will achieve its objectives, as shown in graphs 5 and 8.
Regarding the effects on income, the minimum wage does not play a very important role in it, if there is any effect it is concentrated in the formal sector. This is confirmed by the econometric analysis as it shows that there is no lighthouse effect in Peru, on the contrary, the effects are mostly focused and centralized in specific groups, these are formal wage earners with salaries around the minimum wage level. The results show that for wage earners, there is a significantly weak effect for those earning between 0.9 and 1.2 times the RMV. For independent workers, there is only a significant effect for those who earn between 1.5 and 2 times the RMV. In the case of formal and informal, the effect is positive but not significant. This concludes that the effects of the minimum wage on earnings are minimal or statistically weak.
On the other hand, the effects on employment, according to the results, are not significant on those who earn around the minimum wage, on the contrary, the positive results are directed to the informal sector. According to the probability of continuing to be employed, for wage earners, the effect is positive for those who earn above the minimum wage between 1.2 and 1.5 times the RMV. In the case of formal workers, the effect is statistically weak for those who earn 0.6 and 0.9 times the RMV. For the informal ones, the effects are positive but for the 1.2 and 1.5 times the RMV. Finally, for independent workers, the effect is weak but positive for those who earn between 1.2 and 1.5 of the RMV. This concludes that there are no significant effects of maintaining employment for those who earn around the minimum wage, but there is a positive effect for the informal sector who earn above 20% of the RMV.
Finally, the author concludes that the minimum wage policy is not effective in promoting social inclusion since it does not raise the wages of those who earn less and because the effects on employment materialize in the informal sector of those who earn above the minimum wage.
