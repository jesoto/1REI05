Title: Peruvian labour market adjustments to changes in the minimum wage: The experience of the 2000s


There are several theories on the question of how the labour market adjusts to changes in the minimum wage. Therefore, the research question is what is the effect of the Minimum Living Wage on the labour market in different scenarios such as formal, informal, self-employed or close to the RMV? In this paper the authors aim to assess the effects of the minimum wage on two variables, income and employment. However, these variables are also affected by the status of the worker, i.e. whether the worker is young or female.


One of the strengths is that they used all available panels between 2002-2010 to examine the labour market effects of changes in the RMV, which allowed for greater variability in the data. Moreover, despite excluding the 2001 panel, they were able to analyse the data for the other years and found 229,323 observations. Finally, the use of longitudinal data allows them to examine changes in the labour market variables of individuals affected or unaffected by changes in the minimum wage.
 
 
On the other hand, among the weaknesses, we find that research on the minimum wage is limited in Peru and focuses more on specific dates. Another weakness is that the analysis only compares the coefficients of the "treated and control group" and does not develop a direct comparison, nor are statistically significant differences known. Finally, in Peru, most studies on the effects of the minimum wage emphasise September 2003, which makes it difficult to conduct a proper analysis.
The paper begins by focusing on the effects of an increase or decrease in the minimum wage on an individual's income or employment. The paper attempts to relate the Minimum Living Wage to the Average Wage and non-compliance with the norm.  In Peru, during the decade from 2000 to 2010, there were five increases in the RMV; although the effects of this increase have not been conclusive in the different studies carried out. These objectives are contextualised in different scenarios such as in the case of a formal wage earner, an informal wage earner, a self-employed person and a wage earner with a wage close to the RMV. As for the relationship between the RMV and the earnings of wage earners with wages close to the RMV, there is nothing conclusive. This is partly because it is a little studied topic and the years for which there is more information are from the decade 2000 to 2010.  There is a positive relationship between the RMV and the employment of formal workers, this is confirmed by Del Valle (2009), but he mentions that this positive relationship occurs for wages between the RMV and twice the RMV; there is also a positive relationship between the RMV and the employment of informal workers, when wages are between the RMV and 1.2 times the RMV. One point to note is that as the RMV increases so does non-compliance with the standard, and according to research this is more pronounced in the interior of the country. Therefore, the policy of increasing the RMV becomes less effective. Saavedra mentions a positive relationship between a change in the RMV and income, but a negative relationship between the RMV and employment of formal workers with wages close to the minimum wage.


 Also, different authors can find and confirm more specifically these relationships between RMV with respect to income or employment. For example: Jaramillo and López (2004): positive relationship between RMV and income, when workers' wages are between 0.9*MRV and 1.2*RMV; negative relationship between RMV and employment of private workers, when workers' wages are between 0.9*RMV and 2*RMV. Del Valle (2009) specifies this only for the years between 2003 and 2006. However, the dismissal of employees falls more strongly for a specific sector of workers, the probability of remaining employed for a young person between 25 years old is lower than for an average worker. Similarly, for women, the probability of remaining employed is lower. Finally, increasing or decreasing the RMV as a tool to promote social inclusion is not very effective and its effectiveness decreases in the interior of the country. This is why RMV increases are more often used to alleviate popular discontent. This gives us much insight into the context of the labour market in the interior of the country and the shortcomings of Peru's inclusion tools, as research papers such as Jaramillo's (2004) conclude that the policy of increasing the RMV does not serve the purpose of social inclusion.


One specific next step that would be valuable for this research would be further research or accessible information. This would allow access to more detailed research that could confirm or disconfirm some of the speculations made on the basis of lack of information, as lack of consensus is a common feature of these issues.
