I. Report Section 

This article by Miguel Jaramillo Baanante aims to evaluate the effects produced by the adjustments in the Peruvian labor market in the face of changes in the minimum wage in the 2000s. Likewise, it analyzes how the changes that occur in the minimum wage impact on two important labor market variables: income and employment. In order to respond to this assumption, the author uses the data of a whole decade, with which he manages to cover a total of five increases in the minimum vital remuneration (RMV).

Regarding the strengths of this work, in search of an answer to the research question, the author makes use of different methodologies that enrich the study. Among them, is the descriptive analysis. This observes the strong association between the level of the minimum wage in relation to average wages and the rate of non-compliance with the norm, and the significant growth of the minimum wage from the mid-1990s is presented. There is also the analysis of wage distributions that concludes that the minimum wage does not have a significant impact on the latter, confirmed by the econometric analysis. Lastly, it is in charge of studying the quarterly rotating role of the Permanent Employment Survey (EPE), which in turn focuses on different groups of workers: wage earners, formal wage earners, informal wage earners and independent workers. In turn, Baanante uses a series of previous empirical studies, in which important literature is reviewed. They show the effects of the minimum wage on employment and the relationship of the minimum wage with income. An important part of these studies is the great current heterogeneity obtained from the results with respect to previous current studies carried out.

Following this line of research, Baanante analyzes the effects of the increase in the RMV given in 2003, using econometric analysis. However, this shows weaknesses, since it is limited to comparing coefficients between the "treated" group and the "control" group, without making an estimate that allows them to be compared directly and to know if the differences between the two are significant, since they are not comparable. . Another weakness of the study is that being focused on a single region, Metropolitan Lima, the results shown are limited. In this context, the document does not show reliable support to analyze the data on informality.

With the use of the different variables, the author manages to understand that employment does not have any significant impact on those who earn around the Minimum Wage. However, this article contributes significantly to the knowledge of the subject, since the adjustments of the Peruvian labor market in the 2000s due to the changes in the minimum wage predominantly favored the informal sector. This is due to the substitution of less qualified workers for more qualified ones. Manifesting thus that the active policy of minimum wages is not a potentially effective tool as an instrument to promote social inclusion.

However, a next step not analyzed in this study would be to find out why the probability that income will increase is lower for formal workers with respect to this same probability for informal workers, a statistically significant result.

```python
model_1 = 'logpgp95 ~ avexpr'
results_1= smf.ols(model_1, data=data10).fit(cov_type='HC1')
print(results_1.summary())
```
