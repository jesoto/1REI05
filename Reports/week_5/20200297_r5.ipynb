REPORT

The article presented talks about the effects of the minimum wage on two highly relevant variables in the workplace: income and employment. Under this premise, the research responds to the questioning of the impact of the increases in the minimum vital wage on the Peruvian labor market, taking the 2000s as a reference. The author seeks to answer this question through the use of descriptive techniques. , analysis of salary distributions and an econometric analysis, thus demonstrating that, in reality, the minimum wage policy that is managed in the country is not really an efficient policy that manages to promote social inclusion in the labor scenario of Peru.

The article has a fairly clean organization, the author properly organizes 8 sections that explain the central axis of his research, this being one of the main strengths in terms of structure. On the other hand, if we talk about its effectiveness in answering the main question, it has both strengths and weaknesses. One of the strengths is the level of rigor that exists when evaluating the impact of the independent variable (minimum salary) on the dependent variable (remuneration and employability) since it makes use of Kernel density functions that help us to understand the distribution of influence in each case. In the same way, the use of econometrics in research can be highlighted, since thanks to the application of statistical methods (use of correlation) the analysis of the economic data that it presents is facilitated in addition to giving an empirical content to the data. assumptions made by the author.

However, it also has weaknesses, due to the little existing bibliography on the subject, it does not have a very broad academic support. In addition, although it mentions the heterogeneity that exists in employability and remuneration in the various regions of the country, in the end it is still a centralized study since the econometric analysis is limited only to Metropolitan Lima. The analysis mentions that the rate of non-compliance with respect to the minimum wage is higher in other departments, with Lima being the department that best complies with the minimum wage, however, it only mentions the subject but does not go into depth. Additionally, another weakness lies in the fact that the structure of the data used for the econometric analysis only generates quarterly time intervals, which implies that changes in the labor market are studied, but only in the short term, while studies such as that of David Neumark, Mark Schweitzer and William Wascher in their paper Effects of the minimum wage throughout the wage distribution in 2004 work with annual panels that make the study much more objective.

On the other hand, if we talk about the contributions of this paper, it can be mentioned that it is a very innovative work, although there are studies such as Maloney and Nuñez that study the same thing, but in the case of Colombia, it is one of the few not to say the only one that focuses on Peru and the increase in the minimum wage that occurred in 2003 and 2007. And as a consequence of this opening to this field of research, new proposals related to the subject may arise, dealing with variables that have been been able to exclude in this case, such as the distribution of working hours in relation to the minimum wage and its heterogeneity according to the sector in question. For example, in Washington, for workers who earn between 1.2 and 1.5 times the minimum, there are moderate but statistically significant increases in hours. The effects are more striking, with significant reductions in hours for people with the minimum wage, up to around 1.5 times the minimum wage (Neumark, Schweitzer, & Wascher, 2004), in Peru this issue should have different results that would be the same valuable for future research.
