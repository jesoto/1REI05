The IS Curve shows all the combinations of income (Y) and real interest rates (r) for which the market for real goods and services is in equilibrium. 

"I" represents investment and "S" the saving

This formula is derived from the equilibrium condition where output equals spending
          
 $$ \ Y=DA\ $$  
 $$ \ Y=C+I+G+X-M\ $$
 
Where:

C is household consumption:$$ \ C = C_0 + bY^d\ $$  
I is private investment:$$ \ I = I_0 ‚àí hr\ $$  
G is the level of government spending:$$ \ G = G_0\ $$  
X is exports:$$ \ X = X_0\ $$  
M is imports:$$ \ M = mY^d\ $$  
T is T taxation:$$ \ T = tY\ $$ 8not included in DA formula 


So, the DA formula will end like this:
$$ \ DA= C+I+G+X-M\ $$ 

$$ \ DA= C_0 + bY^d + I_0 ‚àí hr + G_0 + X_0 + mY^d \ $$ 

$$ \ DA=  (C_0 + I_0 + G_0 + X_0 ‚àí hr) + (b ‚àí m)(1 ‚àí t)Y\ $$

$$ \ DA = ùõº_0 + ùõº_1Y,\ $$ 

Where: $$ \ Œ±_0 = (C_0 + I_0 + G_0 + X_0 ‚àí hr)\ $$  is the intercept and $$\ Œ±_1 = (b ‚àí m)(1 ‚àí t)\ $$ is the slope of the function.
 


Now, to find the IS schedule:
$$ \ ùëå = ùê∂ + ùêº + ùê∫ + ùëã ‚àí  ùëÄ\ $$ 

$$ \ ùëå ‚àí ùëá = ùê∂ + ùêº ‚àí ùëá + ùê∫ + ùëã ‚àí  ùëÄ\ $$

$$ \ ùëå^ùëë = ùê∂ + ùêº ‚àí ùëá + ùê∫ + ùëã ‚àí ùëÄ\ $$ 

this can be written like this:
$$ \ (ùëå^ùëë ‚àí ùê∂) + (ùëá ‚àí ùê∫) + (ùëÄ ‚àí ùëã) = I\ $$ 

where "I" , the investment, is equal to "S", the total saving

The total saving is equal to the sum of private savings, the government savings and the external savings:


$$ \ ùëÜùëù + ùëÜùëî + ùëÜùëí = ùêº_0 ‚àí ‚Ñéùëü\ $$ 

$$ \ ùëÜ = ùêº_0 ‚àí ‚Ñéùëü\ $$ 

The IS schedule is equal to the equality between saving and investment:
$$ \ S(ùëå) = ùêº(ùëü)\ $$ 

$$ \ (ùëå^ùëë ‚àí ùê∂_0 ‚àí ùëèùëå^ùëë) + (ùëá ‚àí ùê∫_0) + (ùëöùëå^ùëë ‚àí ùëã_0) = ùêº_0 ‚àí ‚Ñéùëü\ $$ 

$$ \ [1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°)]ùëå ‚àí (ùê∂_0 + ùê∫_0 + ùëã_0) = ùêº_0 ‚àí ‚Ñéùëü\ $$ 

$$ \ r =1 / ‚Ñé(ùê∂_0 + ùêº_0 + ùê∫_0 + ùëã_0) ‚àí ((1 ‚àí (ùëè ‚àí ùëö)(1 ‚àí ùë°))/‚Ñé)*Y\ $$

$$ \ ùëü =(ùõΩ_0 /‚Ñé)‚àí(ùõΩ_1 /‚Ñé)*Y\ $$

Where: $$ \ Œ≤_0 = C_0 + I_0 + G_0 + X_0\ $$ and  $$ \ Œ≤_1 = 1 ‚àí (b ‚àí m)(1 ‚àí t)\ $$


```python
import numpy as np               
import matplotlib.pyplot as plt 
```

```python
"2|DEFINE PARAMETERS AND ARRAYS"
# Parameters
Y_size = 100
C0 = 20                 # Autonomous consumption
b = 0.2                # Marginal propensity to consume
alpha = 5              # Autonomous imports
m = 0.2            # Marginal propensity to import
T = 1                  # Taxes
I0 = 10             # Investment intercept (when r = 0)
G0 = 8              # Government spending
X0 = 2              # Exports (given)
h = 5                  # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE IS-SCHEDULE"
def r_IS(C0, alpha, b, m, T, I0, G0, X0, h, Y):
    r_IS = ((C0 + I0 + G0 + X0) - (1-(b-m)*(1-T))*Y)/h
    return r_IS

def Y_IS(Co, alpha, b, m, T, I0, G0, X0, h, r):
    Y_IS = (C0 + I0 + G0 + X0)/1-(b-m)*(1-T) - h*r/(1-(b-m)*(1-T))
    return Y_IS

r = r_IS(C0, alpha, b, m, T, I0, G0, X0, h, Y)

"4|PLOT THE IS-SCHEDULE"
y_max = np.max(r)
x_max = Y_IS(C0, alpha, b, m, T, I0, G0, X0, h, 0)

v = [0, x_max, 0, y_max]                        # Set the axes range
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r, "k-")
ax.yaxis.set_major_locator(plt.NullLocator())   # Hide ticks
ax.xaxis.set_major_locator(plt.NullLocator())   # Hide ticks
plt.axis(v)                                     # Use 'v' as the axes range
plt.show()
```
