The IS schedule is derived from the equilbrium condition where output (Y) equals spending

\begin{equation}
Y = C + I + G + (X - M)
\end{equation}

Where:

$C$ = Household consumption

\begin{equation}
C = C_0 + bY^d 
\end{equation}

$I$ = Private investment

\begin{equation}
I = I_0 - hr
\end{equation}

$G$ = Level of goverment spending

\begin{equation}
X = X_0
\end{equation}

$T$ = Taxation

\begin{equation}
T = tY
\end{equation}

\begin{equation}
M = mY^d 
\end{equation}


\begin{equation}
DA = C + I + G + X - M
\end{equation}

\begin{equation}
DA = C_0 + bY^d + I_0 - hr + G_0 + X_0 + mY^d
\end{equation}

\begin{equation}
DA = (C_0 + I_0 + G_0 + X_0 - hr) + (b-m)(1-t)Y
\end{equation}

\begin{equation}
DA = α_0 + α_1Y
\end{equation}

Where the intercept is:

\begin{equation}
α_0 = (C_0 + I_0 + G_0 + X_0 - hr)
\end{equation}

and the slope of the function is:

\begin{equation}
α_1 = (b-m)(1-t)
\end{equation}


For the IS:

\begin{equation}
Y = C + I + G + X - M
\end{equation}

\begin{equation}
Y - T = C + I - T + G + X - M
\end{equation}

\begin{equation}
Y^d = C + I - T + G + X - M
\end{equation}


Can be like this:

\begin{equation}
(Y^d - C) + (T - G) + (M - X) = I
\end{equation}


The total saving is equal to the sum of private savings, the government savings and the external savings:

\begin{equation}
S_p + S_g + S_e = I_0 - hr
\end{equation}

\begin{equation}
S = I_0 - hr
\end{equation}

The IS schedule is equal to the equality between saving and investment:

\begin{equation}
S(Y) = I(r) 
\end{equation}

\begin{equation}
(Y^d - C_0 - bY^d) + (T - G_0) + (mY^d - X_0) = I_0 - hr
\end{equation}

\begin{equation}
[1 -(b - m)(1 - t)]Y - (C_0 + G_0 + X_0) = I_0 - hr
\end{equation}

\begin{equation}
r = (Y^1/h(C_0 + G_0 + X_0) - ((1 -(b - m)(1 - t))/h) * Y
\end{equation}

\begin{equation}
r = (β_0/h) - (β_1/h) *Y
\end{equation}

So:

β_0 = C_0 + I_0 + G_0 + X_0
β_1= 1 - (b-m)(1-t)

```python
import numpy as np               
import matplotlib.pyplot as plt  


"2|DEFINE PARAMETERS AND ARRAYS"
# Parameters
Y_size = 100
C0 = 20                 # Autonomous consumption
b = 0.2                # Marginal propensity to consume
alpha = 5              # Autonomous imports
m = 0.2            # Marginal propensity to import
T = 1                  # Taxes
I0 = 10             # Investment intercept (when r = 0)
G0 = 8              # Government spending
X0 = 2              # Exports (given)
h = 5                  # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE IS-SCHEDULE"
def r_IS(C0, alpha, b, m, T, I0, G0, X0, h, Y):
    r_IS = ((C0 + I0 + G0 + X0) - (1-(b-m)*(1-T))*Y)/h
    return r_IS

def Y_IS(Co, alpha, b, m, T, I0, G0, X0, h, r):
    Y_IS = (C0 + I0 + G0 + X0)/1-(b-m)*(1-T) - h*r/(1-(b-m)*(1-T))
    return Y_IS

r = r_IS(C0, alpha, b, m, T, I0, G0, X0, h, Y)

"4|PLOT THE IS-SCHEDULE"
y_max = np.max(r)
x_max = Y_IS(C0, alpha, b, m, T, I0, G0, X0, h, 0)

v = [0, x_max, 0, y_max]                        # Set the axes range
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r, "k-")
ax.yaxis.set_major_locator(plt.NullLocator())   # Hide ticks
ax.xaxis.set_major_locator(plt.NullLocator())   # Hide ticks
plt.axis(v)                                     # Use 'v' as the axes range
plt.show()
```
