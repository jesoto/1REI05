INTERPRETATION:
If potential output increases, an excess supply is generated that must be eliminated by lowering the price level and lowering the interest rate. This decrease increases the real supply of money. This is a positive supply shock that shifts the aggregate supply curve toward the right.

MATH EXPLANATION

EFFECT ON INCOME
dY / Y̅ = −θhM^s0/|A|P0^2 >0

EFFECT ON THE INTEREST RATE
dr/Y̅ = −[1 − b(1 − t)]M^s0/|A|P^20 <0

EFFECT ON PRICE LEVEL
dP/Y̅ = θj[1 − b(1 − t)] + θhk /|A| < 0
