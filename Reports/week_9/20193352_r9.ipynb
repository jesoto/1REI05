### 2.1 Using IS and LM equations you have to find the $Y^e$ and $r^e$. Please show all the steps you follow to arrive at the equilibrium equations.

### $Y^e$ and $r^e$

### IS CURVE: 

$r$ = $\dfrac{β_0}{h}$- $\dfrac{β_1}{h}$*$Y$

- WHERE: $β_0$ = C0 + I0 + G0 + X0, 

$β_1$ = 1 - (b-m)(1-t)

### LM CURVE

$r$ = -$\dfrac{1}{j}$*$\dfrac{M_0^s}{P_0}$ + $\dfrac{k}{j}$Y

\begin{align}
   - \frac {1}{j}\ \frac {M^s _ 0}{P_0}\ + \frac {K}{j} Y \ &=\ \frac {β_0}{h}\ - \frac {β_1}{h}Y\ 
\end{align}

Equilibrium income:

\begin{align}
   \ Y^e &= -  \frac {j β_0}{kh + j β_1} Y \ + \frac {h}{kh + j β_1}\ \frac {M^s _ 0}{P_0}\ 
\end{align}

Equilibrium rate:

\begin{align}
   \ r^e &=  \frac {k β_0}{kh + j β_1} Y \ - \frac {β_1}{kh + j β_1}\ \frac {M^s _ 0}{P_0}\ 
\end{align}
