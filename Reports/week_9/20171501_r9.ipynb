20171501

The objective of the article is to talk about monetary non conventional politics, applied in the international crisis from 2008-2009. It shows how the Federal Reserve used interest rate to fund markets. 
The article works with two non conventional instruments: future path of the short-term interest rate and direct intervention in the long-term bond market. One of the reasons to write it, is to show that old methods (keynes, Hicks and Tobin) can be still useful in nowadays macroeconomic problems.

One of the more identifiable strength of the article is that it gets back to the begining of the topics, explains what are the models before assuming the reader knows about it.
On the other side, it highlights that one of its weakness can be that it may be too focused on the contemporary models rather than the non conventional ones, then rather that explain us details of it, we would know more about contemporary ones.

# Equilibrio simultaneo de bienes y gobierno

```python
Y_size = 100             
k = 20             
j = 10               
Ms = 20000            
P  = 20               
Y = np.arange(Y_size)

def i_LM_0( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM
i_0 = i_LM_0( k, j, Ms, P, Y)   
```

```python
b = 0.01               
m = 0.9           
T = 1              
I_bar = 10           
G_bar = 50             
X_bar = 2             
h = 1                 

Y = np.arange(Y_size)  


def i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS

def Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS
i = i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y)


y_max = np.max(i_0)
v = [0, Y_size, 0, y_max]                      
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Equilibrio simultÃ¡neo en mercados de bienes y de dinero", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i_0, "b-",color='lightblue', label="LM_0")
ax.plot(Y, i, "r-",color='yellow',label="IS_0")
ax.vlines(x=47, ymin=-120, ymax=-3,linestyle='dotted', colors='black')
ax.hlines(y=-4, xmin=-8, xmax=47,linestyle='dotted', colors='black')
plt.text(52, -9, 'E=( Y0, r0 )', color='black')
ax.legend()  
plt.show()
```
