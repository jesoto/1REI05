*Equilibrium in the AS-AD MODEL*

```python
import numpy as np               
import matplotlib.pyplot as plt  

"DEFINE PARAMETERS AND ARRAYS"
# Parameters
Y_size = 100
k = 30                # Transaction money demand
j = 20                # Speculation money demand
Ms = 30000             # Nominal money supply
P  = 30                # Price level
c = 30                 # Autonomous consumption
b = 0.03                # Marginal propensity to consume
m = 0.7            # Marginal propensity to import
T = 1                  # Taxes
I = 10             # Investment intercept (when i = 0)
G = 40              # Government spending
X = 2              # Exports (given)
h = 1                  # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"DEFINE AND POPULATE"
def i_DM(c, b, m, T, I, G, X, h, Y, Ms, j, k):
    i_DM = (h*Ms + j*(c + I + G + X) - (j*(1-(b-m)*(1-T)) + h + k)*Y)/h
    return i_DM

i = i_DM(c, b, m, T, I, G, X, h, Y, Ms, j, k)
#---------------------------------------------------
Y_size = 100
Pe = 28200
O = 20
Yrayita = 10
Y = np.arange(Y_size)

def i_SD(Pe, O, Yrayita, Y):
    i_SD = Pe + O*(Y-Yrayita)
    return i_SD

i_1 = i_SD(Pe, O, Yrayita, Y)

"PLOT THE AD-AS MODEL"
y_max = np.max(i_DM)

v = [0, Y_size, 0, y_max]                       # Set the axes range
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="AD-AS MODEL", xlabel=r'Y', ylabel=r'P')
ax.plot(Y, i, "r-",label="AD")
ax.plot(Y, i_1, "b-",label="SD")
ax.yaxis.set_major_locator(plt.NullLocator())   # Hide ticks
ax.xaxis.set_major_locator(plt.NullLocator())   # Hide ticks
ax.legend()
plt.show()
```
