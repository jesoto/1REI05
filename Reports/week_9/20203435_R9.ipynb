## I. REPORT

In 2008, in the United States, the housing bubble burst which generated one of the biggest financial crises of the last decades. Its central bank, the FED, initially used the benchmark interest rate to try to counter the crisis; this tool was unsustainable when it reached near zero. Consequently, this entity had to intervene in the long-term debt market and announce the future trajectory of the interest rate. Both measures are considered unconventional monetary policies. What this article does is extend the traditional model of aggregate demand and supply to incorporate the innovative monetary policies that were implemented to counter the international crisis of 2008-2009. To do this, it uses the traditional IS-LM model, but enriched with models that include imperfect substitute financial assets. In this extended model, on the one hand, the FED administers the short-term interest rate, not the money supply; on the other hand, a long-term bond market is added. In this way, unconventional monetary policy is modelled and its effects on endogenous variables are discussed. Likewise, the model also serves for the author to analyze the effects of fiscal policy and adverse supply shocks. 

Regarding the strengths, it is possible to find the following. First, the author recognize that the model he uses is based on a market with particular characteristics, in this case, the Anglo-Saxon market. This is considered important because it recognizes the limits of the model, which can be an open door for future research. Second, rational expectations are included. As the author points out, in recent years there have been criticisms of macroeconomic policy for just considering that people take into account past prices (adaptive expectations) and not all available information (rational expectations). Third, another strength of the document is that for each supply shock is analyzed: the short-term effects, the effects on transit to the steady state, the effects on the steady state and the effects with rational expectations. Finally, a strong point of the article is that it also makes a contribution in terms of explaining the different effects of fiscal and monetary policies anticipated or not anticipated. 

As for the weaknesses of the document, two could be found. On the one hand, the author does not deep into how the announcement about the future path of the interest rate is inserted or not in the IS-LM-BB model that he proposes. On the other hand, the author mentions in the article that the evidence of whether or not these policies were effective was described by the IMF. In this sense, it leaves open the possibility that, in practice, these unconventional monetary policies have not been effective in countering the crisis. Therefore, a weakness of the article is that this possibility has not been included, or the reason for this possibility, by means of the proposed model. 

The contribution of this article is that it allows people who have no economic expertise to understand unconventional monetary policies, and their effects, through traditional models and methods. The author considers that these are still useful to address many contemporary macroeconomic problems by simple and more convenient language. In that sense, it is noteworthy that the author has adapted the old models to explain how the US faced one of the biggest financial crises of the last decade. 

The first valuable next step to move forward should be to perform the same analysis, but with the characteristics of the Peruvian financial market. This means that instead of using only the bond market, the most common bank loans and deposits in less developed financial markets would be taken into account. Finally, a second important step would be to conduct a study on how US monetary policy innovations influenced the goods market and the money market of other countries, because the central banks of the other countries take into account the rate set by the FED to fix their reference interest rate.

### 1) DA equation

The function that represents IS curve:

[1 - b(1-$t$)]$Y$ = $C_0$ + $I_0$ + $G_0$ - $hr$

$hr$ = $C_0$ + $I_0$ + $G_0$ - [1 - b(1-$t$)]$Y$

- $β_0$ = $C_0$ + $I_0$ + $G_0$
- $β_1$ = [1 - b(1-$t$)]


<center> IS curve: $r$ = $\dfrac{β_0}{h}$ - $\dfrac{β_1}{h}$*$Y$ ... (i) </center>

The function that represents LM curve:

$\dfrac{M_0^s}{P_0}$ = $kY$ - $jr$

$jr$ = - ($\dfrac{M_0^s}{P_0}$) + $kY$


<center> LM curve: $r$ = - $\dfrac{1}{j}$*($\dfrac{M_0^s}{P_0}$) + $\dfrac{k}{j}$*$Y$ ...(ii) </center>

Equalize (i) and (ii):

-$\dfrac{1}{j}$*$\dfrac{M0^s}{P_0}$ + $\dfrac{k}{j}$*Y = $\dfrac{β_0}{h}$ - $\dfrac{β_1}{h}$*$Y$

$\dfrac{k}{j}$*Y + $\dfrac{β_1}{h}$*Y = $\dfrac{β_0}{h}$ + $\dfrac{1}{j}$*$\dfrac{M0^s}{P_0}$

$Y$*$[\dfrac{k}{j} + \dfrac{β_1}{h}]$ =  $\dfrac{β_0}{h}$ +  $\dfrac{1}{j}$*$\dfrac{M0^s}{P_0}$

$Y$*$[\dfrac{hk + jβ_1}{jh}]$ = $\dfrac{β_0}{h}$ +  $\dfrac{1}{j}$*$\dfrac{M0^s}{P_0}$


<center> $Y^e$ = $\dfrac{jβ_0}{hk+jβ_1}$ + $\dfrac{h}{hk+jβ_1}$*($\dfrac{M0^s}{P_0}$) </center>...(iii)

<center> $r^e$ = $\dfrac{kβ_0}{hk+jβ_1}$ + $\dfrac{β_1}{hk+jβ_1}$*($\dfrac{M0^s}{P_0}$) </center> ...(iv)

Solve $P$ as a function of $Y$ in (iii)

($\dfrac{h}{kh + jβ_1}$)*$\dfrac{M_0^s}{P_0}$ = $Y$ - $\dfrac{jβ_0}{kh + jβ_1}$

$\dfrac{M_0^s}{P_0}$ = ($Y$ - $\dfrac{jβ_0}{kh + jβ_1}$) / ($\dfrac{h}{kh + jβ_1}$)

$\dfrac{M_0^s}{P_0}$ = $\dfrac{Y(kh + jβ_1}{h}$ - $\dfrac{jβ_0}{h}$

$\dfrac{M_0^s}{P_0}$ = $\dfrac{Y(kh + jβ_1) - jβ_0}{h}$

$h$$M_0^s$ = $P_0$*$[Y(kh + jβ_1) - jβ_0]$


<center> DA equation: $P$ = $\dfrac{hM_0^s}{-jβ_0 + (kh + jβ_1)Y}$ </center>


### 2) Effects of an expansive fiscal policy (Δ$G_0$>0) on the endogenous variables $Y$, $r$ and $P$

```python
class class_AD:
    def __init__(self, C0    = 20    ,  # AD: consumo autónomoa 
                       b     = 0.2   ,  # AD: propensión marginal a consumir
                       t     = 1     ,  # AD: recaudación tributaria
                       I0    = 10    ,  # AD: inversión autónoma
                       G0    = 8     ,  # AD: gasto del gobierno
                       h     = 5     ,  # AD: efecto marginal de la tasa de interés
                       c1    = 175   ,  # AD: Precautionary money demand
                       c2    = 2     ,  # AD: Transactions money demand
                       c3    = 50    ,  # AD: Speculation money demand
                       B     = 250   ,  # AD: Base money
                       lmbda = 0.05  ,  # AD: Currency drain ratio
                       rho   = 0.10  ):  # AD: Reserve requirement         
        self.C0    = C0
        self.b     = b
        self.t     = t
        self.I0    = I0
        self.G0    = G0
        self.h     = h
        self.c1    = c1
        self.c2    = c2
        self.c3    = c3
        self.B     = B
        self.lmbda = lmbda
        self.rho   = rho
                 
    "Multiplicador del dinero"             
    def m(self):
        lmbda = self.lmbda
        rho   = self.rho
        return ((1 + lmbda)/(rho + lmbda))
   
    "Oferta monetaria"     
    def M(self):
        B = self.B
        return (B*self.m())
                 
    "DA: Demanda agregada"
    def AD(self, P):
        C0    = self.C0
        b     = self.b
        t     = self.t
        I0    = self.I0
        G0    = self.G0
        h     = self.h
        c1    = self.c1
        c2    = self.c2
        c3    = self.c3

        AD_level1 = (((C0-b)*t+I0+G0)/h-c1/c3)
        AD_level2 = ((1-b)/h-c2/c3)**(-1)
        AD_shape  = self.M()/c3 * ((1-b)/h - c2/c3)**(-1)
        return (AD_level1 * AD_level2 + AD_shape/P)
    
out = class_AD()

size = 50
P = np.arange(1, size)
Y = out.AD(P)
```

```python
y_max = np.max(P)
x_max = np.max(Y)
v = [0, x_max, 0, y_max]
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="DEMANDA AGREGADA", xlabel="Y", ylabel="P")
ax.plot(Y, P,"k-", scalex = True, scaley = True)
plt.axis()
plt.show()
```

#### a. Graphic analysis
The increase in government spending leads to $β_0$ increasing. As $β_0$ increases, the intercept of the demand curve equation increases. It is for this reason that the aggregate demand curve expands.

DA: $P$ = $\dfrac{hM_0^s + jβ_0}{h}$ - $\dfrac{kh + jβ_1}{h}$*$Y$

Where:
- The intercept is $\dfrac{hM_0^s + jβ_0}{h}$
- $β_0$ = $C_0$ + $I_0$ + $G_0$

#### b. Intuition
- On the Aggregate Demand side: ↑ $G0$ --> Aggregate demand increases (↑ DA) --> Which produces an excess in the goods market generating an imbalance: DA>Y --> Consequently, the level of production of the economy must increase to return to equilibrium (↑ $Y$)
- On the aggregate supply side: ↑ $Y$ --> The increase in the product causes the difference between it and the potential product to increase ↑($Y$ - $\bar{Y}$) --> Consequently the price increases (↑$P$). The curve does not shift because there was no change in any exogenous variable.

#### c. Differentials

Where:
- The intercept is $\dfrac{hM_0^s + jβ_0}{h}$


#### b. Intuition

OA: $P$ = ($P^e$ - $θ\bar{Y}$) + $θY$

Where:
- The intercept is $P^e$ - $θ\bar{Y}$


#### b. Intuition

#### a. Graphic analysis
- The increase in potential output (↑$\bar{Y}$) leads to a decrease in the intercept of the supply curve equation. Therefore, the aggregate supply curve expands.

OA: $P$ = ($P^e$ - $θ\bar{Y}$) + $θY$

Where:
- The intercept is $P^e$ - $θ\bar{Y}$

#### b. Intuition
- On the one hand, the increase in potential product (↑$\bar{Y}$) generates a drop in current prices (↓$P$). --> Consequently, the real money supply increases ($↑\dfrac{M_0^s}{P_0}$), which generates an imbalance in the money market: $M^d$<$\dfrac{M_0^s }{P_0}$ --> Therefore, the interest rate decreases to compensate for the excess demand (↓$r$).
- On the other hand, ↓$r$ leads to an increase in investment due to its inverse relationship (↑$I$) --> This generates an increase in demand and leads to an imbalance in the market: $DA$ > $Y$ - -> Consequently, the level of production must increase to return to equilibrium (↑$Y$).
