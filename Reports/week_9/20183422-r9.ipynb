Title: Aggregate demand and supply in the presence of unconventional monetary policies.


The paper presents market equilibrium models and shows that the old IS-LM model still serves to explain and model different cases of today's economy. It also allows to discuss fiscal and monetary policy effects; effects of moving the short-term interest rate and unconventional policies such as buying bonds for financing. In turn, with Walras' law we can dispense with one market since if two of the three markets are in equilibrium the third one will also be in equilibrium. Furthermore, the IS-LM model was developed in detail and the market for both short-term and long-term bonds was added. Therefore, the question could be asked: how does one arrive at the IS-LM model and what modifications does it have when the bond market is added?




One of the strengths of the paper is that thanks to Walras' law, two markets can be analysed in depth and one can be left out, since equilibrium is guaranteed if two of the three markets are in equilibrium. However, one of the weaknesses of this paper is that although several economic concepts are clearly presented, they are not reflected in reality with examples and derivations in the face of an external shock and what policies are expected to bring the economic model back to equilibrium.  Similarly, these models do not allow us to know the trajectory of endogenous variables between short-term equilibrium and final equilibrium; but they do allow us to know the trajectory towards stationary equilibrium. 




We are presented with equations and rules such as Walras' law and then introduced to the IS-LM model which is still very useful for the development of a model that describes well the behaviour of the economy. We are given the equations that make up the IS-LM model. And we are introduced to the concept of money creation by the purchase and sale of bonds by the FED. However, the monetary authority may be left without much power to stabilise the economy if it enters the liquidity trap. However, given the limited content of this model in terms of reality or specific cases, I believe that this paper's main purpose is to mathematically demonstrate the models presented, in order to educate the reader. 




Clearly one would expect that the next valuable step of this paper could be to bring this economic model to specific realities or to the reality of a country. Also, to complement this, one could analyse the comparison of this economic model with other countries, in order to analyse the effects of monetary policies, fiscal policies, unconventional monetary policies, moving short-term interest rates, conventional monetary policies, long-term bond purchases, supply shocks in both the short run and the long run, transit in the stationary equilibrium, etc.

```python
"1|IMPORT PACKAGES"
import numpy as np               # Package for scientific computing with Python
import matplotlib.pyplot as plt  # Matplotlib is a 2D plotting library

"2|DEFINE PARAMETERS AND ARRAYS"
# Parameters
Y_size = 100
             # Precautionary money demand
k = 25             # Transaction money demand
j = 10                # Speculation money demand
Ms = 20000             # Nominal money supply
P  = 20                # Price level
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE LM-SCHEDULE"
def i_LM_0( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM
i_0 = i_LM_0( k, j, Ms, P, Y)
#------------------------------------
Y_size = 100
             # Precautionary money demand
k = 25               # Transaction money demand
j = 10              # Speculation money demand
Ms = 30000             # Nominal money supply
P  = 20                # Price level
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE LM-SCHEDULE"
def i_LM_0( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM
i_1 = i_LM_0( k, j, Ms, P, Y)

#-------------------------------------   

Y_size = 100
c = 20                 # Autonomous consumption
b = 0.01                # Marginal propensity to consume
             # Autonomous imports
m = 0.9            # Marginal propensity to import
T = 1                  # Taxes
I_bar = 15            # Investment intercept (when i = 0)
G_bar = 50              # Government spending
X_bar = 2              # Exports (given)
h = 1                  # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE IS-SCHEDULE"
def i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS

def Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS


i = i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y)
#---------------------------------------------------



"4|PLOT THE IS-SCHEDULE"
y_max = np.max(i_0)

v = [0, Y_size, 0, y_max]                       # Set the axes range
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="LM SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i_0, "b-", label="LM_0")
ax.plot(Y, i, "r-",label="IS_0")
ax.plot(Y, i_1, "r-",linestyle='dashed',label="LM_1")
ax.yaxis.set_major_locator(plt.NullLocator())   # Hide ticks
ax.xaxis.set_major_locator(plt.NullLocator())   # Hide ticks
ax.legend()


plt.text(47, 60.07, 'I', fontsize=12, color='black')
plt.text(27, 2.07, 'II', fontsize=12, color='black')
plt.text(67, 2.07, 'III', fontsize=12, color='black')
plt.text(47, -60, 'IV', fontsize=12, color='black')

plt.text(90, 90, 'LM', fontsize=12, color='black')
plt.text(90, 140, 'LM*', fontsize=12, color='black')
plt.text(90, -75, 'IS', fontsize=12, color='black')

plt.axvline(x=42.5,  ymin= 0, ymax= 0.53 , linestyle = ":", color = "black")
plt.axhline(y=6, xmin= 0, xmax= 0.43, linestyle = ":", color = "black")

plt.axvline(x=54.5,  ymin= 0, ymax= 0.45 , linestyle = ":", color = "black")
plt.axhline(y=-17, xmin= 0, xmax= 0.54, linestyle = ":", color = "black")

plt.show()
```
