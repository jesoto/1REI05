# Codification part

IS curve:

\begin{align}
    \ r &= \frac {β_0}{h}\ - \frac {β_1}{h}Y\        
\end{align}  

LM curve:

\begin{align}
    \ r &= - \frac {1}{j}\ \frac {M^s _ 0}{P_0}\ + \frac {K}{j} Y\  
\end{align}

Differentiating the functions in terms of P

\begin{align}
P = -\frac{h Mo^s}{-j B_o + (jB_1 + hk)Y}
\end{align}

In terms of Y

\begin{align}
Y = \frac{jB_o}{jB_1 + hk} + (\frac{hMo^s}{jB_1 + hk})\frac{1}{P}
\end{align}

Transforming the equation of the LM curve through logarithms, we obtain:

\begin{align}
Mo^s - P = kY - jr
\end{align}

Deriving r, in the Aggregate Demand equation, as a function of P

\begin{align}
P = \frac{h Mo^s + jB_o}{h} - \frac{jB_1 + hk}{h}Y
\end{align}

\begin{align}

Endogenous variable P
\begin{align}
P^e↑ - P↑ 
\end{align}
Given the rise in expected prices, that is, prices in the future are going to rise, there will be a contraction of the OA curve, going from P_0 to P_1. The level of output falls and the price level rises.

Endogenous variable r: 
\begin{align}
Y↓ - Md↓ - Md < \frac {M^s}{P}\ - r ↑ 
\end{align}

Endogenous variable Y:

Due to the inverse relationship between prices and the level of production, if there is an increase in the price, consequently there will be a decrease in the level of production:

\begin{align}
P↑ - Y↓
\end{align}

Another explanation for the fall in the level of production is that, as the price level increases and the interest rate rises, there is a reduction in the AD curve because the level of investment is reduced.

\begin{align}
r ↑ - I ↓ - DA ↓ - DA < Y - Y ↓
\end{align}

Since there is a decrease in the level of production, the demand for money will also decrease. In such, there will be an imbalance in the LM curve, since the demand for money will be less than the real supply of money. To return to equilibrium, the interest rate must be increased. Because of this there is a shift to the left. In conclusion, prices rise and the interest rate, for the purposes of this, production falls

Mathematic explanation:

Endogenous variable P
\begin{align}
   \ ΔP &= ΔP^e
\end{align}

\begin{align}
\frac {ΔP}{ΔP^e}\ &= 1
\end{align}

\begin{align}
1 &> 0
\end{align}

\begin{align}
   ΔP^e &> 0
\end{align}

Endogenous variable Y

\begin{align}
   \ ΔP^e &= -KΔY\
\end{align}

\begin{align}
   \frac {ΔP^e}{ΔY^e}\ & = - K
\end{align}

\begin{align}
   ΔY &< 0
\end{align}

Endogenous variable r

\begin{align}
   \ ΔP^e &= - (-jΔr)
\end{align}

\begin{align}
   \frac {ΔP^e}{Δr}\ & = j
\end{align}

\begin{align}
   \ Δr &> 0
\end{align}

Intuitive explanation:

Endogenous variable r:

\begin{align}
(Y- Y_f)↑ - \frac {M^s}{P}\ ↑ - \frac {M^s}{P}\ > Md - r↓
\end{align}

The increase in marginal product causes the money supply to increase. Which generates an imbalance in the LM curve, since the money supply is greater than the demand for money. So to return to a balance between both parties, the interest rate is reduced.

Endogenous variable Y: 

\begin{align}
r↓ - I↑ - DA↑ > Y - Y↑
\end{align}

With the fall in the interest rate, investment increases, so aggregate demand increases. This creates an imbalance between the amount of aggregate demand and the level of production. For which, to return to equilibrium, the level of production is increased.

Endogenous variable P:

Due to the inverse relationship between prices and the level of production, if there is an decrease in the price, consequently there will be an increase in the level of production:

\begin{align}
(Y- Y_f)↑ - Y ↑ - P ↓ 
\end{align}



Mathematic explanation:

Endogenous variable P
\begin{align}
   \ \frac{ΔP}{ΔP^e} = \frac{-j[1 - b(1 - t)] - hk}{|A|} > 0 
\end{align}

\begin{align}
   ΔP &>0
\end{align}

Endogenous variable Y

\begin{align}
   \frac{ΔY}{ΔP^e} = \frac{hMo^s}{|A|P_o^2} < 0
\end{align}

\begin{align}
   ΔY &<0
\end{align}

Endogenous variable r

\begin{align}
\frac{Δr}{ΔP^e} = \frac{-[1 - b(1 - t)]Mo^s}{|A|P_o^2} > 0
\end{align}

\begin{align}
   \ Δr &> 0
\end{align}
