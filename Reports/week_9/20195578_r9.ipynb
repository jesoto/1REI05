## The old economic methods are still alive!

The main objective of the academic article, entitled "AGGREGATE DEMAND AND SUPPLY IN THE PRESENCE OF UNCONVENTIONAL MONETARY POLICIES", consists in modeling unconventional monetary policy, extending the standard model of closed economy aggregate demand and supply to incorporate innovations of monetary policy and thus counteract the aftermath of the international crisis at the end of the 2000s. Before the crisis, the main instrument of monetary policy of the central bank of the United States, the Federal Reserve, was the interest rate of benchmark for the federal funds markets. The reduction of the interest rate almost to zero due to the crisis forced the monetary authority of the United States to resort to two policy instruments, which are labeled in the article as "unconventional": the announcement about the trajectory future of the short-term interest rate and direct purchase in the market of long-term bonds, which is an essential component of quantitative easing.

The contribution of this work lies in the confirmation and reaffirmation that the traditional models and traditional methods, in reality, have not lost their usefulness for studying and dealing with macroeconomic problems. Unlike the Hicks-Hansen models, in which the monetary policy instrument is the money supply, the Federal Reserve is considered to manage the short-term interest rate, not the money supply, which is endogenous. In addition, in order to deal with the purchase of long-term securities by the Reserve, a long-term bond market is added to the Hicks-Hansen model, in which there is only a short-term bond market. All this constitutes the aggregate demand of the economy. This aggregate demand is combined with a conventional aggregate supply curve, thereby obtaining a macroeconomic model that allows determining the equilibrium values of production, the long-term interest rate and the price level. The model also allows discussion of the effects of fiscal policy and the adverse supply shock.

The two outstanding strengths of the article are the following: the first is based on the figure of the person who wrote it. The author of this document is Dr. Waldo Mendoza, who has been a member of the board of directors of the Central Reserve Bank of Peru (from January 2015 to July 2016), and, subsequently, has held the position of Minister of Economy and Finances during the transition government of Francisco Sagasti. His experience and impeccable track record in the economic framework, specifically in monetary and fiscal policies, make this academic work reliable and efficient. The second strength is that this article, by working together with theory and graphics, demonstrates that old models and methods can still be useful in answering the series of contemporary macroeconomic unknowns. The models of Keynes, Hicks or Tobin and the traditional method of comparative statics have become tools of the past for many, but, in more than one respect, they explain the complex events reported by the crisis better than modern models and methods.

There are a couple of aspects of the article that can be classified as weaknesses: one of them is its lack of validity, manifested in not taking into account a series of current factors that influence the aggregate supply, such as the covid 19 pandemic or the effects of natural phenomena. Another weakness is the very poor figure of opinions, constructive criticism, and/or counterarguments from other academics, experts on the subject, regarding Mendoza's article, despite the fact that it has been published for more than five years. In effect, this situation can be interpreted, on the one hand, as a limit, but, on the other hand, it can be conceived as a valuable opportunity to develop, in the not remote future, academic works that study in depth, for example, the behavior of aggregate demand and supply in the presence of unconventional monetary policies in times of the covid 19 pandemic.

## The equilibrium in the IS-LM model

### IS

$r$=$β_{0}$*$h^{-1}$-$(β_{1}$*$h^{-1})$$y$

### LM

$r$=$-1$*$j^{-1}$*$M_{0}^{s}$$(P_{0})^{-1}$+$K(j)^{-1}$$y$

Considering the proposed equations, it is possible to replace some variables to find $y^{e}$ and $r^{e}$, which represent the equilibrium of the IS-LM model...

$-1$*$j^{-1}$*$M_{0}^{s}$$(P_{0})^{-1}$+$K(j)^{-1}$=$β_{0}$*$h^{-1}$-$(β_{1}$*$h^{-1})$$y$

$y^{e}$=$j$$(β_{0})(Kh+β_{1})^{-1}$+$h$($Kh$+$β_{1})^{-1}$$(M_{0}^{s}(P_{0})^{-1})$

$r^{e}$=$j$$(β_{0})(Kh+β_{1})^{-1}$+$β_{1}$($Kh$+$β_{1})^{-1}$$(M_{0}^{s}(P_{0})^{-1})$

### 2.1 Simultaneous equilibrium in the goods and money markets 

```python
import numpy as np               
import matplotlib.pyplot as plt  


Y_size = 100          
k = 20               
j = 10                
Ms = 20000             
P  = 20                
Y = np.arange(Y_size)  
def i_LM_0( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM
i_0 = i_LM_0( k, j, Ms, P, Y)


Y_size = 100
c = 20                 
b = 0.01                
m = 0.9            
T = 1                 
I_bar = 20             
G_bar = 55              
X_bar = 2             
h = 1                  
Y = np.arange(Y_size)  
def i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS
def Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS


i = i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y)


y_max = np.max(i_0)

v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Graphic of the simultaneous equilibrium in the goods and money markets", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i_0, "b-", label="LM_0")
ax.plot(Y, i, "r-",label="IS_0")

plt.axvline(x=51,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black") 
plt.axhline(y=1, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")   
plt.plot(51,1, marker="o", color="black")
plt.text(50.1, 5, 'E', fontsize=12, color='black')

ax.legend()
plt.show()
```
