## A model for our economy in times of covid-19

The academic article, entitled ‚ÄúThe Macroeconomics of Quarantine: A Two-Sector Model‚Äù, presents, as the name reveals, a two-sector macroeconomic model. This model aims to represent both the drop in GDP, a product of the covid 19 pandemic, and the main tool to contain the spread of viruses, quarantine. Sector 1 includes restaurants, the hotel industry, airlines, and is directly affected by the quarantine. Sector 2, on the other hand, encompasses the production of food, services, and essential goods, and is indirectly affected by the negative demand shock from sector 1. The two proposed sectors are Keynesian, since production is determined by demand; and the price levels depend on the expected level of prices and the GDP gap in each sector. The intersectoral connection occurs because workers consume goods from both sectors. Due to the quarantine, the decrease in potential GDP and autonomous consumption in sector 1 trigger a general recession.

Through the joint work of the theoretical component with the statistical one, the contribution of this article consists of the confirmation and reaffirmation that the decrease in both potential GDP and autonomous consumption is transitory, that is, its values experience a reestablishment after a few periods. Due to the fear that the pandemic produces in consumers, consumption suffers, constituting a shock to demand, which affects production in both the first and second sectors. After the falls in production and prices in both sectors, the economy subsequently presents a gradual recovery, due to the sustained decline in expected prices, which reduce prices, raise real wealth, consumption, demand and production in both sectors. The continuation of the recovery culminates in a new stationary equilibrium, with the new potential output of sector 1, and the same potential output of sector 2.

The two outstanding strengths of the academic work are the following: the first is based on those who elaborated it. The identity of the authors corresponds to Waldo Mendoza, a prominent Peruvian economist and former Minister of Economy and Finance of Peru, Luis Mancilla, a graduate of the specialty of Economics from the Pontifical Catholic University of Peru, and Rafael Velarde, also a graduate of Economics and teacher of this science. As has been seen, they are familiar with Research and Economics, which are topics required for the preparation of an ideal academic work, such as this one. The second strength lies in the relative validity of this article, since at the national, regional, and international levels we are immersed in a scenario of economic recovery largely due to the effects of the mandatory social isolation decreed as a result of the covid 19 pandemic: the collapse of the workers' income, and, consequently, the weakening of the productive capacity and the decrease of the potential GDP.

The weakness of the article consists in the very poor or non-existent figure of the opinions, the questions, and the constructive criticism of other academics regarding the article by Mendoza, Mancilla and Velarde, because it was published only in February of this year. This can be interpreted as a limit, but, from another angle, it can be conceived as a valuable opportunity to develop the theme in other academic works. In this model, emphasis has been placed on the demand components in order to propose alternative economic stabilization policies, so this dimension cannot be rescued. In the not remote future, a model could be developed that provides an analysis of the effect of quarantine duration on potential output.

## The equilibrium in the IS-LM model

### IS

$r$=$Œ≤_{0}$*$h^{-1}$-$(Œ≤_{1}$*$h^{-1})$$y$

### LM

$r$=$-1$*$j^{-1}$*$M_{0}^{s}$$(P_{0})^{-1}$+$K(j)^{-1}$$y$

Considering the proposed equations, it is possible to replace some variables to find $y^{e}$ and $r^{e}$, which represent the equilibrium of the IS-LM model...

$-1$*$j^{-1}$*$M_{0}^{s}$$(P_{0})^{-1}$+$K(j)^{-1}$=$Œ≤_{0}$*$h^{-1}$-$(Œ≤_{1}$*$h^{-1})$$y$

$y^{e}$=$j$$(Œ≤_{0})(Kh+Œ≤_{1})^{-1}$+$h$($Kh$+$Œ≤_{1})^{-1}$$(M_{0}^{s}(P_{0})^{-1})$

$r^{e}$=$j$$(Œ≤_{0})(Kh+Œ≤_{1})^{-1}$+$Œ≤_{1}$($Kh$+$Œ≤_{1})^{-1}$$(M_{0}^{s}(P_{0})^{-1})$

### 2.1 Simultaneous equilibrium in the goods and money markets 

```python
import numpy as np               
import matplotlib.pyplot as plt  


Y_size = 100          
k = 20               
j = 10                
Ms = 20000             
P  = 20                
Y = np.arange(Y_size)  
def i_LM_0( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM
i_0 = i_LM_0( k, j, Ms, P, Y)


Y_size = 100
c = 20                 
b = 0.01                
m = 0.9            
T = 1                 
I_bar = 20             
G_bar = 55              
X_bar = 2             
h = 1                  
Y = np.arange(Y_size)  
def i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS
def Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS


i = i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y)


y_max = np.max(i_0)

v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Graphic of the simultaneous equilibrium in the goods and money markets", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i_0, "b-", label="LM_0")
ax.plot(Y, i, "r-",label="IS_0")

plt.axvline(x=51,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black") 
plt.axhline(y=1, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")   
plt.plot(51,1, marker="o", color="black")
plt.text(50.1, 5, 'E', fontsize=12, color='black')

ax.legend()
plt.show()
```

Intuitive explanation: in the economy, sooner or later everything tends to balance. Under this premise, the graph is born from the simultaneous equilibrium in both markets, that is, at the intersection of the IS and LM curves. This intersection point is $E$=($ùíÄ_{0}$, $ùíì_{0}$).

Mathematical explanation:

$y^{e}$=$j$$(Œ≤_{0})(Kh+Œ≤_{1})^{-1}$+$h$($Kh$+$Œ≤_{1})^{-1}$$(M_{0}^{s}(P_{0})^{-1})$

$r^{e}$=$j$$(Œ≤_{0})(Kh+Œ≤_{1})^{-1}$+$Œ≤_{1}$($Kh$+$Œ≤_{1})^{-1}$$(M_{0}^{s}(P_{0})^{-1})$
