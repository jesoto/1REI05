# The equilibrium in the IS-LM model

IS

\begin{align}
    \ r &= \frac {β_0}{h}\ - \frac {β_1}{h}Y\         (1)
\end{align}               

LM

\begin{align}
    \ r &= - \frac {1}{j}\ \frac {M^s _ 0}{P_0}\ + \frac {K}{j} Y\  (2)
\end{align}

To find the balance in the IS-LM model, we must equate both equations 1 and 2

\begin{align}
   - \frac {1}{j}\ \frac {M^s _ 0}{P_0}\ + \frac {K}{j} Y \ &=\ \frac {β_0}{h}\ - \frac {β_1}{h}Y\ (3)
\end{align}

Finding the equilibrium income level, the following equation results

\begin{align}
   \ Y^e &= -  \frac {j β_0}{kh + j β_1} Y \ + \frac {h}{kh + j β_1}\ \frac {M^s _ 0}{P_0}\  (4)
\end{align}

To find the equilibrium interest rate, we substitute equation 4 into 1

\begin{align}
   \ r^e &=  \frac {k β_0}{kh + j β_1} Y \ - \frac {β_1}{kh + j β_1}\ \frac {M^s _ 0}{P_0}\  (5)
\end{align}

# 2.1. Simultaneous equilibrium in the goods and money markets

```python
import numpy as np               
import matplotlib.pyplot as plt  


Y_size = 100          
k = 20               
j = 10                
Ms = 20000             
P  = 20                
Y = np.arange(Y_size)  
def i_LM_0( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM
i_0 = i_LM_0( k, j, Ms, P, Y)


Y_size = 100
c = 20                 
b = 0.01                
m = 0.9            
T = 1                 
I_bar = 20             
G_bar = 55              
X_bar = 2             
h = 1                  
Y = np.arange(Y_size)  
def i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS
def Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS


i = i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y)


y_max = np.max(i_0)

v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(16, 14))
ax.set(title="Simultaneous equilibrium in the goods and money markets", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i_0, "red", label="LM_0")
ax.plot(Y, i, "black",label="IS_0")

plt.axvline(x=51,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black") 
plt.axhline(y=1, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")   
plt.plot(51,1, marker="o", color="black")
plt.text(50.1, 5, 'E', fontsize=12, color='black')

ax.legend()
plt.show()
```

Intuitive explanation:

On the IS curve: 
\begin{align}
ΔGo - DA ↑ - DA > Y - Y ↑
\end{align}

On the LM curve: 
\begin{align}
Y ↑ - Md ↑ - Md > \frac {M^s}{P}\ - r ↑ 
\end{align}
Although this change in the LM curve has an additional effect on the IS curve:
\begin{align}               
r ↑ - I ↓ - DA < Y - Y ↓ 
\end{align}
This last blow produced in the IS curve has a smaller effect, so the one initially described predominates. 

In the IS curve, with an increase in autonomous government spending, there is an increase in Aggregate Demand. With this increase, this causes the Aggregate Demand to be greater than the Income level. So, to maintain equilibrium, Income must be increased by the same amount as Aggregate Demand increased. On the LM curve, the increase in income will cause the real demand for money to increase. Being this greater than the real supply, to return to equilibrium, the real supply of money must increase, causing the interest rate to increase.

Mathematic explanation:

\begin{align}
   \ ΔY^e &=  \frac {j}{kh + j β_1}
\end{align}

\begin{align}
   ΔG_o &> 0
\end{align}

\begin{align}
   \ Δr^e &=  \frac {k}{kh + j β_1}
\end{align}

\begin{align}
   ΔG_o &> 0
\end{align}

So:

\begin{align}
   \ ΔY^e &> 0
\end{align}

\begin{align}
   \ Δr^e &> 0
\end{align}

In conclusion, before the intervention of the government policy of fiscal increase, increase in public spending, and thus the level of GDP will increase. Although it has a serious consequence, the interest rate will increase, which scares the investment.

# Intuitive explanation:

a. Expansive monetary policy

On the LM curve: 
\begin{align}
ΔM^s ↑ - \frac {M^s}{P}\ ↑ - \frac {M^s}{P}\ > M^d - r ↓
\end{align}

On the IS curve: 
\begin{align}
r ↓ - I ↑ - DA ↑ - DA > Y - Y ↑ 
\end{align}

b. Restrictive fiscal policy

On the IS curve: 
\begin{align}
Go ↓ - DA ↓ - DA < Y - Y ↓
\end{align}

On the LM curve: 
\begin{align}
ΔM^d ↓ - M^d <\frac {M^s}{P}\ - r ↓
\end{align}



An expansionary monetary policy means that there is a larger monetary supply. So, the money supply will be greater than the money demand. And to restore the balance, the interest rate will have to be reduced.

Its effects on the IS curve will be that, by reducing the interest rate, the investment will increase. This will cause the Aggregate Demand to increase, being greater than the level of production. Again to maintain the balance, the level of production will have to be increased.

In the case of a restrictive fiscal policy, a reduction in government spending is assumed. With this, the Aggregate Demand will be diminished and will end up being less than the level of production. So, to maintain equilibrium, the level of production will have to decrease.

The consequence of this on the LM curve is that the demand for money will decrease since there is a decrease in the level of production. The level of demand will be lower than the level of money supply, so, to maintain equilibrium, the interest rate will have to decrease.

Both changes, products of the two policies, cause the level of production to remain at a level of production. But with variations are represented by the decrease in the interest rate.

Intuitive explanation:

In the IS curve:
a.
\begin{align}
t ↓ - T ↑- Y^d ↑ - C_o ↑ - DA ↑ - DA > Y - Y ↑
\end{align}

b.
\begin{align}
t ↓ - Y^d ↑ - M ↑ - (X -M) ↓ - DA ↓ - DA < Y - Y ↓
\end{align}

In the LM curve:
\begin{align}
    Y ↑ - M^d ↑ - M_o > \frac {M^s}{P}\ - r ↑
\end{align}

The reduction in the tax rate produces an increase in disposable income. This effect produces two consequences:

a. The increase in disposable income will cause autonomous consumption to increase, which will cause Aggregate Demand to increase. This will generate an inequality, to counteract this, the level of production must increase.

b. The increase in disposable income will cause the level of imports to increase. What will cause the income from exports to decrease. This will cause the amount of Aggregate Demand to be reduced, being less than the level of production. To maintain equilibrium, the level of production will have to decrease.

The most significant blow between the two cases is the first, since the number of people who matter is less than the number of the entire population that basically consumes. Thus, the reduction of the tax rate will cause the level of production to grow as well as the interest rate.
