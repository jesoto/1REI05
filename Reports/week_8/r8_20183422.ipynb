Title: The Macroeconomics of Quarantine: A Two-Sector Model


In 2020 Latin America suffered a decline in GDP. This was due to covid-19.  The purpose of this article is to reproduce and show a macroeconomic model of sector 1 (directly affected) and sector 2 (indirectly affected) inspired by Blanchard. an IS-LM-DA-OA macroeconomic model and the IS-MR-DA-OA model are used. Given these macroeconomic models, we can ask how it relates to reality in the context of the pandemic.



One of the strengths is that we can easily identify the variables of the macroeconomic model, so we can develop the graph or equilibrium with some precision. This allows us to know the equilibrium values of output and the price level. It also facilitates the study of how these endogenous variables move from the initial stationary equilibrium to the final stationary equilibrium. It can be seen that the interest rate of both sectors does not change as output and prices decrease. However, it declines more sharply in sector 1, as output is reduced due to pandemic constraints, and in sector 2 output continues, but with a decline in demand, leading to excess supply. In response to this, the government used a credit policy applied in the period 2020-2021 that prevented massive company bankruptcies.
One of the weak points is that the reduction of disposable income in sector 1 affects demand, but the extent to which it affects demand is not detailed, and the lack of information in this market, which generates excess supply in sector 2, is not explained. Another weakness is that the paper makes little reference to fiscal policy or bank intervention to help the country's economic recovery.



The paper begins by contextualising a pandemic scenario, where we are shown different effects on the variables that will be developed in the macroeconomic models. Such as the marginal propensity to consume, which decreases; autonomous consumption, which also decreases; the interest rate, which remains constant; the Keynesian multiplier, which indicates the magnitude of the effect of the movement of the variables on production; etc. Then with these variables he develops a model in which aggregate demand and aggregate supply are present, which generates an equilibrium and shows us the equilibrium level of production and what the level of production would be if one of the variables increases or decreases. In the same way, he develops a model where the monetary rule is held constant with respect to the interest rate. And finally, he relates these models to the effects of the pandemic on two sectors, both sector 1 (which was directly affected by the restrictions of the pandemic) and sector 2 (which was indirectly affected, since the decrease in demand caused by sector 1 affected the demand of sector 2, thus generating an excess supply in this sector).
