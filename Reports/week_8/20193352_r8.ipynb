### 2.1 Instruction: Derive mathematically the equilibrium in the IS-LM model. You have to find the equilibrium equations of Y^e and R^e.

### $Y^e$ and $r^e$

### IS CURVE: 

$r$ = $\dfrac{β_0}{h}$- $\dfrac{β_1}{h}$*$Y$

- WHERE: $β_0$ = C0 + I0 + G0 + X0, 

$β_1$ = 1 - (b-m)(1-t)

### LM CURVE

$r$ = -$\dfrac{1}{j}$*$\dfrac{M_0^s}{P_0}$ + $\dfrac{k}{j}$Y

\begin{align}
   - \frac {1}{j}\ \frac {M^s _ 0}{P_0}\ + \frac {K}{j} Y \ &=\ \frac {β_0}{h}\ - \frac {β_1}{h}Y\ 
\end{align}

Equilibrium income:

\begin{align}
   \ Y^e &= -  \frac {j β_0}{kh + j β_1} Y \ + \frac {h}{kh + j β_1}\ \frac {M^s _ 0}{P_0}\ 
\end{align}

Equilibrium rate:

\begin{align}
   \ r^e &=  \frac {k β_0}{kh + j β_1} Y \ - \frac {β_1}{kh + j β_1}\ \frac {M^s _ 0}{P_0}\ 
\end{align}

### 2.2 Simultaneous equilibrium in the goods and money markets

```python
import numpy as np               
import matplotlib.pyplot as plt  
```

```python
Y_size = 100          
k = 20               
j = 10                
Ms = 20000             
P  = 20                
Y = np.arange(Y_size)  
def i_LM_0( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM
i_0 = i_LM_0( k, j, Ms, P, Y)


Y_size = 100
c = 20                 
b = 0.01                
m = 0.9            
T = 1                 
I_bar = 20             
G_bar = 55              
X_bar = 2             
h = 1                  
Y = np.arange(Y_size)  
def i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS
def Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS


i = i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y)


y_max = np.max(i_0)

v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Simultaneous equilibrium in the goods and money markets", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i_0, "b-", label="LM_0", color="pink")
ax.plot(Y, i, "r-",label="IS_0",color="orange")

plt.axvline(x=51,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black") 
plt.axhline(y=1, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")   
plt.plot(51,1, marker="o", color="black")
plt.text(50.1, 5, 'E', fontsize=12, color='black')

ax.legend()
plt.show()
```

### Explanation: 
As we see in the graph, this appears from the simultaneous equilibrium of the two markets (an intersection of the IS and LM curves = $E$=($𝒀_{0}$, $𝒓_{0}$)).

$y^{e}$ $r^{e}$

$y^{e}$=$j$$(β_{0})(Kh+β_{1})^{-1}$+$h$($Kh$+$β_{1})^{-1}$$(M_{0}^{s}(P_{0})^{-1})$

$r^{e}$=$j$$(β_{0})(Kh+β_{1})^{-1}$+$β_{1}$($Kh$+$β_{1})^{-1}$$(M_{0}^{s}(P_{0})^{-1})$
