```python
import numpy as np               
import matplotlib.pyplot as plt  


Y_size = 100             
k = 20             
j = 10               
Ms = 20000            
P  = 20               
Y = np.arange(Y_size)  


def i_LM_0( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM
i_0 = i_LM_0( k, j, Ms, P, Y)

#-------------------------------------   

Y_size = 100
c = 20                
b = 0.01               

m = 0.9           
T = 1              
I_bar = 10           
G_bar = 50             
X_bar = 2             
h = 1                 

Y = np.arange(Y_size)  


def i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y):
    i_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - (1-b+m)*Y)/h
    return i_IS

def Y_IS(c, b, m, T, I_bar, G_bar, X_bar, h, i):
    Y_IS = ((c)-(b-m)*T + I_bar + G_bar + X_bar - h*i)/(1-b+m)
    return Y_IS


i = i_IS(c, b, m, T, I_bar, G_bar, X_bar, h, Y)


y_max = np.max(i_0)
v = [0, Y_size, 0, y_max]                      
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Equilibrio simultáneo en los mercados de bienes y de dinero", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i_0, "b-", label="LM_0")
ax.plot(Y, i, "r-",label="IS_0")
ax.vlines(x=47, ymin=-120, ymax=-3,linestyle='dotted', colors='black')
ax.hlines(y=-4, xmin=-8, xmax=47,linestyle='dotted', colors='black')
plt.text(52, -9, 'E=( Y0, r0 )', color='black')
ax.legend()  
plt.show()
 
```
