```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

IS Curve:

$$ r = \frac{B_o}{h} - \frac{B_1}{h}Y $$

- Where $ B_0 = C_o + I_o + G_o + X_o $ and $ B_1 = 1 - (b - m)(1 - t) $

LM Curve:

$$  r = -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y $$

These equations can be reduced:

$$ -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y = \frac{B_o}{h} - \frac{B_1}{h}Y $$

$$ Y^e = \frac{j B_o}{k h + j B_1} + (\frac{h}{k h + j B_1})\frac{Ms_o}{P_o} $$

- Equilibrium Interest Rate $ (r^e) $

$$ r^e = \frac{kB_o}{kh + jB_1} - (\frac{B_1}{kh + jB_1})\frac{Ms_o}{P_o} $$

Equilibrium of the IS-LM model.

#### Simultaneous equilibrium in money and goods markets

```python
# IS Curve
Y_size = 100 
b = 0.4                 
m = 0.2                 
t = 0.9                  
Co = 20              
Io = 10                  
Go = 50                 
Xo = 2                  
h =  1                 

Y = np.arange(Y_size)
```

```python
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
r
```

```python
Y_size = 100

k = 15                
j = 10                
Ms = 9000             
P  = 20         

Y = np.arange(Y_size)
```

```python
def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
i
```

```python
y_max = np.max(i)

fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Simultaneous equilibrium in money and goods markets", xlabel= r'Y', ylabel= r'r')

ax.plot(Y, r, "k-", label = "IS", color = "green")
ax.plot(Y, i, "b-", label="LM", color = "blue")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())  

plt.axvline(x=51.1,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black")
plt.axhline(y=32, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")
plt.plot(51, 32, marker="o", color="black", label = "E")
plt.text(49,38, 'E', fontsize = 14, color = 'black')

plt.text(99, 99, 'LM', fontsize=13)
plt.text(99, -11, 'IS', fontsize=13)
plt.text(48, 80, 'I', fontsize=13)
plt.text(80, 35, 'II', fontsize=13)
plt.text(48, -13, 'III', fontsize=13)
plt.text(20, 35, 'IV', fontsize=13)

ax.legend()
plt.show()
```

$$ G_o↑ → DA↑ → DA > Y → Y↑ $$
$$ Y↑ → M_d↑ → M_d > M_o → r↑ $$ 
    
An increase in Public Spending mean an increase in Aggregate Demand. So, Aggregate Demand will be greater than the Production Level. Therefore, to return to equilibrium, the production level will increase.

An increase in production level will cause the increase of Money Demand. Therefore, the demand for money will be greater than the Money Supply. For the equilibrium, the interest rate will increase.

##### Math

$$ ∆Y^e = \frac{j∆G_o}{kh + jB_1} $$

$$ ∆Y^e = [\frac{j}{kh + jB_1}] ∆G_o > 0 $$


$$ ∆r^e = \frac{k∆G_o}{kh + jB_1} $$

$$ ∆r^e = [\frac{k}{kh + jB_1}] ∆G_o > 0 $$

In the money market, an increase in the mass of money will cause the money supply to be greater than demand. Therefore, to equalize, the interest rate must be reduced

The interest's reduction will encourage more investment.While the investment increases, aggregate demand also increase. Then, the aggregate demand will be greater than production. And, the level of production must grow.
