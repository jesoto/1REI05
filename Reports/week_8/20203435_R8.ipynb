## I. REPORT

The COVID-19 pandemic began at the beginning of 2020. As expected, it caused a collapse of all economic indicators, Peru was no exception. Thus, that year was marked by the collapse of GDP and employment, becoming the most severe in Peru since 1989. However, a year later, the country registered a vigorous recovery in the aforementioned indicators, with levels equal to or above those registered before the pandemic. In this way, what this article seeks is to analyze the macroeconomic effects generated by the quarantine, as a public health policy to fight the pandemic, in Peru. To do this, the authors present a macroeconomic model that reproduces the effects of the quarantine on the Peruvian economy. Two sectors are distinguished in the model. On the one hand, sector 1, directly affected by the quarantine (hotels, restaurants, airlines). On the other hand, sector 2, indirectly affected by the quarantine (production of food, and essential goods and services). The quarantine is replicated as a decrease in the partial product and autonomous consumption in sector 1. The article concludes that there was a contraction in general production and the level of prices, but very unevenly, due to the different nature of the sectors analyzed. . Likewise, the authors conclude that the scenario that best resembles reality is exercise 2 (where the falls in potential GDP and autonomous consumption in sector 1 are transitory) because the GDP levels were re-established; the article attributes said effect to the credit policy applied ("Reactiva Peru") during the 2020-2021 period. 

Regarding the strengths of the document, three can be found. First, a model is used that makes it possible to see all the effects of the quarantine by disaggregating the elements of the economy's aggregate supply and demand. Second, the authors carry out three exercises to analyze the possible effects of the quarantine. Finally, it is highlighted that the authors have adapted previous models to develop the question of their work; they use an economic model inspired by Blanchard and elaborated with the instruments of Mendoza. 

As for the weaknesses, three can be found. First, that the model does not take into account the particular characteristics of the country. On the one hand, the high levels of informality are overlooked, an aspect that can influence variables such as public spending and the ease of firing workers (which affects their income). On the other hand, it does not take into account the uneven development between the regions. In regions like Lima there may be a balance between sector 1 and sector 2, but other regions clearly depend on sector 2 and their recovery could not have been at the same rate as those that depend on both sectors. Second, it does not take into account that the credit policy was focused on large companies, so only 16.6% of mypes benefited from Reactiva Peru (RRP 2022). Although this could have generated a recovery in production in general, the possible consequences of the collapse of the mypes on the country's future GDP are not taken into account. Finally, it does not consider that other variables can influence price rigidity and, therefore, not lead to the increase in consumption that the model expects; In addition, it does not delve into the explanation of how credit policy influenced the reactivation of the economy. 

This article was probably published because it demonstrates the effectiveness of stimulating demand to get out of an economic slowdown. Peru had one of the worst falls in GDP and in employment due to the pandemic measures (quarantine) and, despite that, managed to overcome the slowdown, registering the levels prior to the pandemic at the end of 2021. Therefore, this paper shows to the rest of the Latin American countries that if they follow this “recipe” (countercyclical measures), their economy can stabilize. 

The first next step would be to carry out an exercise that considers the continuous subsidies that the government granted. In the final part, the authors propose that another exercise should be carried out considering other public policies that were applied, such as the reduction of the interest rate (fiscal relief) and the tax rate (monetary stimulus). Considering that the Peruvian government made continuous use of money transfers to its population, it should be considered as a variable of interest for future research. A second important next step would be to carry out the same study, but at a mathematical level, taking into account the real data of the period to check the validity of the model and its predictions. Finally, another valuable step to move forward would be to carry out the same study, but in the Latin American region to compare the macroeconomic effects of different public policies against the pandemic in countries with similar characteristics.

### 1) Equilibrium equations of $Y^e$ and $r^e$

The function that represents the IS Curve: 

<center> $r$ = $\dfrac{β_0}{h}$- $\dfrac{β_1}{h}$*$Y$ </center>

- $β_0$ = C0 + I0 + G0 + X0
- $β_1$ = 1 - (b-m)(1-t)

The function that represents the LM Curve:

<center> $r$ = -$\dfrac{1}{j}$*$\dfrac{M_0^s}{P_0}$ + $\dfrac{k}{j}$Y </center>

For the economy to be in equilibrium, the two equations are set equal. The objective is to place the endogenous variables (r, Y) in function of the exogenous ones.

-$\dfrac{1}{j}$*$\dfrac{M0^s}{P_0}$ + $\dfrac{k}{j}$*Y = $\dfrac{β_0}{h}$ - $\dfrac{β_1}{h}$*$Y$

$\dfrac{k}{j}$*Y + $\dfrac{β_1}{h}$*Y = $\dfrac{β_0}{h}$ + $\dfrac{1}{j}$*$\dfrac{M0^s}{P_0}$

$Y$*$[\dfrac{k}{j} + \dfrac{β_1}{h}]$ =  $\dfrac{β_0}{h}$ +  $\dfrac{1}{j}$*$\dfrac{M0^s}{P_0}$

$Y$*$[\dfrac{hk + jβ_1}{jh}]$ = $\dfrac{β_0}{h}$ +  $\dfrac{1}{j}$*$\dfrac{M0^s}{P_0}$

<center> $Y^e$ = $\dfrac{jβ_0}{hk+jβ_1}$ + $\dfrac{h}{hk+jβ_1}$*($\dfrac{M0^s}{P_0}$) </center>

To find $r^e$, substitute $Y^e$ into the IS curve:

$r$ = $\dfrac{β_0}{h}$- $\dfrac{β_1}{h}$*[$\dfrac{β_0}{h}$- $\dfrac{β_1}{h}$]

<center> $r^e$ = $\dfrac{kβ_0}{hk+jβ_1}$ + $\dfrac{β_1}{hk+jβ_1}$*($\dfrac{M0^s}{P_0}$) </center>

### 2) Equilibrium IS-LM

```python
import numpy as np               
import matplotlib.pyplot as plt  
```

```python
Y_size = 100
C0 = 20                  # Autonomous consumption
b = 0.2                  # Marginal propensity to consume
m  = 0.2                 # Marginal propensity to import
t = 1                    # Taxes
I0 = 10                  # Investment intercept (when i = 0)
G0 = 8                   # Government spending
X0 = 2                   # Exports (given)
h = 5                    # Investment slope wrt to i

Y = np.arange(Y_size)  
```

```python
def r_IS(C0, b, m, t, I0, G0, X0, h, Y):
    r_IS = (C0-(b-m)*t + I0 + G0 + X0 - (1-b+m)*Y)/h
    return r_IS

def Y_IS(a, b, m, t, I0, G0, X0, h, r):
    Y_IS = (C0-(b-m)*t + I0 + G0 + X0 - h*r)/(1-b+m)
    return Y_IS

r = r_IS(C0, b, m, t, I0, G0, X0, h, Y)
```

```python
Y_size = 100           # Precautionary money demand
k = 10                 # Transaction money demand
j = 10                 # Speculation money demand
Ms = 20000             # Nominal money supply
P  = 20                # Price level

Y = np.arange(Y_size)  # Array values of Y
```

```python
def r_LM(k, j, Ms, P, Y):
    r_LM = (-(1/j)*(Ms/P)) + (k/j)*Y
    return r_LM

rl = r_LM(k, j, Ms, P, Y)
```

```python
y_max = np.max(rl)
x_max = np.max(r)
                        
v = [80, 100, -30, 0]
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Simultaneous equilibrium in the goods and money markets", xlabel='Y', ylabel='r')
ax.plot(Y, r, scalex = True, scaley = True, color = 'black', label = "IS")
ax.plot(Y, rl,scalex= True, scaley = True, color = 'blue', label = "LM")

ax.vlines(x = 90, ymin = -30, ymax = -10, linestyle = "dotted", color = "grey")
ax.hlines(y = -10, xmin = -80, xmax = 90, linestyle = "dotted", color = "grey")

plt.text(89.7, -9, 'E', fontsize = 17, color = 'black')
plt.text(89.7, -5, 'I', fontsize = 15, color = 'green')
plt.text(95, -9, 'II', fontsize = 15, color = 'green')
plt.text(92.5, -17, 'III', fontsize = 15, color = 'green')
plt.text(82.5, -15, 'IV', fontsize = 15, color = 'green')

plt.axis(v)
ax.legend()
plt.show()
```

### 3) Increase government spending ($Go$)

```python
G0 = 20

def r_IS_G(C0, b, m, t, I0, G0, X0, h, Y):
    r_IS = (C0-(b-m)*t + I0 + G0 + X0 - (1-b+m)*Y)/h
    return r_IS

def Y_IS_G(a, b, m, t, I0, G0, X0, h, r):
    Y_IS = (C0-(b-m)*t + I0 + G0 + X0 - h*r)/(1-b+m)
    return Y_IS

r_G = r_IS_G(C0, b, m, t, I0, G0, X0, h, Y)
```

```python
y_max = np.max(rl)
x_max = np.max(r)
                        
v = [80, 100, -30, 0]
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Effect of an increase in Government Spending", xlabel='Y', ylabel='r')
ax.plot(Y, r, scalex = True, scaley = True, color = 'black', label = "IS")
ax.plot(Y, r_G, scalex = True, scaley = True, color = 'red',linestyle="--", label = "IS_G")
ax.plot(Y, rl,scalex= True, scaley = True, color = 'blue', label = "LM")

plt.text (96, -10.8, '↑', fontsize = 25, color = 'green')
plt.text (80, -7.80, '↑', fontsize = 25, color = 'green')

ax.vlines(x = 92.1, ymin = -30, ymax = -8.1, linestyle = "dotted", color = "grey")
ax.hlines(y = -8.1, xmin = 80, xmax = 92.1, linestyle = "dotted", color = "grey")
plt.text(89.7, -9.4, 'E', fontsize = 17, color = 'black')
plt.text(91.8, -7.3, 'E2', fontsize = 17, color = 'black')

plt.axis(v)
ax.legend()
plt.show()
```

When the government decides to apply an expansionary fiscal policy, what increases in the model is autonomous consumption ($G0$). Consequently, the following two effects occur.
- ↑ $G0$ --> Aggregate demand increases (↑ DA) --> Which produces an excess in the goods market generating an imbalance: DA>Y --> Consequently, the level of production of the economy must increase to return to equilibrium (↑ $Y$): The IS curve shifts to the right.
- The increase in income (↑ $Y$) to compensate for the imbalance generates an increase in the demand for money in the market (↑ $M^d$). --> Which generates an imbalance in the money market: $M^d$>$\dfrac{M_0^s}{P_0}$ --> As the monetary mass is fixed, what moves to adjust the market is the interest rate. Therefore, the interest rate increases to compensate for the excess demand (↓ $r$).: The LM curve does not shift because there was no change in any exogenous variable.

<center> Δ$Y$ = $\dfrac{ΔG0}{[1-(b-m)(1-t)]+(kh)/j}$ > 0 </center>



<center> Δ$r$ = $\dfrac{ΔG0*(k/j)}{[1-(b-m)(1-t)]+(kh)/j}$ > 0 </center>

### 3) Increase in the amount of money ($M^s$)

```python
Ms = 20700

def r_LM_M(k, j, Ms, P, Y):
    r_LM = (-(1/j)*(Ms/P)) + (k/j)*Y
    return r_LM

rl_M = r_LM_M(k, j, Ms, P, Y)
```

```python
y_max = np.max(rl)
x_max = np.max(r)
                        
v = [80, 100, -30, 0]
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Effect of an increase in the amount of money", xlabel='Y', ylabel='r')
ax.plot(Y, r, scalex = True, scaley = True, color = 'black', label = "IS")
ax.plot(Y, rl, scalex= True, scaley = True, color = 'blue', label = "LM")
ax.plot(Y, rl_M, scalex= True, scaley = True, color = 'red',linestyle="--", label = "LM_M")

plt.text (81, -21, '→', fontsize = 30, color = 'green')
plt.text (96.5, -5, '→', fontsize = 30, color = 'green')

ax.vlines(x = 92.85, ymin = -30, ymax = -10.3, linestyle = "dotted", color = "grey")
ax.hlines(y = -10.6, xmin = 80, xmax = 92.85, linestyle = "dotted", color = "grey")
plt.text(89.7, -9, 'E', fontsize = 17, color = 'black')
plt.text(92.65, -9.75, 'E3', fontsize = 17, color = 'black')

plt.axis(v)
ax.legend()
plt.show()
```

In the case of applying an expansive monetary policy, what increases is the money supply in the economy ($M_0^s$). Which produces the following two effects:
- ↑ $M_0^s$ --> This creates excess supply in the money market: $\dfrac{M_0^s}{P_0}$ > $M^d$ --> To rebalance the system, the rate of interest is reduced (↓ $r$).: The LM curve shifts to the right as a result of this effect.
- The decrease in the interest rate (↓ $r$) generates an increase in investment (↑ $I0$) --> Consequently, aggregate demand also increases (↑ DA) --> Excess demand is produced: DA > $Y$ --> The production level of the economy must increase to return to equilibrium (↑$Y$): The IS curve does not shift because there is no change in the exogenous variables that affects it.

<center> Δ$Y$ = ($\dfrac{h}{kh + jβ_1}$) * $\dfrac{ΔM_0^s}{P_0}$ > 0 </center>


<center> Δ$r$ = ($\dfrac{β_1}{kh + jβ_1}$) * $\dfrac{ΔM_0^s}{P_0}$ < 0 </center>

### 4) Effects of an expansive Monetary Policy and restrictive Fiscal Policy

In this case we have a mixture of politics. On the one hand, an expansive monetary policy, which increases the money supply (↑ $M_0^s$). On the other hand, a restrictive fiscal policy, which reduces government spending (↓ $G0$). Consequently, the following effects occur:
1) Expansionary monetary policy
- ↑ $M_0^s$ --> Increase in nominal money supply: $\dfrac{$MM_0^s$}{$P_0$}$ --> This produces an imbalance in the money market due to excess supply : $\dfrac{$MM_0^s$}{$P_0$}$ > $M^d$ --> Therefore, the interest rate has to decrease to return to equilibrium (money has to become more expensive to reduce demand): ↓ $r$.: This effect moves the LM curve to the right.
- As a result of the reduction in the interest rate, investment increases (↑ $I_0$) --> Consequently, aggregate demand also increases (↑ DA) --> Excess demand is produced: DA > 𝑌 --> The production level of the economy must increase to return to equilibrium (↑ 𝑌)

2) Restrictive fiscal policy
- ↓ $G0$ --> Aggregate demand decreases (↓ DA) --> Which produces an excess supply in the goods market, generating an imbalance: DA < Y --> Therefore, the level of production in the economy must decrease to return to equilibrium (↓ $Y$): The IS curve shifts to the left.
- As a result of the decrease in the level of production, the demand for money in the market decreases: ↓ $M^d$. --> Consequently, there is an imbalance in the money market: $M^d$ < $\dfrac{$MM_0^s$}{$P_0$}$ --> To rebalance the market, the rate interest decreases (↓ $r$) to encourage the demand for money.

### 5) Tax rate reduction ($t$)

```python
t = 0.3

def r_IS_t(C0, b, m, t, I0, G0, X0, h, Y):
    r_IS = (C0 + I0 + G0 + X0 - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS_t(a, b, m, t, I0, G0, X0, h, r):
    Y_IS = (C0 + I0 + G0 + X0 - (h*r))/(1-(b-m)*(1-t))
    return Y_IS

r_t = r_IS_t(C0, b, m, t, I0, G0, X0, h, Y)
```

```python
y_max = np.max(rl)
x_max = np.max(r)
                        
v = [80, 100, -30, 0]
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Effect of a reduction in Tax rate", xlabel='Y', ylabel='r')
ax.plot(Y, r, scalex = True, scaley = True, color = 'black', label = "IS")
ax.plot(Y, r_t, scalex = True, scaley = True, color = 'red',linestyle="--", label = "IS_t")
ax.plot(Y, rl,scalex= True, scaley = True, color = 'blue', label = "LM")

plt.text (96, -10.8, '↑', fontsize = 25, color = 'green')

ax.vlines(x = 92.1, ymin = -30, ymax = -8.1, linestyle = "dotted", color = "grey")
ax.hlines(y = -8.1, xmin = 80, xmax = 92.1, linestyle = "dotted", color = "grey")
plt.text(89.7, -9.4, 'E', fontsize = 17, color = 'black')
plt.text(91.8, -7.3, 'E5', fontsize = 17, color = 'black')

plt.axis(v)
ax.legend()
plt.show()
```

In this case, we are dealing with a case of reduction of the tax rate (↓ $t$). The effects that this generates are described below:
- ↓ $t$ --> This leads to higher tax collection (↓ $T$) --> This generates a drop in disposable income (↓ $Y_d$). This last fall produces two effects.

   a) On the one hand, it causes consumption to fall (↓ $C0$), then aggregate demand (↓ DA), which produces an imbalance leading  to a reduction in the level of production to return to equilibrium: ↓ $Y$.
   
   b) On the other hand, the reduction in disposable income generates a fall in imports (↓ M), which increases net exports (↑ XN) leading to an increase in aggregate demand (↑ DA) for which the level of production has to increase: ↑ $Y$.
   
   
- Normally, the effect that predominates is a) over b), so this scenario is assumed for the exercise. Consequently, the slope of the IS curve decreases, that is, the curve rotates upwards.
- ↑ $Y$ --> generates an increase in the demand for money in the market (↑ $M^d$). --> Which generates an imbalance in the money market: $M^d$>$\dfrac{M_0^s}{P_0}$ --> As the monetary mass is fixed, what moves to adjust the market is the interest rate . Therefore, the interest rate increases to compensate for the excess demand (↓ $r$).: The LM curve does not shift because there was no change in any exogenous variable.
