# Equilibrium in the IS-LM model

#1. Equation of the IS Curve:
    
$$r = \frac{B_o}{h} - \frac{B_1}{h}Y $$

Where $B_0 = C_o + I_o + G_o + X_o$ and  $B_1 = 1 - (b - m)(1 - t)$

#2. Equation of the LM Curve:

$$  r = -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y $$

To find the equilibrium level of Income $ (Y^e) $, interest rates of each model are equalized $ (r) $:

$$ -\frac{1}{j}\frac{Mo^s}{P_o} + \frac{k}{j}Y = \frac{B_o}{h} - \frac{B_1}{h}Y $$

$$ Y^e = \frac{j B_o}{k h + j B_1} + (\frac{h}{k h + j B_1})\frac{Ms_o}{P_o} $$

To find the equilibrium Interest Rate $ (r^e) $, the levels of Income $ (Y) $ of each model are equalized:

$$ r^e = \frac{kB_o}{kh + jB_1} - (\frac{B_1}{kh + jB_1})\frac{Ms_o}{P_o} $$

These represent the equilibrium of the IS-LM model.

```python
# IS Curve
Y_size = 100 
b = 0.3                 #propensión marginal a consumir
m = 0.3                 #propensión a importar
t = 0.5                   #tasa impositiva
Co = 35                 #consumo autónomo
Io = 20                  #inversión autónoma
Go = 55                  #gasto gobierno 
Xo = 4                  #exportación 
h =  2                  #constante de decisión a invertir

Y = np.arange(Y_size)
```

```python
def r_IS(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r = r_IS(b, m, t, Co, Io, Go, Xo, h, Y)
r
```

```python
# LM Curve
Y_size = 100

k = 15                # Transaction money demand
j = 10                # Speculation money demand
Ms = 9000             # Nominal money supply
P  = 20                # Price level

Y = np.arange(Y_size)
```

```python
def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
i
```

```python
y_max = np.max(i)

fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS-LM Model", xlabel= r'Y', ylabel= r'r')

ax.plot(Y, r, "k-", label = "IS", color = "blue")
ax.plot(Y, i, "b-", label="LM", color = "red")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())  

plt.axvline(x=51.1,  ymin= 0, ymax= 0.52, linestyle = ":", color = "black")
plt.axhline(y=32, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")
plt.plot(51, 32, marker="o", color="black", label = "Eq")
plt.text(49,38, 'Eq', fontsize = 14, color = 'black')

ax.legend()
plt.show()
```

# Expansionary Monetary Policy and Restrictive Fiscal Policy

```python
#Expansionary Monetary Policy: increase in Money Supply (M^s)
Ms = 13000

def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
i_Ms
```

```python
#Restrictive Fiscal Policy: reduction in Goverment Supply (Go)
Go = 20
def r_IS_Go(b, m, t, Co, Io, Go, Xo, h, Y):
    r_IS = (Co + Io + Go + Xo - Y * (1-(b-m)*(1-t)))/h
    return r_IS

def Y_IS_Go(b, m, t, Co, Io, Go, Xo, h, r):
    Y_IS = (Co + Io + Go + Xo - h*r)/(1-(b-m)*(1-t))
    return Y_IS

r_Go = r_IS_Go(b, m, t, Co, Io, Go, Xo, h, Y)
r_Go
```

```python
y_max = np.max(r)
x_max = Y_IS(b, m, t, Co, Io, Go, Xo, h, 0)

v = [0, x_max, 0, y_max]                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Effects of Expansionary Monetary Policy and Restrictive Fiscal Policy", xlabel= 'Y', ylabel= 'r')

ax.plot(Y, r, "k-", label = "IS", color = "black")
ax.plot(Y, r_Go, "k-", label = "IS_G", color = "orange")
ax.plot(Y, i, "b-", label="LM", color = "grey")  
ax.plot(Y, i_Ms, "b-", label="LM_Ms", color = "skyblue")  

plt.axvline(x=51.1,  ymin= 0, ymax= 0.57, linestyle = ":", color = "black")
plt.axhline(y=32, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")
plt.plot(51, 32, marker="o", color="black", label = "Eo")
plt.text(49,38, 'Eo', fontsize = 14, color = 'black')

plt.axhline(y=14.5, xmin= 0, xmax= 0.52, linestyle = ":", color = "black")
plt.plot(52, 14.5, marker="o", color="blue", label = "E_1")
plt.text(49,16, 'E_1', fontsize = 14, color = 'blue')

plt.text(65, 44, '→', fontsize=15, color='black')
plt.text(65, 6, '←', fontsize=15, color='black')

ax.legend()
plt.show()
```
