# **The Keynesian Income - Expenditure model**

## 1. The equations of the Income-Expenditure model and the Aggregate Demand function:

*Consumption function:* $C = C_o + bY^d$  
*Invenstment function:* $I = I_o - hr$  
*Government spending:* $G = G_o$  
*Taxation:* $T = tY$  
*Exports:* $X = X_o$  
*Imports:* $M = mY^d$ 

###  *Find the Equilibrium equation for the Aggregate Income. Please, write all the equations in steps.*

$$Aggregate Demand (DA) = C + I + G + X - M$$

$DA = C_o + bY^d + I_o - hr + G_o + X_o - mY^d$

$DA = C_o + I_o + G_o + X_o - hr + bY^d - mY^d$

$DA = C_o + I_o + G_o + X_o - hr + (b-m)Y^d$

If:  
$Y^d = Y - T$  
$Y^d = Y - tY$  
$Y^d = (1 - t)Y$  
  
Then:  
$$DA = C_o + I_o + G_o + X - hr + (b-m)(1-t)Y$$


*Based on the equilibrium condition:*
  $$Y = DA$$

$Y = C_o + I_o + G_o + X - hr + (b-m)(1-t)Y$  
  
$Y - (b-m)(1-t)Y = C_o + I_o + G_o + X - hr$  
  
$Y[1-(b-m)(1-t)] = C_o + I_o + G_o + X - hr$  

**Equilibrium equation for the Aggregate Income**

$$Y =\frac{1}{1-(b-m)(1-t)} * (C_o + I_o + G_o + X - hr)$$

Where:  
  
$$\frac{1}{1-(b-m)(1-t)} = K$$  
  
$K = Keynesian  Multiplier$  
  
$$ Y = K*(C_o + I_o + G_o + X - hr)$$

### *Show the derivation of the IS curve from equilibrium Y=DA. Explain using intuition, Math, and a Figure.*

*Consumption function:* $C = C_o + bY^d$ 
  
*Invenstment function:* $I = I_o - hr$  
  
*Government spending:* $G = G_o$  
  
*Taxation:* $T = tY$  
  
*Exports:* $X = X_o$  
  
*Imports:* $M = mY^d$  
  
*Aggregate demand:* $DA=C+I+G+X-M$  
  
*Equilibrium condition:* $DA=Y$

$$Y=DA$$

$DA = C_o + bY^d + I_o - hr + G_o + X_o - mY^d$  
  
$DA = C_o + I_o + G_o + X_o - hr + bY^d - mY^d$  
  
$DA = C_o + I_o + G_o + X_o - hr + (b-m)Y^d$

If:  
$Y^d = Y - T$  
  
$Y^d = Y - tY$  
  
$Y^d = (1 - t)Y$  
  
Then:  
$$DA = C_o + I_o + G_o + X - hr + (b-m)(1-t)Y$$


$Y = C_o + I_o + G_o + X - hr + (b-m)(1-t)Y$  
  
$Y - (b-m)(1-t)Y = C_o + I_o + G_o + X - hr$  
  
$Y[1-(b-m)(1-t)] = C_o + I_o + G_o + X - hr$  
  
$$Y =\frac{1}{1-(b-m)(1-t)} * (C_o + I_o + G_o + X - hr)$$  

$$IS$$  
  
$[1-b(1 - t)] * Y = C_o + I_o + G_o - hr$
  
$hr = (C_o + I_o + G_o) - [1-b(1 - t)]Y$  
  
$$ r = \frac{C_o + I_o + G_o}{h} - \frac{[1-b(1 - t)]Y}{h}$$

```python
# Parameters
Y_size = 100
Co = 20             # Autonomous consumption
b = 0.2             # Marginal propensity to consume
m  = 0.2            # Marginal propensity to import
t = 1               # Taxes
Io = 10             # Investment intercept (when i = 0)
Go = 8              # Government spending
Xo = 2              # Exports (given)
h = 5               # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

# Define and populate THE IS-Schedule
def i_IS(Co, b, m, t, Io, Go, Xo, h, Y):
    i_IS = (Co + Io + Go + Xo - Y*(1-(b-m)*(1-t)))/h
    return i_IS

def Y_IS(Co, b, m, t, Io, Go, Xo, h, r):
    Y_IS = (Io + Co + Go + Xo-r*h)/(1-(b-m)*(1-t))
    return Y_IS

r = i_IS(Co, b, m, t, Io, Go, Xo, h, Y)

# PLOT THE IS-SCHEDULE
y_max = np.max(r)
x_max = Y_IS(Co, b, m, t, Io, Go, Xo, h, 0)

v = [0, x_max, 0, y_max]                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r, "k-")
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())   
plt.axis(v)                                     
plt.show()
```

### *Derive the IS curve from the Save-Investment Equilibrium.*

$$Y=DA$$

$$Y = C + I + G + X - M$$

*Subtract Taxes (T) to the two sides of the ecuation:*  
  
$Y - T = C + I - T + G + X - M$  
  
$Y^d = C + I - T + G + X - M$  

$$(Y^d - C) + (T-G) + (M-X) = I$$ 
  
$I = Io - hr$  


*S = Total saving = Private saving + Government's saving + Extern's saving*
  
$S_p = Y^d - C$  
  
$$S_p = Y^d -(C_o + bY^d)$$  
  
$S_g = T - G$      
  
$S_e = M - X$  
    
$$S_e = mY^d - X_o$$

$$S = I$$  
  
$$S_p + S_g + S_e = I_o - hr$$
   
   

$$Y^d - (C_o + bY^d) + T - G + mY^d - X_o = I_o - hr$$  
  
$ [1 - (b - m)(1 - t)]Y - (C_0 + G_0 + X_0) = I_0 - hr $  
  
$ hr = (C_0 + G_0 + I_0 + X_0) - (1 - (b - m)(1 - t))Y $
  
$ r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - \frac{1 - (b - m)(1 - t)}{h}Y $
  
$$ r = \frac{B_0}{h} - \frac{B_1}{h}Y $$

Where $ B_0 = C_0 + G_0 + I_0 + X_0  $ and $  B_1 = 1 - (b - m)(1 - t) $
  

### *Show the derivation of the LM curve from equilibrium $M^s = M^d$ . Explain using intuition, Math, and a Figure.*

$$M^s = M^d$$  
  
$M^s = \frac{M_s^o}{P}$
  
$M^d = kY - ji$  
  
*If:*
  
$r=i-π$  
  
$π = 0$  
  
$r = i$  
  
*Then:*  
  
$M^d = kY - jr$  

$$\frac{M_s^o}{P} = ky - jr$$  
  
    
$$\frac{M_s^o}{P} - ky = - jr$$  
  
    
$$r = \frac{-1}{j} *\frac{(M_s^o)}{P} + \frac{k}{j}Y $$  

```python
# Parameters
Y_size = 100
k = 10                 # Transaction money demand
j = 10                 # Speculation money demand
Ms = 20000             # Nominal money supply
P  = 20                # Price level

Y = np.arange(Y_size)

# Define and populate THE LM-Schedule
def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
i

y_max = np.max(i)

v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="LM SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i, "k-", label="LM")
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

ax.legend()
plt.show()
```

## 1. Use the next Equations and find the equilibrium

## *Find the equilibrium equation of $Y^e$, $P^e$ and $r^e$.*

- **Equilibrium equation of $Y^e$**

$IS:$ 
  
$[1-b(1 - t)]Y = C_o + I_o + G_o - hr$  
  
$hr = (C_o + I_o + G_o) - [1-b(1 - t)]Y$  
  
$ r = \frac{C_o + I_o + G_o}{h} - \frac{[1-b(1 - t)]Y}{h}$

$LM:$  
  
$\frac{M_o^s}{P_o} = kY - jr$  
  
$jr = kY - \frac{M_o^s}{P_o}$  
  
$r = \frac{kY}{j} - \frac{M_o^s}{P_o*j}$

$$\frac{C_o + I_o + G_o}{h} - \frac{[1-b(1 - t)]Y}{h} = \frac{kY}{j} - \frac{M_o^s}{P_o*j}$$  
  
$\frac{C_o + I_o + G_o}{h} + \frac{M_o^s}{P_o*j} = \frac{kY}{j} + \frac{[1-b(1 - t)]Y}{h}$  
  
$\frac{C_o + I_o + G_o}{h} + \frac{M_o^s}{P_o*j} =  Y * (\frac{k}{j} + \frac{1-b(1 - t)}{h}$  
  
$ \frac{C_o + I_o + G_o}{h} + \frac{M_o^s}{P_o*j} = Y * (\frac{kh + [1-b(1 - t)]j}{hj})$  
  
$ (\frac{C_o + I_o + G_o}{h} + \frac{M_o^s}{P_o*j}) * (\frac{hj}{kh + [1-b(1 - t)]j})  = Y$  
  
$Y = \frac{(C_o + I_o + G_o)hj}{h*(kh + [1-b(1 - t)]j} +  \frac{M_o^s*h}{P_o*kh + [1-b(1 - t)]j}$  
  
$$ Y^e= \frac{(C_o + I_o + G_o)j}{(kh + [1-b(1 - t)]j} + \frac{h}{kh + [1-b(1 - t)]j} * \frac{M_o^s}{P_o^s} $$




- **Equilibrium equation of $P^e$**

$B_0 = C_o + I_o + G_o$  
  
$B_1 = \frac{[1-b(1 - t)]Y}{h}$  

$IS:$  
  
  $ r = \frac{B_o}{h} - \frac{B_1Y}{h}$


$LM:$  
  
$r = \frac{kY}{j} - \frac{M_o^s}{P_o*j}$

$$\frac{B_o}{h} - \frac{B_1Y}{h} = \frac{kY}{j} - \frac{M_o^s}{P_o*j}$$  
  
$\frac{B_o}{h} - \frac{B_1Y}{h} - \frac{kY}{j} = - \frac{1}{P_o} * \frac{M_o^s}{j}$  
  
$(\frac{B_o}{h} - \frac{B_1Y}{h} - \frac{kY}{j}) * \frac{j}{M_o^s} =  - \frac{1}{P_o}$  
  
$ (\frac{(B_o - B_1Y)j - khY}{hj} * \frac{j}{M_o^s} = - \frac{1}{P_o}$  
  
$ - P =  \frac{M_o^s h}{jB_o - (jB_1 + hk)Y}$  
    
$$ P^e =  \frac{M_o^s h}{- jB_o + (jB_1 + hk)Y}$$  
  
$$ P^e =  \frac{M_o^s h}{- j(C_o + I_o + G_o) + (j\frac{[1-b(1 - t)]Y}{h}+ hk)Y}$$ 




- **Equilibrium equation of $r^e$**

$IS:$  
  
  $ r = \frac{B_o}{h} - \frac{B_1Y}{h}$
  
$ rh = B_o - B_1Y$    
  
$$Y= \frac{B_o - rh}{B_1Y}$$
  
$LM:$  
  
$r = \frac{kY}{j} - \frac{M_o^s}{P_oj}$  
  
$r =  \frac{1}{j}(-\frac{M_o^s}{P} + Y)$  
  
$rj =  (-\frac{M_o^s}{P} + Y)$  
  
$$Y = rj + \frac{M_o^s}{P}$$ 


$$\frac{B_o - rh}{B_1Y} = rj + \frac{M_o^s}{P}$$  
  
$B_o - rh =  B_1Y (rj + \frac{M_o^s}{P}$  
  
$B_o - rh =  B_1Yrj + \frac{B_1 Y M_o^s}{P}$   
    
$B_o - \frac{B_1 Y M_o^s}{P} =  B_1Yrj + rh$   
    
$B_o - \frac{B_1 Y M_o^s}{P} =  r(B_1Yj + h)$
  
$$ r = \frac{1}{B_1Yj + h}(B_o - \frac{B_1 Y M_o^s}{P})$$

### 2. Now you have to use the model below. But now the interest rate $r$  and $p$ are exogenous. Because $r$ is the instrument of monetary policy. The endogenous variables of this model are $Y$ and  $M_o^s$.

$Y = C + I + G + X - M$  
  
$C = C_o + b(1-t)Y$  
  
$I = I_o - hr$  
  
$G = tY$  
  
$X = X_o$  
  
$M = mY$

#### *Find the equilibrium equation of $Y$ and $M_o^s$.*

- Equilibrium of $Y$

$$Y = C + I + G + X - M$$  
  
$Y = C_o + b(1-t)Y + I_o - hr + tY + X_o - mY$  
  
$Y = C_o + I_o + X_o - hr + (b(1-t) + t - m)Y$  
    
$Y - (b(1-t) + t - m)Y = C_o + I_o + X_o - hr$  
  
$Y(1 - (b(1-t) + t - m)) = C_o + I_o + X_o - hr$  
  
$$Y = \frac{C_o + I_o + X_o - hr}{1 - (b(1-t) + t - m)}$$   



- Equilibrium of $M_o^s$

$\frac{M_o^s}{P_o} = kY - jr$  
  
$$M_o^s = (kY - jr)P_o$$

```python
"EQUILIBRIUM IN THE IS-LM MARKET"
# IS - Parameters
Y_size = 100
C = 600                 # Autonomous consumption
b = 0.01               # Marginal propensity to consume
m = 5                # Marginal propensity to import
t = 0.1                # Tax rate
I = 10                 # Investment 
X = 20                  # Exports 
h = 1                  # Investment slope wrt to i  
r = 50                #Interest rate

r_1 = 30 
# Arrays
Y = np.arange(Y_size)       # Array values of Y

# LM - Parameters
k = 20                      # Transaction money demand
j = 10                      # Speculation money demand
P  = 1                     # Price level


# DEFINE AND POPULATE THE IS AND LM SCHEDULE
def i_IS(C, b, m, t, I, X, h, r, Y):
    i_IS = C + I + X - (h*r) + (b - (b*t) + t - m)*Y
    return i_IS

i = i_IS(C, b, m, t, I, X, h, r, Y)

def i_IS_1(C, b, m, t, I, X, h, r_1, Y):
    i_IS_1 = C + I + X - (h*r_1) + (b - (b*t) + t - m)*Y
    return i_IS_1

i_2 = i_IS_1(C, b, m, t, I, X, h, r_1, Y)

def i_LM( k, j, r, P, Y):
    i_LM = (k*Y - j*r)*P
    return i_LM

i_1 = i_LM( k, j, r, P, Y)

def i_LM_1( k, j, r_1, P, Y):
    i_LM_1 = (k*Y - j*r_1)*P
    return i_LM_1

i_3 = i_LM( k, j, r_1, P, Y)

# PLOT THE IS-SCHEDULE
y_max = np.max(i)

fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS-LM Model", xlabel= r'Y', ylabel= r'r')

ax.plot(Y, i, "b-", label = "IS_1")
ax.plot(Y, i_1, "r-", label="LM_1")
ax.plot(Y, i_2, "k-", label = "IS")
ax.plot(Y, i_3, "k-", label="LM")

plt.axhline(y=355, xmin= 0, xmax= 1, linestyle = ":", color = "skyblue", label = "r_1")
plt.axhline(y=410, xmin= 0, xmax= 1, linestyle = ":", color = "grey", label = "r")

plt.plot(43.5, 370, marker="o", color="blue", label = "Eo")
plt.plot(36, 425, marker="o", color="black", label = "Eo_1")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())  

ax.legend()
plt.show()
```

- $Intuition:$   
On the one hand, if $r$ is decreased by the central bank, investments will increase. As a result, the aggregate demand will increase and become higher than the production. Thus the production have to increase.     
On the other hand, the decreased of $r$ will make the demand of money higher than the demand. In this markent, what the central bank will do is to increase the real money supply to match the excess demand for money in the economy, that is why we can see an expansion in the LM curve.
Overall, we can see in the graphic an expansion in both the LM and the IS curve in which the production increases and the interest rate is lower.  
  
- $Math:$   
  
    $∆r → r↓ → I↑ → DA↑ → DA > Y → Y↑$
  
    $∆r → r↓ → M^d↑ → M^d > M^o → M^o↑$
