### We have the aggregate demand. "DA" fuction:

$$ DA = C + I + G + (X - M) $$

It refers:

- ### $C$: Consumption function: $C_{0}$$+$$b$$y^{d}$

- ### $I$: Investment function: $I_{0}$$-$$h_{r}$

- ### $G$: Government spending: $G_{0}$

- ### $X$ : exports ($X_{0}$)

- ### $M$ : imports ( $m$$y^{d}$). 

As $"DA"$

$DA$=$C_{0}$+$b$$(y-ty)$+$I_{0}$-$h_{r}$+$G_{0}$+$X_{0}$-$m(y-xy)$

$DA$=$(C_{0}+I_{0}+G_{0}+X_{0}-h_{r})$+$(b-m)(1-t)Y$

$DA$=$α_{0}$+$α_{1}Y$ 

where:

- $α_{0}$ is the intercept


- $α_{1}$, the slope

### The Equilibrium equation for the Aggregate Income

$$DA=Y$$

- $Y$=$(C_{0}+I_{0}+G_{0}+X_{0}-h_{r})$+$(b-m)(1-t)Y$


- $Y$-$(b-m)(1-t)Y$=$[C_{0}+I_{0}+G_{0}+X_{0}-h_{r}]$


- $[1-(b-m)(1-t)]Y$=$[C_{0}+I_{0}+G_{0}+X_{0}-h_{r}]$


- $Y$=$[1-(b-m)(1-t)]^{-1}$$[C_{0}+I_{0}+G_{0}+X_{0}-h_{r}]$ 

Substitute values:

$Y$=$[\frac{1}{1-(b-m)(1-T)}]$$[C_{0}+I_{0}+G_{0}+X_{0}-h_{r}]$
