# I) The Keynesian Income-Expenditure model

## 1.-

### Consumption function: $C$$=$$C_{0}$$+$$b$$y^{d}$

It can also be represented as follows: $C$=$C_{0}$+$b(1-t)y$

$C_{0}$ is the level of autonomous consumption (indepenent of the level of income); $b$ is the marginal propensity to consume; and $t$ is the dollar-amount of taxes.

### Investment function: $I$$=$$I_{0}$$-$$h_{r}$

$I_{0}$ represents autonomous investment; and $h_{r}$ represents the interest rate. Both have an inverse relationship.

### Government spending: $G$$=$$G_{0}$

$G_{0}$ represents autonomous spending.

### Taxation (not included in the aggregate demand formula): $T$$=$$tY$

### Export: $X$$=$$X_{0}$

$X_{0}$ represents autonomous export.

### Import: $M$$=$$m$$y^{d}$

$m$ is the marginal propensity to import.

### The aggregate demand 

Let's represent it as "$DA$"...

$DA$=$C$+$I$+$G$+($X$−$M$)

$DA$=$C_{0}$+$b$$(y-ty)$+$I_{0}$-$h_{r}$+$G_{0}$+$X_{0}$-$m(y-xy)$

$DA$=$(C_{0}+I_{0}+G_{0}+X_{0}-h_{r})$+$(b-m)(1-t)Y$

$DA$=$α_{0}$+$α_{1}Y$ 

$α_{0}$ is the intercept; and $α_{1}$, the slope.

### On the way to a perfect balance: the equilibrium economy formula

Let's represent it as "$Y$"...

$Y$=$(C_{0}+I_{0}+G_{0}+X_{0}-h_{r})$+$(b-m)(1-t)Y$

$Y$-$(b-m)(1-t)Y$=$[C_{0}+I_{0}+G_{0}+X_{0}-h_{r}]$

$[1-(b-m)(1-t)]Y$=$[C_{0}+I_{0}+G_{0}+X_{0}-h_{r}]$

$Y$=$[1-(b-m)(1-t)]^{-1}$$[C_{0}+I_{0}+G_{0}+X_{0}-h_{r}]$ 

The part of the equation "$[1-(b-m)(1-t)]^{-1}$" is called the Keynesian multiplier, which must be bigger than zero, and can be represented as $k$.

## $Y^{e}$ and $DA^{e}$ represent in the next graphic the function of Aggregate Demand...

```python
"1|IMPORT PACKAGES"
import numpy as np               
import matplotlib.pyplot as plt  
Y_size = 10
c1 = 80              
c2 = 25                
c3 = 11                
Ms = 60             
P  = 23                
Y = np.arange(Y_size)  
def i_LM(c1, c2, c3, Ms, P, Y):
    i_LM = (c1 - Ms/P)/c3 + c2/c3*Y
    return i_LM
i = i_LM(c1, c2, c3, Ms, P, Y)
y_max = np.max(i)
v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="The function of Aggregate Demand", xlabel=r'Y', ylabel=r'DA')

plt.text(7, 18, 'DA=α_0+α_1(Y)', fontsize=20, color='black')

ax.plot(Y, i, "k-")
plt.axis(v)  
plt.show()
```

## 1.-

## The equilibrium in the IS-LM model

### IS

$r$=$β_{0}$*$h^{-1}$-$(β_{1}$*$h^{-1})$$y$

### LM

$r$=$-1$*$j^{-1}$*$M_{0}^{s}$$(P_{0})^{-1}$+$K(j)^{-1}$$y$

Considering the proposed equations, it is possible to replace some variables to find $y^{e}$ and $r^{e}$, which represent the equilibrium of the IS-LM model...

$-1$*$j^{-1}$*$M_{0}^{s}$$(P_{0})^{-1}$+$K(j)^{-1}$=$β_{0}$*$h^{-1}$-$(β_{1}$*$h^{-1})$$y$

$y^{e}$=$j$$(β_{0})(Kh+β_{1})^{-1}$+$h$($Kh$+$β_{1})^{-1}$$(M_{0}^{s}(P_{0})^{-1})$

$r^{e}$=$j$$(β_{0})(Kh+β_{1})^{-1}$+$β_{1}$($Kh$+$β_{1})^{-1}$$(M_{0}^{s}(P_{0})^{-1})$
