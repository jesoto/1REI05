C) Derivation of the LM curve from equilibrium M^s = M^d

* Real money supply:

\begin{align}
    \frac {M^s}{P}\ = \frac {M_0^S}{P} Y\ 
\end{align}

* Real money demand:

$$\ M^d = kY âˆ’ ji \$$

* Equilibrium:

$$\ M^s = M^d \$$

\begin{align}
    \frac {M^s}{P}\ = \ kY âˆ’ ji \ 
\end{align}

$$\ M^s =  P(kY âˆ’ ji) \$$

* The LM schedule 

\begin{align}
    \ r &= -\frac{1}{j}\frac {M_0^s}{P_0}\ + \frac {k}{j} Y\ 
\end{align}

Where: \begin{align}
    -\frac{1}{j}\frac {M_0^s}{P_0}\
\end{align} is the intercept and 

\begin{align}
    \frac {k}{j} Y\ 
\end{align} is the slope


```python
# Parameters
Y_size = 100

k = 10                # Transaction money demand
j = 10                # Speculation money demand
Ms = 20000             # Nominal money supply
P  = 20                # Price level

Y = np.arange(Y_size)

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)
i

y_max = np.max(i)

v = [0, Y_size, 0, y_max]                       
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Derivation of the LM curve from equilibrium M^s= M^d", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, i, "b-", label="LM", color = 'black')
ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.text(99, -8.7, 'LM', fontsize=13)

plt.axvline(x=20,  ymin= 0, ymax= 0.22, linestyle = ":", color = "black")
plt.axhline(y=-80, xmin= 0, xmax= 0.23, linestyle = ":", color = "black")  

plt.axvline(x=40,  ymin= 0, ymax= 0.41, linestyle = ":", color = "black")
plt.axhline(y=-60, xmin= 0, xmax= 0.41, linestyle = ":", color = "black")

plt.axvline(x=60,  ymin= 0, ymax= 0.60, linestyle = ":", color = "black")
plt.axhline(y=-40, xmin= 0, xmax= 0.60, linestyle = ":", color = "black")

plt.plot(20, -80, marker="o", color="red", label = "Eo")
plt.text(19,-75, 'A', fontsize = 14, color = 'red')

plt.plot(40, -60, marker="o", color="red", label = "E_1")
plt.text(39, -55, 'B', fontsize = 14, color = 'red')

plt.plot(60, -40, marker="o", color="red", label = "E_1")
plt.text(59, -35, 'C', fontsize = 14, color = 'red')

plt.show()
```
