# a. 

To find the equilibrium equation for Aggregate Income, we must first assume the conditions established by the Income-Expenditure model. Therefore we have the following two equilibria:

- The value of the total Savings will be equivalent to the value of the Investment
- Aggregate Demand will have equivalence with Aggregate Income

From this equivalence, it is assumed:

 $$ \ DA = Y\ $$ 

$$ \ Y=C+I+G+X-M\ $$

Where: 
 - $Consumption function$: $$ \ C= C_0 + b*Y^d\ $$
$C_o$ is the amount basic of autonomous consumption (indepenent of the income thar evereyone receive),  $Y^d$ is the incomes available. Finally, $b$ is the marginal propensity to consume and t is the dollar-amount of taxes. Both $b$ and $Y$ have a value between 0 and 1.

 - $Inversion function$:  $$ \ I = Io − hr\ $$
$I_o$ is the autonomous inversion.There is a negative relationship, since $r$ is the real interest rate. So, during the increase in $r$, the autonomous investment decreases. 

 - $Government spending$: $$ \ G = Go\ $$ 
$G_o$, is the expenditure of the autonomous government. The money the government spends comes from taxes.

 - $Exportation$:  $$ \ X = Xo\ $$
 
 - $Importation$:  $$ \ M = m*Y^d\ $$

Before finding the equilibrium equation for Aggregate Income, we will first find the equation for Aggregate Demand.

The Aggregate Demand equation is as follows:

$$ \ DA = C+I+G+X-M\ $$

First, each element of the Aggregate Demand will be replaced by its functions, shown above:

$$ \ DA = Co + bY^d + Io - hr + Go + Xo − mY^d \ $$

The variables that are affected by $Y^d$ will move to the right. Additionally, all autonomous variables will be placed on the left

$$ \ DA = Co + Io - hr + Go + Xo + (b-m)(1-t)Y \ $$

In short, the Aggregate Demand formula would be:
$$ \ DA = α_0 + (α_1)*Y\ $$

Since the Aggregate Income, Y, must be equal to the Aggregate Demand $DA$. This equation is obtained:

$$ \ Y = Co + Io - hr + Go + Xo + (b-m)(1-t)Y \ $$ or

$$ \ Y = α_0 + (α_1)*Y\ $$
 

Then, the equilibrium Aggregate Income equation:

$$ \ Y - (b-m)(1-t)Y = (Co + Io - hr + Go + Xo) \ $$
$$ \ [1 - (b-m)(1-t)] Y= (Co + Io - hr + Go + Xo) \ $$

\begin{align}
    \ Y &= \frac {1 }{[1 - (b-m)(1-t)]}\ (Co + Io - hr + Go + Xo)
\end{align}

Derivation of the IS curve from equilibrium $Y=DA$

$$ \ Y = DA\ $$

$$ \ Y = C+I+G+X-M\ $$

Differentiating in terms of $Y$:

\begin{align}
  Y &=\frac {1}{[1- (b -m) (1-t)]}* (C_o + I_o + G_o+ X_o) \ - \frac {[1- (b -m) (1-t)]}{h}* r
\end{align}

Or differentiating in terms of $r$

\begin{align}
  r &=\frac {1}{h}* (C_o + I_o + G_o+ X_o) \ - \frac {[1- (b -m) (1-t)]}{h}* Y
\end{align}

```python
import numpy as np               
import matplotlib.pyplot as plt 

"2|DEFINE PARAMETERS AND ARRAYS"
# Parameters
Y_size = 100
C0 = 20                 # Autonomous consumption
b = 0.2                # Marginal propensity to consume
alpha = 5              # Autonomous imports
m = 0.2            # Marginal propensity to import
T = 1                  # Taxes
I0 = 10             # Investment intercept (when r = 0)
G0 = 8              # Government spending
X0 = 2              # Exports (given)
h = 5                  # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y
```

```python
"3|DEFINE AND POPULATE THE IS-SCHEDULE"
def r_IS(C0, alpha, b, m, T, I0, G0, X0, h, Y):
    r_IS = ((C0 + I0 + G0 + X0) - (1-(b-m)*(1-T))*Y)/h
    return r_IS

def Y_IS(Co, alpha, b, m, T, I0, G0, X0, h, r):
    Y_IS = (C0 + I0 + G0 + X0)/1-(b-m)*(1-T) - h*r/(1-(b-m)*(1-T))
    return Y_IS

r = r_IS(C0, alpha, b, m, T, I0, G0, X0, h, Y)

"4|PLOT THE IS-SCHEDULE"
y_max = np.max(r)
x_max = Y_IS(C0, alpha, b, m, T, I0, G0, X0, h, 0)

v = [0, x_max, 0, y_max]                        # Set the axes range
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Derivation of the IS curve from equilibrium Y=DA ", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r, "k-")

plt.axvline(x=25,  ymin= 0, ymax= 0.38, linestyle = ":", color = "black")
plt.axhline(y=3, xmin= 0, xmax= 0.62, linestyle = ":", color = "black")  

plt.axvline(x=15,  ymin= 0, ymax= 0.63, linestyle = ":", color = "black")
plt.axhline(y=5, xmin= 0, xmax= 0.38, linestyle = ":", color = "black")

plt.plot(15, 5, marker="o", color="black", label = "Eo")
plt.text(14.5,5.3, '$E_o$', fontsize = 14, color = 'black')

plt.plot(25, 3, marker="o", color="blue", label = "E_1")
plt.text(24.5, 3.3, '$E_1$', fontsize = 14, color = 'blue')
plt.axis(v)                                     # Use 'v' as the axes range
ax.legend()
plt.show()
```

# b.

To find the equation that models the IS equation, we must first understand the elements that make up savings. Total savings includes Private Savings, Government Savings and External Savings.

$$ \ S = Sp + Sg + Se\ $$

With the equivalence condition that the total savings is the same as the investment, the following equation is proposed.

$$ \ Sp + Sg + Se = Io − hr\ $$

Then we get the equation that models the IS curve
$$ \ S(Y) = I(r)\ $$

Replacing this equation with the equations and functions previously found, we obtain:

$$ \ Sp + Sg + Se = Io − hr\ $$
$$ \ (Yd - Co - bY) + (T - Go) + (mY^d - Xo) = Io − hr\ $$
$$ \ Y [1- (b -m) (1-t)]- (Co + Go + Xo)  = Io − hr\ $$
$$ \ hr = Y [1- (b -m) (1-t)]- (Co + Go + Xo + Io)\ $$

\begin{align}
  r &=\frac {1}{h}* [1- (b -m) (1-t)]\ - \frac {1}{h}* [1- (b -m) (1-t)]Y
\end{align}
 or
\begin{align}
  r &=\frac {β_o}{h} - \frac {β_1}{h}* Y

# 4. DA-OA Model

a. Aggregate Demand

Both the equation of the IS curve and the LM curve must be derived as a function of $r$

* $IS$:

\begin{align}
    \ [1-b (1-t)] Y &= C_o + I_o + G_o - hr\         
\end{align} 

\begin{align}
    \ r &= \frac {β_0}{h}\ - \frac {β_1}{h}Y\     (1)    
\end{align}    

* $LM$:

\begin{align}
    \frac {M^s_o}{P_o}\ &= kY - jr  
\end{align}

\begin{align}
    \ r &= - \frac {1}{j}\ \frac {M^s _ 0}{P_0}\ + \frac {K}{j} Y\  (2)
\end{align}


To linearize the curve, the function of $LM$ will be defined by logarithms:

\begin{align}
     M^s_0 - P_0 &= ky - jr 
\end{align}

So, we get

\begin{align}
   r &= - \frac {M^s _ 0 - P}{j}\ + \frac {K}{j} Y \ (3)
\end{align}

The equilibrium equation of $r^e$

\begin{align}
   \ r^e &= -  \frac {k β_0}{kh + j β_1} \ - \frac {h}{kh + j β_1}\ \frac {M^s _ 0}{P_0}\   (4)
\end{align}

Based on what was found, we will begin to obtain the equilibrium equation of $Y^e$

\begin{align}
   - \frac {1}{j}\ \frac {M^s _ 0}{P_0}\ + \frac {K}{j} Y \ &=\ \frac {β_0}{h}\ - \frac {β_1}{h}Y\ 
\end{align}

\begin{align}
   \ Y^e &= \frac {j β_0}{kh + j β_1} \ + \frac {h}{kh + j β_1}\ \frac {M^s _ 0}{P_0}\   (5)
\end{align}

Deriving r, in the Aggregate Demand equation, as a function of P

\begin{align}
P = \frac{h Mo^s + jB_o}{h} - \frac{jB_1 + hk}{h}Y  (5)
\end{align}

b. Aggregate Incomes

\begin{align}
P = P^e
\end{align}

\begin{align}
P^e = \frac{h Mo^s + jB_o}{h} - \frac{jB_1 + hk}{h}Y  (6)
\end{align}
