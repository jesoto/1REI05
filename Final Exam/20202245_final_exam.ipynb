```python
Consumption: C = C0 + bY^d
Investment:  I = I0 ‚àí hr
Government:  G = G0
Taxation:    T = tY
Export:      X = X0
Import:      M = mY^d
Expenditure: DA = C + I + G + X ‚àí M

Aggregate demand
ùê∑ùê¥ = C0 + bY^d + I0 ‚àí hr + G0 + X0 ‚àí mY^d
ùê∑ùê¥ =(b-m)(1-t)Y+C0 +I0 -hr+G0 +X0
ùê∑ùê¥ =(C0 +I0 +G0 +X0 ‚àíhr)+(b-m)(1-t)Y

Shortening the equation:
ùê∑ùê¥ = ùõº0 + ùõº1Y,
Œ±0 =(C0 +I0 +G0 +X0 ‚àíhr)---intercept
Œ±1 =(b-m)(1-t)-------------slope 
```

```python
Equilibrium equation for the Aggregate Income:
Y = DA
ùëå = (C0 + I0 + G0 + X0 ‚àí hr) + (b - m)(1 - t)Y
Y - (b - m)(1 - t)Y = [C0 + I0 + G0 + X0 - hr]
[1 - (b-m)(1-t)]Y = [C0 + I0 + G0 + X0 - hr]

Y* = Equilibrium equation
Y* = (1/[1 - (b-m)(1-t)]) [C0 + I0 + G0 + X0 - hr]
K = (1/[1 - (b-m)(1-t)]) = keynesian multiplier
```

```python
Aggregate demand
IS : 
r = (√ü0/h) - (√ü1/h)Y

√ü0 = C0 + G0 + I0 + X0 
√ü1 = 1 - (b-m)(1-t)

LM:
r = - (1/j)(M0^s/P0)+(k/j)Y

Endogenous: r, Y
Exogenous: everything else
```

```python
Equalize:
- (1/j)(M0^s/P0)+(k/j)Y = (√ü0/h) - (√ü1/h)Y

Y^e = j√ü0/kh+j√ü1 + (h/kh+j√ü1)M0^s/P0

r = (√ü0/h) - (√ü1/h)(j√ü0/kh+j√ü1 + (h/kh+j√ü1)M0^s/P0)
r^e = k√ü0/kh+j√ü1 - (√ü1/kh+j√ü1)M0^s/P0

Y^e y r^e represent the equilibrium in the IS-LM model
The slope of the IS curve is negative and the slope
of the LM is positive.
```

```python
Without "r", P:
P = h(M0^s)/-j√ü0+(j√ü1+hk)Y

The aggregate demand equation
Y = j√ü0/j√ü1+hk + (h(M0^s)/j√ü1+hk)1/P
```
