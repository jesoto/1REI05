### I. The Keynesian Income-Expenditure model

#### 1) First model

##### (i) Equilibrium equation for the Aggregate Income

Considering, first, the Aggregate Demand equation:

<center> $DA = (C_0 + bY^d) + (I_0 - hr) + G_0 + X_0 - (mY^d)$


* $Y^d = (1-t)Y$

<center> $DA = (C_0 + b(1-t)Y) + (I_0 - hr) + G_0 + X_0 - m(1-t)Y$

<center> $DA = (C_0 + I_0 + G_0 + X_0 - hr) + b(1-t)Y - m(1-t)Y$

<center> $DA = (C_0 + I_0 + G_0 + X_0 - hr) + (b-m)(1-t)Y$

In this model, to find the equilibrium income, what is needed is to equalize everything that is demanded in the economy to everything that has been produced: 


<center> $Y = DA$

<center> $Y = C_0 + I_0 + G_0 + X_0 - hr + (b-m)(1-t)Y$

<center> $Y - (b-m)(1-t)Y = C_0 + I_0 + G_0 + X_0 - hr$

<center> $Y[1 - (b-m)(1-t)] = C_0 + I_0 + G_0 + X_0 - hr$

After equalizing the equations, we proceed to clear the $Y$ to find the Equilibrium equation for Aggregate Income:

<center> $Y^e$ = $\dfrac{1}{[1 - (b-m)(1-t)]}$ * $C_0 + I_0 + G_0 + X_0 - hr$

##### (ii) Figure

* Intuition

The IS curve is composed of the set of ordered pairs of $Y$ and $r$ that balance the goods market.

In the agreggate demanda (DA), given the equilibrium income $Y_0$ that corresponds to an interest rate equal to $r_0$, we assume the case that the interest rate increases to $r_1$, that is, $r_0<r_1$. This increase in the interest rate decreases the intercept of the $DA$ function, contracting the curve. Consequently, a lower equilibrium income $Y_1$ is generated, that is, $Y_1<Y_0$. These two ordered pairs, product of the increase of $r$ in DA, can be seen at points $Eo$ and $E1$ of the IS curve. 

#### 2) Derivation of the IS curve from the Save-Investment Equilibrium

The balance between income and aggregate demand is equivalent to the balance between total saving and investment.

<center> Total saving = $S_p + S_g + S_e$

* $S_p = Y^d - (C_0 + bY^d)$
* $S_g = T - G_0$
* $S_e = mY^d - X_0$

<center> $S_p + S_g + S_e = I_0 - hr$

<center> $Y^d - (C_0 + bY^d) + (T - G_0) + (mY^d - X_0) = I_0 - hr$

<center> $[1 - (b-m)(1-t)]Y - (C_0 + G_0 + X_0) = I_0 - hr$

<center> $hr = I_0 - [1- (b-m)(1-t)]Y + (C_0 + G_0 + X_0)$

Clearing r as a function of y, we obtain:

<center> $r$ = $\dfrac{1}{h}$$(C_0 + I_0 + G_0 + X_0)$ - $\dfrac{1 - (b-m)(1-t)}{h}$$Y$

$B_0 = C_0 + I_0 + G_0 + X_0$

$B_1 = 1 + (b-m)(1-t)$

<center> $IS:$ $r$ = $\dfrac{B_0}{h}$ - $\dfrac{B_1}{h}$$Y$

#### 3) Derivation of the LM curve from equilibrium $M^s=M^d$

* Math

<center> Taking into account that $M^s$ = $\dfrac{M^s}{P}$ and $M^d = kY - jr$ 

In equilibrium, all money demand has to be equal to money supply $(M^s=M^d)$:

<center> $\dfrac{M^s}{P}$ = $kY - jr$ 

<center> $jr$ = $kY$ - $\dfrac{M^s}{P}$

Clearing r as a function of y, we obtain:

<center> $LM:$ $r$ = - $\dfrac{1}{j}$$\dfrac{M^s}{P}$ + $\dfrac{k}{j}$$Y$

* Intuition

The LM curve is composed of the set of ordered pairs $r$ and $Y$ that balance the money market.

When income $Y$ increases, the demand for money also increases. Therefore, if the market was in equilibrium, this increase in income generates an excess demand that has to be reestablished with an increase in the interest rate. In this case, starting from $Y_0 = 80$, $Y = 90$ is increased. Consequently, what happens in the money market is that a higher interest rate equal to $r_1$ is generated, that is, $r_0<r_1$. These two ordered pairs, the product of the increase in income ($Y$), can be seen at points $A$ and $B$ on the LM curve.

### III. DA-OA Model

#### 1) Equilibrium equation of $Y^e$, $r^e$ and $P^e$

First, we form the IS curve ($r$ as a function of $Y$).

<center> $[1 - b(1-t)]$$\bar{Y}$ $= C_0 + I_0 + G_0 - hr$

* Where: $\bar{Y}$ = $Y$

<center> $hr = C_0 + I_0 + G_0 - [1 - b(1-t)]Y$

<center> $IS:$ $r =$ $\dfrac{C_0 + I_0 + G_0}{h}$ - $\dfrac{[1 - b(1-t)]}{h}$$Y$

Second, we form the LM curve ($r$ as a function of $Y$).

<center> $\dfrac{M^s_0}{P_0}$ $= kY - jr$

<center> $jr$ = $kY -$ $\dfrac{M^s_0}{P_0}$

<center> $LM:$ $r = -$$\dfrac{M^s_0}{P_0}$ $\dfrac{1}{j}$ + $\dfrac{k}{j}$ $Y$

To find $Y^e$ and $r^e$ it is necessary to equalize both equations:

<center> $\dfrac{C_0 + I_0 + G_0}{h}$ $-$ $\dfrac{[1 - b(1-t)]}{h}$$Y$ $=$ $-$$\dfrac{M^s_0}{P_0}$ $\dfrac{1}{j}$ + $\dfrac{k}{j}$ $Y$

<center> $\dfrac{C_0 + I_0 + G_0}{h}$ $+$ $\dfrac{M^s_0}{P_0}$ $\dfrac{1}{j}$ $=$ $\dfrac{k}{j}$ $Y$ $+$ $\dfrac{[1 - b(1-t)]}{h}$

<center> $\dfrac{C_0 + I_0 + G_0}{h}$ $+$ $\dfrac{M^s_0}{P_0}$ $\dfrac{1}{j}$ $=$ $\dfrac{hk + j[1 - b(1-t)]}{jh}$ $Y$

<center> $j(C_0 + I_0 + G_0) + h$$\dfrac{M^s_0}{P_0}$ $= hk + j[1 - b(1-t)]Y$

<center> $Y^e =$ $\dfrac{j(C_0 + I_0 + G_0)}{hk + j[1 - b(1-t)]}$ $+$ $\dfrac{h}{hk + j[1 - b(1-t)]}$ $\dfrac{M^s_0}{P_0}$

To find the equilibrium $r$, we must proceed to replace $Y^e$ in the LM equation:

<center> $r = -$$\dfrac{M^s_0}{P_0}$ $\dfrac{1}{j}$ + $\dfrac{k}{j}$ $[\dfrac{j(C_0 + I_0 + G_0)}{hk + j[1 - b(1-t)]}$ $+$ $\dfrac{h}{hk + j[1 - b(1-t)]}$ $\dfrac{M^s_0}{P_0}]$

<center> $r = -$$\dfrac{M^s_0}{P_0}$ $\dfrac{1}{j}$ + $\dfrac{k(C_0 + I_0 + G_0)}{hk + j[1 - b(1-t)]}$ + $\dfrac{hk}{j[hk + j[1 - b(1-t)]]}$ $\dfrac{M^s_0}{P_0}$

<center> $r =$ $\dfrac{k(C_0 + I_0 + G_0)}{hk + j[1 - b(1-t)]}$ $-$ $\dfrac{M^s_0}{P_0}$ [$\dfrac{hk}{j[hk + j[1 - b(1-t)]]}$ $+$ $\dfrac{1}{j}$]

<center> $r^e =$ $\dfrac{k(C_0 + I_0 + G_0)}{hk + j[1 - b(1-t)]}$ $-$ $\dfrac{M^s_0}{P_0}$ [$\dfrac{hk}{hk + j[1 - b(1-t)]}$ $+1$]

Finally, to find $P^e$, $P$ is cleared as a function of $Y$ in the equation for $Y^e$.

<center> $Y =$ $\dfrac{j(C_0 + I_0 + G_0)}{hk + j[1 - b(1-t)]}$ $+$ $\dfrac{h}{hk + j[1 - b(1-t)]}$ $\dfrac{M^s}{P}$

* $B_1 = hk + j[1 - b(1-t)$
* $B_0 = C_0 + I_0 + G_0$

<center> $Y =$ $\dfrac{jB_0}{B_1}$ $+$ $\dfrac{h}{B_1}$ $\dfrac{M^s}{P}$

<center> $Y -$ $j$$\dfrac{B_0}{B_1}$ $= h$ $\dfrac{M^s}{P}$

<center> $B_1Y - jB_0 = h$ $\dfrac{M^s}{P}$

<center> $P^e =$ $\dfrac{hM^s}{B_1Y - jB_0}$

### IIII. IS-LM Model with exogenous interest set by the central bank

#### 1) Equilibrium equation of $Y$ and $M^s$

<center> $DA = [C_0 + b(1-t)Y] + I_0 - hr + tY - X_0 - mY$

In this model, to find the equilibrium income, what is needed is to equalize everything that is demanded in the economy to everything that has been produced: 


<center> $Y = DA$

<center> $Y = [C_0 + b(1-t)Y] + I_0 - hr + tY + X_0 - mY$

<center> $Y = tY - mY + b(1-t)Y + C_0 + I_0 + X_0 - hr$

<center> $Y = (t-m)Y + b(1-t)Y + C_0 + I_0 + X_0 - hr$

<center> $Y - [(t-m) + b(1-t)]Y = C_0 + I_0 + X_0 - hr$

<center> $Y[1 - [(t-m) + b(1-t)]] = C_0 + I_0 + X_0 - hr$

* Considering: $r = r_0$

<center> $Y^e =$ $\dfrac{1}{1 - [(t-m) + b(1-t)]}$ $C_0 + I_0 + X_0 - hr_0$ 

Equilibrium equation of monetary market:

<center> $\dfrac{M^s}{P}$ = $kY - jr$ 

*  Considering: $r = r_0$

<center> $\dfrac{M^s}{P}$ = $- jr_0 + kY$ 

#### 2) Equilibrium in the IS-LM (graphic)

```python
Y_size = 100

C0 = 20   #consumo autónomo
I0 = 10   #inversión autónoma
X0 = 2    #exportacion
b = 0.4   #propension marginal a consumir
m = 0.2   #propension a importar
t = 0.4   #tasa impositiva
h = 2     #constante de decisión a invertir
r = 5   #tasa de interés real

B2 = (t-m) + b*(1-t)
B3 = C0 + I0 + X0

Y = np.arange(Y_size)
```

```python
def r_IS3(C0, b, m, t, I0, X0, h, Y):
    r_IS3 = B3/h - ((1-B2)*Y)/h
    return r_IS3

def Y_IS3(b, m, t, I0, X0, h, r):
    Y_IS3 = (B3 - h*r)/1-B2
    return Y_IS3

r_IS3 = r_IS3(C0, b, m, t, I0, X0, h, Y)
```

```python
y_max = np.max(r)
                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Derivation of the IS curve from equilibrium $Y=DA$", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r_IS3, label = "IS", color="purple")
ax.hlines(y = 5, xmin = 0, xmax = 100, label = "LM = $r_0$", color ="orange")

plt.axvline(x=39.7,  ymin= 0, ymax= 0.59, linestyle = ":", color = "gray")
plt.axhline(y=5, xmin= 0, xmax= 0.40, linestyle = ":", color = "gray")

plt.plot(39.7, 5, marker="o", color="black")
plt.text(39.8, 5.5, 'Eo', fontsize = 14, color = 'black')

plt.text(39, -16, '$Y^e$', fontsize = 16, color = 'black')
plt.text(-18, 5, '$r_0 = r^e$', fontsize = 16, color = 'black')

ax.legend()                                  
plt.show()
```

#### 3) Effects of a Contractive Fiscal Policy ($ΔG < 0$)

Considering ($G = tY$):
* $t_0 = 0.4$
* $t_1 = 0.1$

```python
Y_size = 100

C0 = 20   #consumo autónomo
I0 = 10   #inversión autónoma
X0 = 2    #exportacion
b = 0.4   #propension marginal a consumir
m = 0.2   #propension a importar
t = 0.1   #tasa impositiva
h = 2     #constante de decisión a invertir
r = 5   #tasa de interés real

B2 = (t-m) + b*(1-t)
B3 = C0 + I0 + X0

Y = np.arange(Y_size)
```

```python
def r_IS4(C0, b, m, t, I0, X0, h, Y):
    r_IS4 = B3/h - ((1-B2)*Y)/h
    return r_IS4

def Y_IS4(b, m, t, I0, X0, h, r):
    Y_IS4 = (B3 - h*r)/1-B2
    return Y_IS4

r_IS4 = r_IS4(C0, b, m, t, I0, X0, h, Y)
```

```python
y_max = np.max(r)
                        
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="Equilibrium in the IS-LM after ΔG($↓t$)", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r_IS3, label = "IS", color="purple")
ax.plot(Y, r_IS4, label = "IS2", linestyle = "--", color="purple")
ax.hlines(y = 5, xmin = 0, xmax = 100, label = "LM = $r_0$", color ="orange")

plt.axvline(x=39.7,  ymin= 0, ymax= 0.69, linestyle = ":", color = "gray")
plt.axhline(y=5, xmin= 0, xmax= 0.40, linestyle = ":", color = "gray")

plt.plot(39.7, 5, marker="o", color="black")
plt.text(39.8, 5.5, 'Eo', fontsize = 14, color = 'black')

plt.text(39, -26, '$Y^e$', fontsize = 16, color = 'black')
plt.text(-23, 5, '$r_0 = r_1 = r^e$', fontsize = 16, color = 'black')

plt.axvline(x=30,  ymin= 0, ymax= 0.69, linestyle = ":", color = "gray")
plt.plot(30, 5, marker="o", color="black")
plt.text(30.5, 5.5, 'E1', fontsize = 14, color = 'black')
plt.text(28.5, -25.8, '$Y^e_1$', fontsize = 16, color = 'black')

plt.text(33, -10, '←', fontsize = 22, color = 'gray')
plt.text(78, -10, '↙', fontsize = 22, color = 'firebrick')
plt.text(82, -12, 'ΔG($↓t$)', fontsize = 14, color = 'black')

ax.legend()                                  
plt.show()
```

* Math

<center> $ΔY =$ $\dfrac{C_0 + I_0 + X_0 - hr_0}{(1-b)Δt}$ $< 0$

* Intuition

<center> $↓t ⇢ ↓G ⇢ ↓DA ⇢ DA<Y ⇢ ↓Y$

<center> $↓Y ⇢ ↓M^d ⇢ M^d<$$\dfrac{M^s}{P}$ ⇢ ↓$\dfrac{M^s}{P}$ / considering $r=r_0$

Considering that $G=tY$, a decrease in $G$ produces a fall in aggregate demand. Consequently, an excess is generated in the production level, so $(Y_1<Y_0)$ has to be adjusted and reduced.

On the other hand, the fall in $Y$ causes the demand for money to decrease as well. Therefore, an excess supply is generated in the money market. Since the interest rate is already fixed by the Central Reserve Bank, it cannot be used as a tool to balance the market. Consequently, what is reduced is the real supply of money to maintain equilibrium.

* Math

<center> $ΔY =$ $\dfrac{-h}{1 -[(t-m) + b(1-t)]}$ $Δr > 0$

<center> $ΔM^s = -jΔr > 0$

* Intuition

<center> $↓r ⇢ ↑I ⇢ ↑DA ⇢ DA>Y ⇢ ↑Y$

<center> $↑Y ⇢ ↑M^d ⇢ M^d>$$\dfrac{M^s}{P}$ ⇢ ↑$\dfrac{M^s}{P}$ / considering $r=r_0$

On the one hand, the fall in the interest rate generates an increase in investment. Consequently, aggregate demand increases, which generates production shortage. To compensate for this imbalance, the level of production is adjusted and $(Y_0<Y_1)$ increases.

On the other hand, the increase in $Y$ causes the demand for money to increase as well. Therefore, an excess demand is generated in the money market. Since the interest rate is already fixed by the Central Reserve Bank, it cannot be used as a tool to balance the market. Consequently, what is increased is the real supply of money to maintain equilibrium.
