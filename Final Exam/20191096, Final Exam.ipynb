#### 1.1 Find the Equilibrium equation for the Aggregate Income. Please, write all the equations in steps.

The equilibrium equation for the Aggregate Income is derived form the equilibrium condition where income equals aggregate demand: $ DA = Y $

First, the aggregate demand $DA$, is defined as follows:

$$ DA = C + I + G + X - M $$

Where:
$$ C = C_0 + bY^d $$
$$ I = I_0 - hr $$
$$ G = G_0 $$
$$ T = tY $$
$$ X = X_0 $$
$$ M = mY^d $$

Therefore, substituting the values, the Aggregate Demand function is obtained

$$ DA = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $$

$$ DA = (b - m)(1 - t)Y + C_0 + I_0 -hr + G_0 + X_0 $$

$$ DA = (C_0 + I_0 + G_0 + X_0 -hr) + (b - m)(1 - t)Y $$

In short form:

$$ DA = α_0 + α_1Y $$

Where $ α_0 = (C_0 + I_0 + G_0 + X_0 -hr)$ is the intercept and $ α_1 = (b - m)(1 - t) $ is the function slope

Second, considering the equilibrium condition $Y = DA$, the short-term equilibrium income or product equation is obtained

$$ Y = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $$

$$ Y = (C_0 + I_0 + G_0 + X_0 - hr) + (b - m)(1 - t)Y $$

$$ [1 - (b-m)(1 - t)]Y = C_0 + I_0 + G_0 + X_0 - hr $$

Finally, the equilibrium equation for the Aggregate Income is:

$$ Y = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

#### 2. Show the derivation of the LM curve from equilibrium $ M^s = M^d $. Explain using intuition, Math, and a Figure

The equilibrium in the Money Market is derived from the equilibrium between Money Supply $ (M^o) $ and Money Demand $ (M^d) $: 

$$ M^o = M^d $$

Then the equilibrium equation in the money market, expressed in real terms, is:

$$ \frac{M^s}{P} = kY - ji $$

Where $ M^s $ is the nominal money supply, $ P $ the price level, $ k $ the transaction money demand, $ j $ the speculation money demand and $ i $ is the nominal interest rate. It must be noted that $ Y $ or the income is given.

Expected inflation can be assumed to be zero, since in the short run the price level is fixed and exogenous. Therefore, there would not be a big difference between the nominal $ (i) $ and the real interest rate $ (r) $.

The equation and the graph of equilibrium in the money market are:

$$ \frac{M^s}{P} = kY - jr $$

Following the last equilibrium equation, the money market interest rate $ (r) $ is equal to:

$$ jr = \frac{M^s_o}{P_o} - kY $$

$$ r = -\frac{1}{j} \frac{M^s_o}{P_o} + \frac{k}{j}Y $$


Where the slope, $ \frac{k}{j} $ , is positive and the intercept, $ \frac{M^s_o}{j P_o} $ , is negative.

## DA-OA Model

#### 1. Find the equilibrium equation of $Y^e$, $P^e$ and $r^e$

Considering the equations of the Aggregate Demand and the Aggregate Supply:

$$ IS: [1 - b(1 - t)]\bar{Y} = C_0 + I_0 + G_0 -hr $$

$$ LM: \frac{Mo^s}{P_0} = k\bar{Y} - jr $$

Both, in function of the interest rate $(r)$:

$$ IS: r = \frac{B_0}{h} - \frac{B_1}{h}Y $$

Where $ B_0 = C_0 + G_0 + I_0 + X_0  $ and $  B_1 = 1 - (b - m)(1 - t) $

And

$$ LM : r = \frac{-1}{j} \frac{Mo^s}{P_0} + \frac{k}{j}Y $$

For $Y^e$, equalize both equations:

$$ \frac{-1}{j} \frac{Mo^s}{P_0}+ \frac{k}{j}Y = \frac{B_0}{h} - \frac{B_1}{h}Y $$

$$ Y^e = \frac{jB_0}{kh + jB_1} + (\frac{h}{kh + jB_1}) \frac{Mo^s}{P_0} $$

For $r^e$:

$$ r^e = \frac{kB_0}{kh + jB_1} - \frac{B_1}{kh + jB_1} \frac{Mo^s}{P_0} $$

Then, $r^e$ linearized:

$$ r^e = -\frac{Mo^s - P}{j} + \frac{k}{j}Y $$

Last, $P^e$:

$$ P^e = \frac{hMo^s + jB_0}{h} - \frac{jB_1 + hk}{h}Y $$

### IS-LM Model with exogenous interest set by the central bank

#### 1. Now you have to use the model below. But now the interest rate $r$ and $p$ are exogenous. Because $r$ is the instrument of monetary policy. The endogenous variables of this model are $Y$ and $Mo^s$

#### 1.1 Find the equilibrium equation of $Y$ and $Mo^s$ 

For $Y$, considering that

$$ Y = C + I + G + X - M $$

And that this components are defined by:

$$ C = C_0 + b(1-t)Y $$
$$ I = I_0 - hr $$
$$ G = tY $$
$$ X = X_0 $$
$$ M = mY $$

Substituting in each component

$$ Y = C_0 + b(1-t)Y + I_0 - hr + G = tY + X = X_0 + M = mY $$

$$ Y - b(1-t)Y - tY + mY = C_o + I_o + X_o - hr $$

$$ Y(1 - b(1-t) - t + m) = C_o + I_o + X_o - hr $$

The equilibrium equation for $Y$ is:

$$ Y = \frac{1}{1 - b(1-t) - t + m} C_o + I_o + X_o - hr $$

And, for $Mo^s$, considering that

$$ \frac{Mo^s}{P_0} = kY - jr $$

Then, the equilibrium equation for $Mo^s$ is:
