**1) In the Equilibrium equation, the aggregate income ($Y$) must be equal to the aggregate demand ($AD$):** 

$$Y = AD$$

$$Y = (C_0 + I_0 + G_0 + X_0 - hr) + (b - m)(1 - t)Y$$

$$Y - (b - m)(1 - t)Y = (C_0 + I_0 + G_0 + X_0 - hr)$$

$$Y [1-(b-m)(1-t)]=(C_0 + I_0 + G_0 + X_0 - hr)$$

$$Y* = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr)$$

Where:


$$\frac{1}{1 - (b - m)(1 - t)} = k$$ 


and $k$ is the keynesian multiplier

## IS-LM

**1) Show the derivation of the IS curve from equilibrium Y=DA. Explain using intuition, Math, and a Figure.**

**Math:**

$$ Y = C + I + G + X − M$$

Where:

$C$ = household consumption

$I$ = private investment

$G$ = level of government spending

$E$ = exports

$M$ = imports

Each of the elements that equals to $Y$ has its own functions:

- $C = C0 + bY^d$

- $I = I0 - h*r$
    - $I0$: Investment intercept
    
- $G = G0$

- $T = tY$

- $X = X0$

- $M = mY^d$

$$Y - T = C + I - T + G + X - M $$

$$Y^d = C + I - T + G + X - M$$

This equality can be rewritten as follows:

$$(Y^d - C) + (T - G) + (M - X) = I$$

Then the private sector ($S_p$), the public sector ($S_g$) and the external sector ($S_e$) can be put on a equation: 

$$S_p + S_g + S_e = I_0 - hr$$ , this equals to $$S(Y) = I(r)$$

Each of these sector also has its own funtions:

- $S_p = Y^d - (C_0 + bY^d)$

- $S_g = T - G$

- $S_e = mY^d -  X_0$

Finally, 

$$S = I_0 - hr$$

The equality between saving and investment is obtained:

$$S(Y) = I(r)$$

Then we get:

$$(Y^d - C_0 - bY^d) + (T - G_0) + (mY^d - X_0) = I_0 - hr$$

The equation of the IS curve is obtained: 

$$[1-(b-m)(1-t)]Y-(C_0+G_0+X_0) = I_0 - hr$$

We obtain an equivalence for the equilibrium function: 

$$r = \frac{1}{h}(C_0 + G_0 + I_0 + X_0) - \frac{1-(b-m)(1-T)}{h}Y$$

Wich also means:

$$r=\frac{\beta_0}{h}-\frac{\beta_1}{h}Y$$

Where:

$$\beta = C_0 + G_0 + I_0 + X_0$$

and 

$$\beta_1 = 1-(b-m)(1-t)$$

**Figure:**

```python
# Parameters
Y_size = 100
C0 = 20                 # Autonomous consumption
b = 0.2                # Marginal propensity to consume
alpha = 5              # Autonomous imports
m = 0.2            # Marginal propensity to import
T = 1                  # Taxes
I0 = 10             # Investment intercept (when r = 0)
G0 = 8              # Government spending
X0 = 2              # Exports (given)
h = 5                  # Investment slope wrt to i
# Arrays
Y = np.arange(Y_size)  # Array values of Y

"3|DEFINE AND POPULATE THE IS-SCHEDULE"
def r_IS(C0, alpha, b, m, T, I0, G0, X0, h, Y):
    r_IS = ((C0 + I0 + G0 + X0) - (1-(b-m)*(1-T))*Y)/h
    return r_IS

def Y_IS(Co, alpha, b, m, T, I0, G0, X0, h, r):
    Y_IS = (C0 + I0 + G0 + X0)/1-(b-m)*(1-T) - h*r/(1-(b-m)*(1-T))
    return Y_IS

r = r_IS(C0, alpha, b, m, T, I0, G0, X0, h, Y)

"4|PLOT THE IS-SCHEDULE"
y_max = np.max(r)
x_max = Y_IS(C0, alpha, b, m, T, I0, G0, X0, h, 0)

v = [0, x_max, 0, y_max]                        # Set the axes range
fig, ax = plt.subplots(figsize=(10, 8))
ax.set(title="IS SCHEDULE", xlabel=r'Y', ylabel=r'r')
ax.plot(Y, r, "k-")
ax.yaxis.set_major_locator(plt.NullLocator())   # Hide ticks
ax.xaxis.set_major_locator(plt.NullLocator())   # Hide ticks
plt.axis(v)                                     # Use 'v' as the axes range
plt.show()
```

## DA-OA Model

**1) Find the equilibrium equation of $Y^e$, $P^e$ and $r^e$.**

**For $Y^e$:**

$$\frac{C_0 + I_0+ G_0}{h} - \frac{[1-b(1 - t)]Y}{h} = \frac{kY}{j} - \frac{M_0^s}{P_0 j}$$  
  
$$\frac{C_0 + I_0+ G_0}{h} + \frac{M_0^s}{P_0 j} = \frac{kY}{j} + \frac{[1-b(1 - t)]Y}{h}$$  
  
$$\frac{C_0 + I_0+ G_0}{h} + \frac{M_0^s}{P_0 j} =  Y  (\frac{k}{j} + \frac{1-b(1 - t)}{h}$$  
  
$$\frac{C_0 + I_0+ G_0}{h} + \frac{M_0^s}{P_0 j} = Y  (\frac{kh + [1-b(1 - t)]j}{hj})$$  
  
$$(\frac{C_0 + I_0+ G_0}{h} + \frac{M_0^s}{P_0 j})  (\frac{hj}{kh + [1-b(1 - t)]j}) = Y$$  
  
$$Y = \frac{(C_0 + I_0+ G_0)hj}{h(kh + [1-b(1 - t)]j} +  \frac{M_0^s h}{P_0 kh + [1-b(1 - t)]j}$$  
  
$$Y^e= \frac{(C_0 + I_0+ G_0)j}{(kh + [1-b(1 - t)]j} + \frac{h}{kh + [1-b(1 - t)]j}  \frac{M_0^s}{P_0^s}$$

$$Y^e =\frac{j \beta_0}{kh + j\beta_1} + (\frac{h}{kh + j\beta_1})\frac{M^s_0}{P_0}$$



**For $P^e$:**

IS: $r = \frac{\beta_0}{h} - \frac{\beta_1}{h}Y$

Where:

$$\beta_0 = C_0 + I_0 + G_0 + X_0$$

$$\beta_1 = 1 - (b - m) (1 - t)$$

LM: $r = -\frac{1}{j}\frac{M^s_0}{P_0} + \frac{k}{j}Y$

$$\frac{\beta_0}{h} - \frac{\beta_1 Y}{h} = \frac{kY}{j} - \frac{M^s_0}{P_0 j}$$

$$\frac{\beta_0}{h}-\frac{\beta_1 Y}{h}-\frac{kY}{j}=-\frac{1}{P_0}\frac{M^s_0}{j}$$

$$\frac{\beta_0}{h}-\frac{\beta_1 Y}{h}-\frac{kY}{j}\frac{j}{M^s_0}=-\frac{1}{P_0}$$

$$\frac{(\beta_0 - \beta_1Y)j - khY}{hj}\frac{j}{M_0^s} = - \frac{1}{P_0}$$ 

$$- P =  \frac{M_0^s h}{j\beta_0 - (j\beta_1 + hk)Y}$$  

$$P^e = \frac{M^s_0 h}{-j\beta_0 + (j \beta_1 + hk)Y}$$





**For $r^e$:**

$$r^e =\frac{k \beta_0}{kh + j\beta_1} - \frac{\beta_1}{kh + j\beta_1}\frac{M^s_0}{P_0}$$
